## Problem 1

Professor Per-Olof Persson's `dgconvect0` function works very well, but unfortunately, it only allows us to use Legendre basis functions that are up to 2nd-order (i.e. at most up to $p=2$ out of the Legendre polynomials) on each element. Furthermore, when it plots the solution, it only plots these basis functions at the $p$ nodes used within each element, which makes our plot look rough and jagged, and even less smooth than we already have with the discontinuities. Finally, even the nodes picked for each element are equidistant, which means that Runge's phenomenon will be quite prevalent if we want to implement a local higher-order Legendre polynomial basis on each element. We implement a few improvements to address these issues.

**Part (a)**

We do not want to run into overdetermined or singular matrices when we try to find the local polynomial basis (we could technically deal with the former case via linear least squares, but why do so when we could just use less nodes and evaluate the resultant polynomial we get from those node values on more nodes?). Thus, we still assume that we have $p$ nodes for each element. However, equidistant nodes are prone to exhibit Runge's phenomenon, where higher-degree polynomial interpolation actually leads to a cascading error that tends to start near the edges of an interpolation interval. This occurs because of two reasons: generally, for polynomial interpolation, the $n$th derivatives will grow quickly, and the equidistance leads to a constant which grows with a product of the distances between a given interpolation point and the other nodes. 

Chebyshev nodes remedy this issue, because choosing these nodes effectively minimizes this error bound introduced via polynomial interpolation by getting rid of this product term. (These nodes also happen to be the roots of the $k$th-degree Chebyshev polynomial of the first kind, for $k$ nodes.) The usual error bound for $p$th degree polynomial interpolation with equidistant nodes in an interval $[a,b]$ is

$$
\left|\frac{f^\left(p+1\right)\left(\xi\right)}{(p+1)!}\prod_{i=0}^{p}\left(\xi-x_{i}\right)\right|
$$

for some $\xi\in[a,b]$. However, if the nodes we pick are the Chebyshev nodes, the error bound is 

$$
\left|\frac{f^\left(p+1\right)\left(\xi\right)}{2^{p}(p+1)!}\right|\quad(1)
$$

for $p$ and $\xi$ defined likewise. However, even though interpolating with these nodes is in a sense optimal, we will still run into an issue if the derivatives of the function we want to interpolate grow too quickly, i.e. quicker than $2^{p}(p+1)!$, showing that our interpolant may still exhibit Runge's phenomenon in these cases. 

***1.***

In our case, we see that our solution looks similar in shape to the Runge function, a function notorious for exhibiting Runge's phenomenon quite prevalently. Thus, we should especially use Chebyshev nodes (which we will see later do not necessarily save us from Runge's phenomenon taking over if $p$ is too large). For the interval $[-1,1]$ with $p$ nodes, these nodes occur at $s_{i}=\cos\left(\pi i/p\right)$ for $i=0,\dots,p$. Since we want to input these for each element, we also scale and translate these to $[0,h]$ using an affine transformation of the form

$$
x_{i}=\frac{1}{2}\left(a+b\right)+\frac{1}{2}\left(b-a\right)s_{i},
$$

where in particular, for our case, $a=0$ and $b=h$. Then, we put these in increasing order. 

We test this out in code below, and compare with the original equidistant nodes created in `dgconvect0` to make sure that our new nodes are in the correct order and lie in the interval $[0,h]$.

```julia
n = 20
p = 5
h = 1/n
i = collect(0 : p)
s = @. cos.(pi * i / p) #Chebyshev nodes from 1 to -1
s = reverse(h * (s .+ 1) / 2) #Chebyshev nodes shifted and scaled to [0, h]
println(s)

s = @. collect(0:p) / p
println(s * h)
```

Sure enough, for our test case of $h=1/20=0.05$ and $p=5$, both arrays contained $p+1=6$ nodes and went from $0$ to $h=0.05$.

***2.***

Next, we need to implement support for higher-order Legendre polynomial interpolants on each element. To start off, we load in the `gauss_quad` and `legendre_poly` utility functions from Math 228B course webpage. `gauss_quad` returns `p` quadrature nodes and their corresponding weights in the interval $[-1,1]$ (which we must shift if we want to evaluate an integral over some arbitrary interval), while `legendre_poly` evaluates the Legendre polynomials up to the `p`th order at the points specified in vector `x`.

```julia
using LinearAlgebra

function gauss_quad(p)
    n = ceil((p + 1) / 2)
    b = 1 : n - 1
    b = @. b / sqrt(4 * b ^ 2 - 1)
    eval, evec = eigen(diagm(1 => b, -1 => b))
    return eval, 2 * evec[1, :] .^ 2
end

function legendre_poly(x, p)
    z = zeros(size(x))
    o = ones(size(x))
    y = hcat(o, x, repeat(z, 1, p - 1))
    dy = hcat(z, o, repeat(z, 1, p - 1))
    for i = 1 : p - 1
        @. y[:,i+2] = ((2i+1)*x*y[:,i+1] - i*y[:,i]) / (i+1)
        @. dy[:,i+2] = ((2i+1)*(x*dy[:,i+1] + y[:,i+1]) - i*dy[:,i]) / (i+1)
    end
    y, dy
end
```

From here, how can we generate a mass and stiffness matrix for an arbitrary Legendre polynomial basis on each element? To start off, we note that nodal basis functions are endowed with the property that $\varphi_{i}\left(s_{k}\right)=\delta_{ij}$, where each basis function $\varphi_{i}$ can be written as a linear combination of Legendre polynomials:

$$
\varphi_{i}\left(s\right)=\sum_{j=0}^{p}{c_{i}^{j}P_{j}(s)}.
$$

Therefore, this gives us a linear system of equations:

$$
V\left(s_{0},\dots,s_{p}\right)C:=\begin{bmatrix} P_{0}\left(s_{0}\right) & P_{1}\left(s_{0}\right) & \cdots & P_{p}\left(s_{0}\right) \\ P_{0}\left(s_{1}\right) & P_{1}\left(s_{1}\right) & \cdots & P_{p}\left(s_{1}\right) \\ \vdots & \vdots & \ddots & \vdots \\ P_{0}\left(s_{p}\right) & P_{1}\left(s_{p}\right) & \cdots & P_{p}\left(s_{p}\right) \end{bmatrix}\begin{bmatrix} c_{0}^{0} & c_{1}^{0} & \cdots & c_{p}^{0} \\ c_{0}^{1} & c_{1}^{1} & \cdots & c_{p}^{1} \\ \vdots & \vdots & \ddots & \vdots \\ c_{0}^{p} & c_{1}^{p} & \cdots & c_{p}^{p} \end{bmatrix}=I_{p},
$$

where $I_{p}$ is the $p\times p$ identity matrix. Thus, effectively, each $\varphi_{i}$ lies in the dual to our node space. From here, we can evaluate matrix $V$ on the Chebyshev nodes in each element, and by inverting this matrix, we get our matrix of coefficients for our basis functions, $C$.

Fortunately, `legendre_poly` gives us the matrix $V$ exactly if the input vector is equal to `x`=`[s_{0},...,s_{p}]`. Note that, to compute the entries of the elementary matrices $M$ and $C$ (which are `Mel` and `Kel` in our code, respectively), we need to evaluate the integrals

$$
M_{ij}=\int_{0}^{h}{\varphi_{i}\varphi_{j}ds},\quad(2)\quad C_{ij}=\int_{0}^{h}{\varphi'_{i}\varphi_{j}ds},\quad(3)
$$

which can be done via $k$th-order Gauss-Legendre quadrature, which is exact for $k\geq 2p$. Namely, for some $k$th-order polynomial $F(s)$, the Gauss-Legendre quadrature on the standard interval $[-1,1]$ can be evaluated as 

$$
\int_{-1}^{1}{F\left(s\right)ds}=\sum_{i=1}^{k}{w_{i}F\left(s_{i}\right)},
$$

for quadrature points $\left(s_{i}\right)$ and weights $w_{i}$ given by utility function `gauss_quad`. Hence, to evaluate an integral over the interval $[0,h]$, we need to utilize a change of coordinates. In general, 

$$
\int_{a}^{b}{F\left(t\right)dt}=\frac{b-a}{2}\int_{-1}^{1}{F\left(\frac{b-a}{2}s+\frac{b+a}{2}\right)ds},
$$

and therefore, letting $a=0$ and $b=h$, 

$$
\int_{0}^{h}{F\left(t\right)dt}=\frac{h}{2}\int_{-1}^{1}{F\left(\frac{h}{2}s+\frac{h}{2}\right)ds}=\frac{h}{2}\sum_{i=1}^{k}{w_{i}F\left(\frac{h}{2}s_{i}+\frac{h}{2}\right)}.\quad(4)
$$

To use $(4)$ to evaluate $(2)$ and $(3)$, we must evaluate the basis functions $\varphi_{i}$ at the quadrature points shown in $(4)$. We can do this by merely evaluating the Legendre polynomials $P$ at these points using `legendre_poly`, then evaluate the nodal basis functions at these points by evaluating the product $PC$, which multiples the evaluated Legendre polynomials by the necessary coefficients to give us our basis. Then, $(2)$ and $(3)$ can be evaluated with ease, giving us `Mel` and `Kel`.

***3.***

We also would like to make our solution plot each time look smoother by evaluating our solution polynomials at $3p$ nodes per element, rather than just $p$. Furthermore, since we want this plot to look even, these $3p$ nodes should be equidistant, as opposed to the $p$ Chebyshev nodes used for finding the basis functions. To accomplish this, we note that the numerical solution $u_{h}$, regardless of what the final solution looks like, can still be written in terms of a Legendre polynomial basis on each element, i.e. for the $k$th element, $u_{h,k}(x)=\sum_{i=1}^{p}{\hat{u}_{i}^{k}\varphi_{i}^{k}(x)}$. Therefore, to evaluate our solution at three times as many nodes, for *each* element $k$, we extract the values of $u_{h}$ and the $p$ nodes that correspond to each of these values. Then, we evaluate the Legendre polynomials at these $p$ nodes, and by solving the system of equations

$$
P\left(\mathbf{s}_{p}\right)\hat{\mathbf{u}}:=\begin{bmatrix} P_{0}\left(s_{0}\right) & P_{1}\left(s_{0}\right) & \cdots & P_{p}\left(s_{0}\right) \\ P_{0}\left(s_{1}\right) & P_{1}\left(s_{1}\right) & \cdots & P_{p}\left(s_{1}\right) \\ \vdots & \vdots & \ddots & \vdots \\ P_{0}\left(s_{p}\right) & P_{1}\left(s_{p}\right) & \cdots & P_{p}\left(s_{p}\right) \end{bmatrix}\begin{bmatrix} \hat{u}_{1}^{k} \\ \hat{u}_{2}^{k} \\ \vdots \\ \hat{u}_{k}^{k} \end{bmatrix}=\begin{bmatrix} u_{1}^{k} \\ u_{2}^{k} \\ \vdots \\  u_{k}^{k} \end{bmatrix}:=\mathbf{u}_{k}
$$

for $\hat{\mathbf{u}}$, we get the coefficients $\hat{u}_{i}^{k}$ for our local Legendre polynomial basis. From here, we reevaluate $P$ at the $3p$ equidistant nodes by calculating $P\left(\mathbf{s}_{3p}\right)$ using `legendre_poly` as usual, then evaluating the product $P\left(\mathbf{s}_{3p}\right)\hat{\mathbf{u}}$. This returns $u_{h}$ at our new nodes.

***4.***

Finally, instead of using the discrete max-norm error $\left||u-u_{h}|\right|_{\inf}=\max_{x\in[0,1]}{\left|u(x)-u_{h}(x)\right|}$, from the theory for FEM methods, we know that the error tends to depend more directly on the $L_{2}$-norm error, $\left||u-u_{h}|\right|_{2}=\left(\int_{0}^{1}{\left(u(x)-u_{h}(x)\right)^{2}dx}\right)^{1/2}$. Since we have a piecewise polynomial solution for $u_{h}$ and an exact solution of $u$ to compare with, we can in fact compute this error with arbitrarily high accuracy and precision, and do not need to resort to some grid approximation. Hence, in a similar vein as we did for part (c), we evaluate $\int_{x_{k-1}}^{x_{k}}{\left(u(x)-u_{h}(x)\right)^{2}dx}$ on each $k$th element via Gauss-Legendre quadrature to some arbitrarily high order (we use $100$th order GL quadrature in our code), then add up the contributions over all of the $n$ elements. Then, adding up these integrals and square-rooting the result, we can compute $\left||u-u_{h}|\right|_{2}$.

```julia
using LinearAlgebra
using PyPlot

function dgconvect(; n = 10, p = 1, T = 1.0, dt = 1e-3)
    # Discretization
    h = 1 / n
    i = collect(0 : p)
    s = @. cos.(pi * i / p) #Chebyshev nodes from 1 to -1
    s = reverse(h * (s .+ 1) / 2) #Chebyshev nodes shifted and scaled to [0, h]
    x = @. s + (0 : h : 1 - h)'

    # Gaussian initial condition (and exact solution if shifted)
    uinit(x) = exp(-(x - 0.5) ^ 2 / 0.1 ^ 2)

    gx, gw = gauss_quad(2 * p)
    V, dV = legendre_poly(s, p) #Legendre polynomials evaluated at Chebyshev nodes
    C = V \ I #coefficients for basis
    sq = h * (gx .+ 1) / 2 #quadrature points (after change of interval)
    P, dP = legendre_poly(sq, p) #Legendre polynomials and their derivatives evaluated at quadrature points
    phi = P * C #basis functions evaluated at quadrature points
    dphi = dP * C #derivative of basis functions evaluated at quadrature points
    M = zeros(p + 1, p + 1)
    K = zeros(p + 1, p + 1)
    for i = 1 : p + 1 #construct elementary matrices
        for j = 1 : p + 1
            M[i, j] = dot(phi[:, i] .* phi[:, j], gw) 
            K[i, j] = dot(dphi[:, i] .* phi[:, j], gw) 
        end
    end
    Mel = (h / 2) * Symmetric(M) #multiply by Jacobian constant
    Kel = (h / 2) * K #""
        
    # RHS function in semi-discrete system
    function rhs(u)
        r = Kel * u
        r[end,:] = r[end,:] - u[end,:]
        r[1,:] = r[1,:] + u[end, [end; 1:end-1]]
        r = Mel \ r
    end

    # Setup
    splot = @. collect(0 : 3 * p) / (3 * p) #vector for 3p local nodes
    xplot = @. h * splot + (0 : h : 1 - h)'
    uplot = uinit.(xplot)
    u = uinit.(x)
    nsteps = round(Int, T / dt)

    # Setup plotting
    xx = collect(0 : 0.01 : 1) # For exact solution
    clf(); axis([0, 1, -0.1, 1.1]); grid(true);
    ph1 = plot(xplot, uplot, "b")   # Many line object (since discontinuous)
    ph2, = plot(xx, uinit.(xx), "k") # Single line object
    
    # Main loop
    for it = 1 : nsteps
        # Runge-Kutta 4
        k1 = dt * rhs(u)
        k2 = dt * rhs(u + k1 / 2)
        k3 = dt * rhs(u + k2 / 2)
        k4 = dt * rhs(u + k3)
        u += (k1 + 2 * k2 + 2 * k3 + k4) / 6

        # Plotting
        if mod(it, round(nsteps / 100)) == 0 #plot final solution.
            uexact = @. uinit(mod(xx - dt * it, 1.0))
            for i = 1 : n #loop through each element
                uloc0 = u[:, i] #extract values of u at local nodes on the element
                xloc0 = x[:, i] #extract nodes in this element
                Ploc0, ~ = legendre_poly(xloc0, p) #evaluate Legendre polynomials at local nodes
                coeff = Ploc0 \ uloc0 #calculate coefficients of local Legendre polynomial basis
                #define new nodes, such that there are 3p of them per element (i=0,...,3p, where i=3p overlaps with the next)
                xlocf = xloc0[1] .+ (collect(0 : 3 * p) * (xloc0[end] - xloc0[1]) / (3 * p)) 
                Plocf, ~ = legendre_poly(xlocf, p) #evaluate Legendre polynomials at the 3p local nodes
                uplot[:, i] = Plocf * coeff #find the values of u at the 3p local nodes
            end
            for line in ph1 line.remove(); end
            ph1 = plot(xplot, uplot, "b")
            ph2.set_data(xx, uexact)
            gca().legend((L"u\left(x,t\right)", L"u_{h}\left(x,t\right)"))
            pause(1e-3)
        end
    end

    #calculate L2 norm 
    int = 0
    for i = 1 : n #integrate using Gauss-Legendre quadrature of arbitrarily high order
        uloc0 = u[:, i] #extract values of u at local nodes on the element
        xloc0 = x[:, i] #extract nodes in this element
        gx, gw = gauss_quad(100)
        Ploc0, ~ = legendre_poly(xloc0, p) #evaluate Legendre polynomials at local nodes
        coeff = Ploc0 \ uloc0 #calculate coefficients of local Legendre polynomial basis
        #define new nodes, such that there are 3p of them
        gxq = 0.5 * ((xloc0[end] - xloc0[1]) * gx .+ xloc0[1] .+ xloc0[end])
        Plocq, ~ = legendre_poly(gxq, p) #evaluate Legendre polynomials at quadrature points
        ulocq = Plocq * coeff #find the values of u at quadrature points
        uexactlocq = @. uinit(mod(gxq - T, 1.0)) #exact solution at quadrature points
        int += 0.5 * (xloc0[end] - xloc0[1]) * dot((uexactlocq - ulocq) .^ 2, gw) 
        #add contribution of this element to the integral
    end
    error = sqrt(int)  #calculate L2 norm error
    return u, error
end
```

As we can see, `dgconvect` behaves exactly as desired, and we can increase the order `p` of our piecewise Legendre polynomial Sobolev space and our number of elements `n` arbitrarily. However, looking back at our error estimate $(1)$ for using Chebyshev nodes, we observe that, while increasing the number of elements usually tends to decrease the error, if our derivatives of the actual solution, $u$, become too large for our given choice of $p$, this might still lead to Runge's phenomenon. For example, if we use $5$ elements and a $12$th-degree local Legendre polynomial basis, we certainly have problems.

```julia
u, error = dgconvect(p = 12, n = 5)
error
```

The denominator of our error bound for $p=12$ is $2^{12}(12+1)!\approx 2.55059\times 10^{13}$, so how could the $13$th derivative of $u$, our actual solution, possibly become that large in the numerator? We note that the $13$th derivative of $u(x,0)=e^{-(x-0.5)^{2}/{0.1^{2}}}$ is the monstrosity

$$
u^{(13)}(x,0)\approx u\left(x,0\right)\left(-8.192\times 10^{29}x^{13}+5.3248\times 10^{30}x^{12}-1.56549\times 10^{31}x^{11}+2.75292\times 10^{31}x^{10}-3.2259\times 10^{31}x^{9}+2.65554\times 10^{31}x^{8}-1.57681\times 10^{31}x^{7}+6.82183\times 10^{30}x^{6}-2.14427\times 10^{30}x^{5}+4.81934\times 10^{29}x^{4}-7.49441\times 10^{28}x^{3}+7.59092\times 10^{27}x^{2}-4.45216\times 10^{26}x+1.12845\times 10^{25}\right).
$$

We also state without proof that the maximum of $u^{(13)}(x,0)$ in the interval $[0,1]$ (which is also a global maximum) is a whopping $\approx 3.18674\times 10^{19}$, which occurs at $x\approx 0.470825$, which means that the maximum possible error for our solution is $\approx\frac{3.18674\times 10^{19}}{2.55059\times 10^{13}}\approx 1.249414\times 10^{6}$. This example shows that, for a very high-order Legendre polynomial Sobolev space, if our derivatives happen to grow too much, we could still fail to produce a solution that does not exhibit Runge's phenomenon, since the error becomes bounded by a very large constant.

Note that we also might have rounding errors from the inversion of matrix equations in our code. As $p$ increases, we get matrices which might become more and more singular, i.e. ill-conditioned. This means that small rounding errors might become extremely amplified if $p$ is large, also further contributing to the errors.

**Part (b)**

For this next part, we assume that, as usual, the global error in our finite element formulation depends on the stepsize in some kind of power law, i.e. $\left||u-u_{h}|\right|_{2}\approx Ch^{r}$, where $u$ is the exact solution, $u_{h}$ is the approximate (numerical) solution, $h$ is some one-dimensional spatial quantity (say, the average spacing between neighboring nodes) and $C,r$ constants greater than zero. Then, by plotting this max norm error as $h$ varies in a log-log plot, the slope of the resulting graph will give us $r$, because 

$$
\left||u-u_{h}|\right|_{max}\approx Ch^{r}\Rightarrow\log\left||u-u_{h}|\right|_{max}\approx \log\left(Ch^{r}\right)=\log C+\log h^{r}=\log C+r\log h.\quad(5)
$$

In our case, we can let our "spatial quantity" be $n\cdot p$, which is the total number of nodes used in calculating our solution. However, this is roughly inversely proportional to the average spacing between neighboring nodes, so when we look at our loglog plot, we will get the same desired convergence plot, but with a negative correlation vs. a positive one (so the slope will be negative of what it should be). 

We now run our function `dgconvect` for $p=1,2,4,8,12$ (we use $p=12$ because the case for $p=16$ already exhibits Runge's phenomenon for $n=2$, which quickly jeopardizes the accuracy of our solution), with $\Delta t = 2\cdot 10^{-4}$ (i.e. `dt` in our code), $T=1.0$, and with $n$ chosen such that $n\cdot p=16,32,64,128,256$. Of course, we can calculate $n$ each time by dividing $16$, $32$, etc. by our given value of $p$. Then, we calculate the $L_{2}$ error for each of these $25$ cases and store it in a $5$-by-$5$ array called `errors`, where the rows will contain the errors for each $p$ value, and the columns for each $n$ value. 

From there, given the theory in $(5)$, we can estimate the slopes of the logarithm of the decreasing error for each given value of $p$. Using the theory in $(5)$, this slope will give us $r$, our rate of convergence for using the DG method on the convection equation with a $p$th-order Legendre polynomial Sobolev space. Of course, to estimate each of these slopes, we use a linear least squares best-fit slope on our log-log data.

This can all be done with our `dgconvect_convergence()` function, which we define below. Since we do not want our function to print a plot each time we run our function for each of the $25$ cases, we define another function `dgconvectnoplot` which is identical to `dgconvect`, except with the plotting portion of the code removed. We also remove errors from our `errors` matrix if they increase from the previous ones, since this demonstrates either rounding errors near machine tolerance, Runge's phenomenon, or general rounding errors in the various calculations using matrix equations from earlier.

```julia
using LinearAlgebra
using PyPlot

function dgconvectnoplot(; n = 10, p = 1, T = 1.0, dt = 1e-3)
    # Discretization
    h = 1 / n
    i = collect(0 : p)
    s = @. cos.(pi * i / p) #Chebyshev nodes from 1 to -1
    s = h * (1 .- s) / 2 #Chebyshev nodes shifted and scaled to [0, h]
    x = @. s + (0 : h : 1 - h)'

    # Gaussian initial condition (and exact solution if shifted)
    uinit(x) = exp(-(x - 0.5) ^ 2 / 0.1 ^ 2)

    gx, gw = gauss_quad(2 * p)
    V, dV = legendre_poly(s, p) #Legendre polynomials evaluated at Chebyshev nodes
    C = V \ I #coefficients for basis
    sq = h * (gx .+ 1) / 2 #quadrature points (after change of interval)
    P, dP = legendre_poly(sq, p) #Legendre polynomials and their derivatives evaluated at quadrature points
    phi = P * C #basis functions evaluated at quadrature points
    dphi = dP * C #derivative of basis functions evaluated at quadrature points
    M = zeros(p + 1, p + 1)
    K = zeros(p + 1, p + 1)
    for i = 1 : p + 1 #construct elementary matrices
        for j = 1 : p + 1
            M[i, j] = dot(phi[:, i] .* phi[:, j], gw) 
            K[i, j] = dot(dphi[:, i] .* phi[:, j], gw) 
        end
    end
    Mel = (h / 2) * Symmetric(M) #multiply by Jacobian constant
    Kel = (h / 2) * K #""
        
    # RHS function in semi-discrete system
    function rhs(u)
        r = Kel * u
        r[end,:] = r[end,:] - u[end,:]
        r[1,:] = r[1,:] + u[end, [end; 1:end-1]]
        r = Mel \ r
    end

    # Setup
    u = uinit.(x)
    nsteps = round(Int, T / dt)
    
    # Main loop
    for it = 1 : nsteps
        # Runge-Kutta 4
        k1 = dt * rhs(u)
        k2 = dt * rhs(u + k1 / 2)
        k3 = dt * rhs(u + k2 / 2)
        k4 = dt * rhs(u + k3)
        u += (k1 + 2 * k2 + 2 * k3 + k4) / 6
    end

    #calculate L2 norm 
    int = 0
    for i = 1 : n #integrate using Gauss-Legendre quadrature of arbitrarily high order
        uloc0 = u[:, i] #extract values of u at local nodes on the element
        xloc0 = x[:, i] #extract nodes in this element
        gx, gw = gauss_quad(100)
        Ploc0, ~ = legendre_poly(xloc0, p) #evaluate Legendre polynomials at local nodes
        coeff = Ploc0 \ uloc0 #calculate coefficients of local Legendre polynomial basis
        #define new nodes, such that there are 3p of them
        gxq = 0.5 * ((xloc0[end] - xloc0[1]) * gx .+ xloc0[1] .+ xloc0[end])
        Plocq, ~ = legendre_poly(gxq, p) #evaluate Legendre polynomials at quadrature points
        ulocq = Plocq * coeff #find the values of u at quadrature points
        uexactlocq = @. uinit(mod(gxq - T, 1.0)) #exact solution at quadrature points
        int += 0.5 * (xloc0[end] - xloc0[1]) * dot((uexactlocq - ulocq) .^ 2, gw) 
        #add contribution of this element to the integral
    end
    error = sqrt(int)  #calculate L2 norm error
    return u, error
end
```

```julia
function dgconvect_convergence()
    pq = [1 2 4 8 12]
    ntot = [16 32 64 128 256]
    errors = zeros(length(pq), length(ntot))
    slopes = zeros(length(pq))
    for i = 1 : length(pq)
        p = pq[i]
        for j = 1 : length(ntot)
            n = round(Int, ntot[j] / p) #calculate corresponding value of n
            ~, errors[i, j] = dgconvectnoplot(p = p, n = n, T = 1.0, dt = 2 * 1e-4)
            if j > 1
                if errors[i, j] > errors[i, j - 1] 
                    #if there are rounding errors (e.g. the error starts increasing), flag with NaN
                    errors[i, j] = NaN
                    break
                end
            end
        end
        counter = 1 
        for k = 2 : length(ntot) - 1 #throw out "bad" points from error slope calculation
            if isnan(errors[i, k]) || log(errors[i, k]) > log(errors[i, k - 1])
                break
            end
            counter += 1
        end
        logerror = log.(errors[i, 1 : counter])
        logntot = log.(ntot[1 : counter])
        #calculate best-fit slope using linear regression techniques
        logerrormean = sum(logerror) / counter
        logntotmean = sum(logntot) / counter
        slopes[i] = -dot(logerror .- logerrormean, logntot .- logntotmean) / dot(logntot .- logntotmean, logntot .- logntotmean)
    end
    return errors, slopes
end
```

```julia
errors, slopes = dgconvect_convergence()

npq = [16, 32, 64, 128, 256] * transpose(ones(5))

clf()
loglog(npq, transpose(errors))
title("Convergence plots for the convection equation")
xlabel(L"\log\left(n\cdot p\right)")
ylabel(L"\log\left||u-u_{exact}|\right|_{2}")
gca().legend(("p=1", "p=2", "p=4", "p=8", "p=12"))

println(slopes)
```

The slopes for the error convergence rates with $p=1,2,4,8,12$ are $\approx 2.28853, \approx 3.60685, \approx 5.23873, \approx 5.85332$, and $\approx 9.19059$, respectively.

Thus, our solution exhibits roughly linear convergence up to the points where we started having to exclude the error estimates. Also, visually, for the $p=8$ case, we might have needed to exclude the $n\cdot p=64$ case as well, because the calculated slope does not fit the trend of the rest of the data points, and had we excluded that point, the slope would better match. Nevertheless, we find a mostly consistent convergence rate that increases with our $p$ value, up to when $n$ becomes too large and other errors (rounding, Runge's phenomenon, etc.) begin to dominate.

## Problem 2

**Part (a)** 

We now desire to write a similar script as `dgconvect` which runs the same way and also applies the LDG (local discontinuous Galerkin) method, except on the convection-diffusion equation. We will call this script `dgconvdiff`. The convection-diffusion equation is the same as the convection equation which we solved with the `dgconvect`,

$$
\frac{\partial u}{\partial t}+\frac{\partial u}{\partial x}=0,
$$

except we have an extra second term added to it, which represents diffusion:

$$
\frac{\partial u}{\partial t}+\frac{\partial u}{\partial x}-k\frac{\partial ^{2}u}{\partial x^{2}}=0.\quad(6)
$$

In many ways, we can think of the convection-diffusion equation as representing some quantity (whether heat, fluid density, etc.) under both simultaneous convection and diffusion.


Regarding the LDG method we use to numerically solve $(6)$, the scheme (and therefore our code) is identical with regards to using the Chebyshev nodes, finding the local $p$th-order Legendre polynomial basis functions, calculating $M$ and $C$ (or, in the code, `Mel` and `Kel`, respectively), etc. The only part that changes is in computing the righthand side in the semi-discrete system which we use to timestep our solution forward (and of course, the exact solution which we use to compare with our numerical solution). Namely, we can rewrite $(6)$ as a system of first-order PDEs, or

$$
\frac{\partial u}{\partial t}+\frac{\partial u}{\partial x}-k\frac{\partial\sigma}{\partial x}=\frac{\partial u}{\partial t}+\frac{\partial\left(u-k\sigma\right)}{\partial x}=0,\quad
\frac{\partial u}{\partial x}=\sigma.
$$

From here, we derive the Galerkin form for this equation all the same, as outlined in Professor Persson's DG slides, except we have an extra term depending on $k$ and $\sigma$, as well as an extra equation for $\sigma$. At the end of it all, the righthand side of our semi-discrete system for timestepping requires us to solve *two* matrix equations as opposed to just one, so this is the only part we modify in our code from `dgconvect`, aside from defining the exact solution as a function, `uexact(x)`, and evaluating it instead. The two equations that we have to solve at each timestep are

$$
M^{k}\dot{\mathbf{u}}^{k}=C^{k}\left(\mathbf{u}^{k}-k\mathbf{\sigma}^{k}\right)+\begin{bmatrix} u_{p}^{k-1}-k\sigma_{p}^{k-1} \\ 0 \\ \vdots \\ 0 \\ -\left(u_{p}^{k}-k\sigma_{p}^{k}\right) \end{bmatrix},\quad M^{k}\dot{\mathbf{\sigma}}^{k}=-C^{k}\mathbf{u}^{k}+\begin{bmatrix} -u_{0}^{k} \\ 0 \\ \vdots \\ 0 \\ u_{0}^{k+1} \end{bmatrix},\quad(7)
$$

and as expected, we observe that we retrieve the same DG method for the convection equation if $k\rightarrow 0$, because then, the second equation for $\mathbf{\sigma}^{k}$ becomes irrelevant in calculating the first equation for $\dot{\mathbf{u}}^{k}$. 

Finally, to timestep $(7)$, we use RK4 (i.e. *the* Runge-Kutta method), just as we did in `dgconvect`. We also let $k=10^{-3}$ as a particular example for plotting.

```julia
using LinearAlgebra
using PyPlot

function dgconvdiff(; n = 10, p = 1, T = 1.0, dt = 1e-3, k = 1e-3)
    # Discretization
    h = 1 / n
    i = collect(0 : p)
    s = @. cos.(pi * i / p) #Chebyshev nodes from 1 to -1
    s = h * (1 .- s) / 2 #Chebyshev nodes shifted and scaled to [0, h]
    x = @. s + (0 : h : 1 - h)'

    # Gaussian initial condition (and exact solution if shifted)
    uinit(x) = exp(-100 * (x - 0.5) ^ 2)
    
    # exact solution (N=-1,0,1)
    uexact(x, t) = (1 + 400 * k * t) ^ (-1 / 2) * (exp(-100 * ((x - 0.5 - 1) ^ 2) / (1 + 400 * k * t)) + exp(-100 * ((x - 0.5) ^ 2) / (1 + 400 * k * t)) + exp(-100 * ((x - 0.5 + 1) ^ 2) / (1 + 400 * k * t)))
    
    gx, gw = gauss_quad(2 * p)
    V, dV = legendre_poly(s, p) #Legendre polynomials evaluated at Chebyshev nodes
    C = V \ I #coefficients for basis
    sq = h * (gx .+ 1) / 2 #quadrature points (after change of interval)
    P, dP = legendre_poly(sq, p) #Legendre polynomials and their derivatives evaluated at quadrature points
    phi = P * C #basis functions evaluated at quadrature points
    dphi = dP * C #derivative of basis functions evaluated at quadrature points
    M = zeros(p + 1, p + 1)
    K = zeros(p + 1, p + 1)
    for i = 1 : p + 1 #construct elementary matrices
        for j = 1 : p + 1
            M[i, j] = dot(phi[:, i] .* phi[:, j], gw) 
            K[i, j] = dot(dphi[:, i] .* phi[:, j], gw) 
        end
    end
    Mel = (h / 2) * Symmetric(M) #multiply by Jacobian constant
    Kel = (h / 2) * K #""
        
    # RHS function in semi-discrete system
    function rhs(u)
        w = -Kel * u
        w[end, :] += u[1, [2 : end; 1]] 
        w[1, :] -= u[1, :]
         = Mel \ w
        r = Kel * (u - k * )
        r[end, :] += k * [end, :] - u[end, :]
        r[1, :] += u[end, [end; 1 : end - 1]] - k * [end, [end; 1 : end - 1]]
        r = Mel \ r
    end

    # Setup
    splot = @. collect(0 : 3 * p) / (3 * p) #vector for 3p local nodes
    xplot = @. h * splot + (0 : h : 1 - h)'
    uplot = uinit.(xplot)
    u = uinit.(x)
    nsteps = round(Int, T / dt)

    # Setup plotting
    xx = collect(0 : 0.01 : 1) # For exact solution
    clf(); axis([0, 1, -0.1, 1.1]); grid(true);
    ph1 = plot(xplot, uplot, "b")   # Many line object (since discontinuous)
    ph2, = plot(xx, uinit.(xx), "k") # Single line object
    ph3, = plot(xx, uinit.(xx), "r", linestyle="--") # Single line object
    
    # Main loop
    for it = 1 : nsteps
        # Runge-Kutta 4
        k1 = dt * rhs(u)
        k2 = dt * rhs(u + k1 / 2)
        k3 = dt * rhs(u + k2 / 2)
        k4 = dt * rhs(u + k3)
        u += (k1 + 2 * k2 + 2 * k3 + k4) / 6

        # Plotting
        if mod(it, round(nsteps / 100)) == 0 #plot final solution.
            uexactq = @. uexact(xx, dt * it)
            for i = 1 : n #loop through each element
                uloc0 = u[:, i] #extract values of u at local nodes on the element
                xloc0 = x[:, i] #extract nodes in this element
                Ploc0, ~ = legendre_poly(xloc0, p) #evaluate Legendre polynomials at local nodes
                coeff = Ploc0 \ uloc0 #calculate coefficients of local Legendre polynomial basis
                #define new nodes, such that there are 3p of them per element (i=0,...,3p, where i=3p overlaps with the next)
                xlocf = xloc0[1] .+ (collect(0 : 3 * p) * (xloc0[end] - xloc0[1]) / (3 * p)) 
                Plocf, ~ = legendre_poly(xlocf, p) #evaluate Legendre polynomials at the 3p local nodes
                uplot[:, i] = Plocf * coeff #find the values of u at the 3p local nodes
            end
            for line in ph1 line.remove(); end
            ph1 = plot(xplot, uplot, "b")
            ph2.set_data(xx, uexactq)
            ph3.set_data(xx, uinit.(xx))
            gca().legend((L"u\left(x,t\right)", L"u\left(x,0\right)", L"u_{h}\left(x,t\right)"))
            pause(1e-3)
        end
    end

    #calculate L2 norm 
    int = 0
    for i = 1 : n #integrate using Gauss-Legendre quadrature of arbitrarily high order
        uloc0 = u[:, i] #extract values of u at local nodes on the element
        xloc0 = x[:, i] #extract nodes in this element
        gx, gw = gauss_quad(100)
        Ploc0, ~ = legendre_poly(xloc0, p) #evaluate Legendre polynomials at local nodes
        coeff = Ploc0 \ uloc0 #calculate coefficients of local Legendre polynomial basis
        #define new nodes, such that there are 3p of them
        gxq = 0.5 * ((xloc0[end] - xloc0[1]) * gx .+ xloc0[1] .+ xloc0[end])
        Plocq, ~ = legendre_poly(gxq, p) #evaluate Legendre polynomials at quadrature points
        ulocq = Plocq * coeff #find the values of u at quadrature points
        uexactlocq = @. uexact(gxq, T) 
        int += 0.5 * (xloc0[end] - xloc0[1]) * dot((uexactlocq - ulocq) .^ 2, gw) 
        #add contribution of this element to the integral
    end
    error = sqrt(int)  #calculate L2 norm error
    return u, error
end
```

With this, we plot some test solutions which show the exact solution in solid black, the initial condition (which we recall has not changed from problem 1) in dotted red, and the numerical solution in solid blue.

The numerical solution appears to track the convection better than the diffusion, but this might just have to do with how the errors seem to be larger closer to the peak of the Gaussian, which intuitively makes sense. (After all, the higher-order derivatives which bound the error in our numerical solution are greatest near there).

```julia
using LinearAlgebra
using PyPlot

function dgconvdiffnoplot(; n = 10, p = 1, T = 1.0, dt = 1e-3, k = 1e-3)
    # Discretization
    h = 1 / n
    i = collect(0 : p)
    s = @. cos.(pi * i / p) #Chebyshev nodes from 1 to -1
    s = h * (1 .- s) / 2 #Chebyshev nodes shifted and scaled to [0, h]
    x = @. s + (0 : h : 1 - h)'

    # Gaussian initial condition (and exact solution if shifted)
    uinit(x) = exp(-100 * (x - 0.5) ^ 2)
     
    # exact solution (N=-1,0,1)
    uexact(x, t) = (1 + 400 * k * t) ^ (-1 / 2) * (exp(-100 * ((x - 0.5 - 1) ^ 2) / (1 + 400 * k * t)) + exp(-100 * ((x - 0.5) ^ 2) / (1 + 400 * k * t)) + exp(-100 * ((x - 0.5 + 1) ^ 2) / (1 + 400 * k * t)))

    gx, gw = gauss_quad(2 * p)
    V, dV = legendre_poly(s, p) #Legendre polynomials evaluated at Chebyshev nodes
    C = V \ I #coefficients for basis
    sq = h * (gx .+ 1) / 2 #quadrature points (after change of interval)
    P, dP = legendre_poly(sq, p) #Legendre polynomials and their derivatives evaluated at quadrature points
    phi = P * C #basis functions evaluated at quadrature points
    dphi = dP * C #derivative of basis functions evaluated at quadrature points
    M = zeros(p + 1, p + 1)
    K = zeros(p + 1, p + 1)
    for i = 1 : p + 1 #construct elementary matrices
        for j = 1 : p + 1
            M[i, j] = dot(phi[:, i] .* phi[:, j], gw) 
            K[i, j] = dot(dphi[:, i] .* phi[:, j], gw) 
        end
    end
    Mel = (h / 2) * Symmetric(M) #multiply by Jacobian constant
    Kel = (h / 2) * K #""
        
    # RHS function in semi-discrete system
    function rhs(u)
        w = -Kel * u
        w[end, :] += u[1, [2 : end; 1]] 
        w[1, :] -= u[1, :]
         = Mel \ w
        r = Kel * (u - k * )
        r[end, :] += k * [end, :] - u[end, :]
        r[1, :] += u[end, [end; 1 : end - 1]] - k * [end, [end; 1 : end - 1]]
        r = Mel \ r
    end

    # Setup
    splot = @. collect(0 : 3 * p - 1) / (3 * p - 1) #vector for 3p local nodes
    xplot = @. h * splot + (0 : h : 1 - h)'
    uplot = uinit.(xplot)
    u = uinit.(x)
    nsteps = round(Int, T / dt)
    
    # Main loop
    for it = 1 : nsteps
        # Runge-Kutta 4
        k1 = dt * rhs(u)
        k2 = dt * rhs(u + k1 / 2)
        k3 = dt * rhs(u + k2 / 2)
        k4 = dt * rhs(u + k3)
        u += (k1 + 2 * k2 + 2 * k3 + k4) / 6
    end

    #calculate L2 norm 
    int = 0
    for i = 1 : n #integrate using Gauss-Legendre quadrature of arbitrarily high order
        uloc0 = u[:, i] #extract values of u at local nodes on the element
        xloc0 = x[:, i] #extract nodes in this element
        gx, gw = gauss_quad(100)
        Ploc0, ~ = legendre_poly(xloc0, p) #evaluate Legendre polynomials at local nodes
        coeff = Ploc0 \ uloc0 #calculate coefficients of local Legendre polynomial basis
        #define new nodes, such that there are 3p of them
        gxq = 0.5 * ((xloc0[end] - xloc0[1]) * gx .+ xloc0[1] .+ xloc0[end])
        Plocq, ~ = legendre_poly(gxq, p) #evaluate Legendre polynomials at quadrature points
        ulocq = Plocq * coeff #find the values of u at quadrature points
        uexactlocq = @. uexact(gxq, T) 
        int += 0.5 * (xloc0[end] - xloc0[1]) * dot((uexactlocq - ulocq) .^ 2, gw) 
        #add contribution of this element to the integral
    end
    error = sqrt(int)  #calculate L2 norm error
    return u, error
end
```

```julia
function dgconvdiff_convergence()
    pq = [1 2 4 8 12]
    ntot = [16 32 64 128 256]
    errors = zeros(length(pq), length(ntot))
    slopes = zeros(length(pq))
    for i = 1 : length(pq)
        p = pq[i]
        for j = 1 : length(ntot)
            n = round(Int, ntot[j] / p)
            ~, errors[i, j] = dgconvdiffnoplot(p = p, n = n, T = 1.0, dt = 2 * 1e-4, k = 1e-3)
            if j > 1
                if errors[i, j] > errors[i, j - 1] 
                    #if there are rounding errors (e.g. the error starts increasing), flag with NaN
                    errors[i, j] = NaN
                    break
                end
            end
        end
        counter = 1 
        for k = 2 : length(ntot) - 1 #throw out "bad" points from error slope calculation
            if isnan(errors[i, k]) || log(errors[i, k]) > log(errors[i, k - 1])
                break
            end
            counter += 1
        end
        logerror = log.(errors[i, 1 : counter])
        logntot = log.(ntot[1 : counter])
        #calculate best-fit slope using linear regression techniques
        logerrormean = sum(logerror) / counter
        logntotmean = sum(logntot) / counter
        slopes[i] = -dot(logerror .- logerrormean, logntot .- logntotmean) / dot(logntot .- logntotmean, logntot .- logntotmean)
    end
    return errors, slopes
end
```

```julia
errors, slopes = dgconvdiff_convergence()

npq = [16, 32, 64, 128, 256] * transpose(ones(5))

clf()
loglog(npq, transpose(errors))
title("Convergence plots for the convection-diffusion equation")
xlabel(L"\log\left(n\cdot p\right)")
ylabel(L"\log\left||u-u_{exact}|\right|_{2}")
gca().legend(("p=1", "p=2", "p=4", "p=8", "p=12"))

println(slopes)
```

The slopes for the error convergence rates with $p=1,2,4,8,12$ are $\approx 2.35519, \approx 3.60695, \approx 5.33688, \approx 5.33763$, and $\approx 8.58837$, respectively.

As in problem 1, our solution exhibits roughly linear convergence up to the points where we started having to exclude the error estimates, and we also have the same issue with the $p=8$ case. Hence, we again find a mostly consistent convergence rate that increases with our $p$ value, up to when $n$ becomes too large and other errors appear. Also, we note that our numerical solution demonstrates a remarkable consistency in the convergence rates, which are very similar as those for the DG method as applied in problem 1, which had error convergence rates of $\approx 2.28853, \approx 3.60685, \approx 5.23873, \approx 5.85332$, and $\approx 9.19059$, respectively.
