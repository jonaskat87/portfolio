## Problem 1

**Part (a)**

We want to formulate a Galerkin method for the following boundary-value problem:

$$
u''''(x)=f(x)=480x-120,\quad \forall x\in(0,1)\quad (1)\\
u(0)=u'(0)=u(1)=u'(1)=0 \quad (2)
$$

Suppose that we choose a space of discrete test functions $V_{h}$ which also satisfy the boundary conditions that the solution to our BVP, $u$, satisfies in $(2)$ (i.e. $v(0)=v'(0)=v(1)=v'(1)=0\hspace{0.15cm}\forall v\in V_{h}$). Then, let us integrate both sides of $(1)$ over the domain on which our BVP is defined, and multiply some test function $v\in V_{h}$ to both sides.

$$
\int_{0}^{1}{u''''(x)v(x)dx}=\int_{0}^{1}{f(x)v(x)dx}\quad\forall v\in V_{h}.\quad (3)
$$

We apply integration by parts to $(3)$ by letting $df=u''''(x)dx\Rightarrow\int{df}=\int{u''''(x)dx}\Rightarrow f=u'''$ and $v=v(x)\Rightarrow dv=v'(x)dx$. Then, the left side of $(3)$ becomes

$$
\int_{0}^{1}{u''''(x)v(x)dx}=\left.u'''(x)v(x)\right|_{0}^{1}-\int_{0}^{1}{u'''(x)v'(x)dx}.\quad (4)
$$

Since $v\in V_{h}$, by construction, $v(0)=v(1)=0$. Therefore, as long as $u$ is well-behaved (and it should be, given that the BVP assumes that $u$ is at least four times continuously differentiable within the domain and at least once along the boundary of this domain), $(4)$ reduces to

$$
\int_{0}^{1}{u''''(x)v(x)dx}=-\int_{0}^{1}{u'''(x)v'(x)dx}.\quad (5)
$$

Again, we use integration by parts on the right integral in $(5)$ by letting $dg=u'''(x)dx\Rightarrow\int{dg}=\int{u'''(x)dx}\Rightarrow g=u''$ and $w=v'(x)\Rightarrow dw=v''(x)dx$. Then, 

$$
\int_{0}^{1}{u'''(x)v'(x)dx}=\left.u''(x)v'(x)\right|_{0}^{1}-\int_{0}^{1}{u''(x)v''(x)dx}.\quad(6)
$$

As for $(4)$, by construction of members of $V_{h}$, $v'(0)=v'(1)=0$. Thus, $(6)$ becomes

$$
\int_{0}^{1}{u'''(x)v'(x)dx}=-\int_{0}^{1}{u''(x)v''(x)dx}.\quad(7)
$$

Finally, substituting $(7)$ into $(5)$, and then into $(3)$,

$$
\int_{0}^{1}{u''''(x)v(x)dx}=-\left(-\int_{0}^{1}{u''(x)v''(x)dx}\right)=\int_{0}^{1}{u''(x)v''(x)dx}\Rightarrow\int_{0}^{1}{u''(x)v''(x)dx}=\int_{0}^{1}{u''''(x)v(x)dx}=\int_{0}^{1}{f(x)v(x)dx}.\quad(8)
$$

Now, suppose that we approximate $u$ by a discrete linear combination of members in $V_{h}$, $u_{h}$. Then, $u\approx u_{h}\in V_{h}$, so under the Galerkin formation, $u$ can be replaced by $u_{h}$ in $(8)$, giving us

$$
\boxed{\int_{0}^{1}{u_{h}''(x)v''(x)dx}=\int_{0}^{1}{f(x)v(x)dx}\quad\forall v\in V_{h}}.\quad(9)
$$

**Part (b)**

Following the suggestion given in the problem statement, we want to look for a basis of Hermite polynomials. That is, for each element (of which there are two in this case), we want $u$ to be represented by a space of cubic functions, such that these cubic functions also have continuous derivatives at the boundaries. This will give us four degrees of freedom on each element $K_{j}=\left[x_{j},x_{j+1}\right]$, enough to form a cubic basis: 1) the value of $u$ matches at $x_{j}$, 2) the value of $u'$ matches at $x_{j}$, 3) the value of $u$ matches at $x_{j+1}$, and 4) the value of $u'$ matches at $x_{j+1}$. By allowing these to vary, we can assert that the derivative of a basis function in two neighboring elements is the same at the node points.

We now explicitly find a basis $\{\varphi_{i}\}$ for elements $K_{1}$ and $K_{2}$. We note that each member of the basis can be written as $\varphi_{i}(x)=\sum_{k=0}^{3}{a_{ik}x^{k}}$, because each is a polynomial of degree *at most* 3, and that since there are four degrees of freedom for each element, there are four members for the basis of each element. Furthermore, to solve for the basis functions associated with this element, we note the following property: namely, $\varphi_{i}\left(x_{j}\right)=\delta_{ij}\hspace{0.15cm}\forall\varphi_{i}\in\{\varphi_{i}\}$.

***Basis for*** $K_{1}=\left[0,\frac{1}{2}\right]$.

Because of the last property aforementioned, we have the system of sixteen equations for sixteen unknowns, the four coefficients for each of the four functions in our basis:

$$
\varphi_{0}(0)=1,\varphi_{0}'(0)=\varphi_{0}\left(\frac{1}{2}\right)=\varphi_{0}'\left(\frac{1}{2}\right)=0,\\
\varphi_{1}'(0)=1,\varphi_{1}(0)=\varphi_{1}\left(\frac{1}{2}\right)=\varphi_{1}'\left(\frac{1}{2}\right)=0,\\
\varphi_{2}\left(\frac{1}{2}\right)=1,\varphi_{2}(0)=\varphi_{2}'\left(0\right)=\varphi_{2}'\left(\frac{1}{2}\right)=0,\\
\varphi_{3}'\left(\frac{1}{2}\right)=1,\varphi_{3}(0)=\varphi_{3}'\left(0\right)=\varphi_{3}\left(\frac{1}{2}\right)=0.\quad(10)
$$

More generally, for any element $K_{j}=\left[x_{j},x_{j+1}\right]$,

$$
X\left(x_{j}\right):=\begin{bmatrix} 1 & x_{j} & x_{j}^{2} & x_{j}^{3} \\ 0 & 1 & 2x_{j} & 3x_{j}^{2} \\ 1 & x_{j+1} & x_{j+1}^{2} & x_{j+1}^{3} \\ 0 & 1 & 2x_{j+1} & 3x_{j+1}^{2} \end{bmatrix},\quad \mathbf{a}_{i}:=\begin{bmatrix} a_{i,0} \\ a_{i,1} \\ a_{i,2} \\ a_{i,3} \end{bmatrix},\quad(11)
$$

and with $(11)$ defined, we can actually recast $(10)$ into a more compact form as four matrix equations. Namely,

$$
X\left(0\right)\mathbf{a}_{0}=\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix},\quad X\left(0\right)\mathbf{a}_{1}=\begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix},\quad X\left(0\right)\mathbf{a}_{2}=\begin{bmatrix} 0 \\ 0 \\ 1 \\ 0 \end{bmatrix},\quad X\left(0\right)\mathbf{a}_{3}=\begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \end{bmatrix}.\quad(12)
$$

Furthermore, we can invert the same $X\left(0\right)$ to solve all of the four matrix equations in $(12)$ at once. Avoiding tedious computation by hand (which is possible to do though), Mathematica gives us that

$$
X^{-1}\left(0\right)=\begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \\ -12 & -4 & 12 & -2 \\ 16 & 4 & -16 & 4 \end{bmatrix}.
$$

Using this to solve the four equations in $(12)$,

$$
\mathbf{a}_{0}=X^{-1}\left(0\right)\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix}=\begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \\ -12 & -4 & 12 & -2 \\ 16 & 4 & -16 & 4 \end{bmatrix}\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix}=\begin{bmatrix} 1 \\ 0 \\ -12 \\ 16 \end{bmatrix},\quad\mathbf{a}_{1}=X^{-1}\left(0\right)\begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix}=\begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \\ -12 & -4 & 12 & -2 \\ 16 & 4 & -16 & 4 \end{bmatrix}\begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix}=\begin{bmatrix} 0 \\ 1 \\ -4 \\ 4 \end{bmatrix},\\
\mathbf{a}_{2}=X^{-1}\left(0\right)\begin{bmatrix} 0 \\ 0 \\ 1 \\ 0 \end{bmatrix}=\begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \\ -12 & -4 & 12 & -2 \\ 16 & 4 & -16 & 4 \end{bmatrix}\begin{bmatrix} 0 \\ 0 \\ 1 \\ 0 \end{bmatrix}=\begin{bmatrix} 0 \\ 0 \\ 12 \\ -16 \end{bmatrix},\quad\mathbf{a}_{3}=X^{-1}\left(0\right)\begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \end{bmatrix}=\begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \\ -12 & -4 & 12 & -2 \\ 16 & 4 & -16 & 4 \end{bmatrix}\begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \end{bmatrix}=\begin{bmatrix} 0 \\ 0 \\ -2 \\ 4 \end{bmatrix}.\quad(13)
$$

Thus, by $(13)$, our four basis functions for $K_{1}$ without other constraints would be

$$
\begin{matrix} \phi_{0}(x)=1-12x^{2}+16x^{3}, & \phi_{1}(x)=x-4x^{2}+4x^{3}, \\ \phi_{2}(x)=12x^{2}-16x^{3}, & \phi_{3}(x)=-2x^{2}+4x^{3} \end{matrix}\Rightarrow Span\left\{1-12x^{2}+16x^{3},x-4x^{2}+4x^{3},12x^{2}-16x^{3},-2x^{2}+4x^{3}\right\}.
$$

However, our boundary conditions enforce that $v(0)=v'(0)=0\hspace{0.25cm}\forall v\in V_{h}$. Thus, we must remove $\phi_{0}$ and $\phi_{1}$ (as defined above) from our basis, since these are nonzero at this boundary. Therefore, enforcing constraints imposed by our function space, an *actual* basis for $K_{1}$ is 

$$
\boxed{Span\left\{12x^{2}-16x^{3},-2x^{2}+4x^{3}\right\}}.\quad(14)
$$

***Basis for*** $K_{2}=\left[\frac{1}{2},1\right]$.

We can use $(11)$ again this time, except with $x_{j}=1/2$ and $x_{j+1}=1$. Then, to solve for the coefficients of each cubic function in our basis, we need the coefficients to satisfy the following four matrix equations, similar in form to $(12)$:

$$
X\left(\frac{1}{2}\right)\mathbf{a}_{0}=\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix},\quad X\left(\frac{1}{2}\right)\mathbf{a}_{1}=\begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix},\quad X\left(\frac{1}{2}\right)\mathbf{a}_{2}=\begin{bmatrix} 0 \\ 0 \\ 1 \\ 0 \end{bmatrix},\quad X\left(\frac{1}{2}\right)\mathbf{a}_{3}=\begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \end{bmatrix}.\quad(15)
$$

As before, we can invert $X\left(\frac{1}{2}\right)$ and solve all of the equations in $(15)$ at once. Mathematica gives us that

$$
X^{-1}\left(\frac{1}{2}\right)=\begin{bmatrix} -4 & -2 & 5 & -1 \\ 24 & 8 & -24 & 5 \\ -36 & -10 & 36 & -8 \\ 16 & 4 & -16 & 4 \end{bmatrix}.
$$

Thus, using this to solve the equations in $(15)$,

$$
\mathbf{a}_{0}=X^{-1}\left(0\right)\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix}=\begin{bmatrix} -4 & -2 & 5 & -1 \\ 24 & 8 & -24 & 5 \\ -36 & -10 & 36 & -8 \\ 16 & 4 & -16 & 4 \end{bmatrix}\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix}=\begin{bmatrix} -4 \\ 24 \\ -36 \\ 16 \end{bmatrix},\quad\mathbf{a}_{1}=X^{-1}\left(0\right)\begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix}=\begin{bmatrix} -4 & -2 & 5 & -1 \\ 24 & 8 & -24 & 5 \\ -36 & -10 & 36 & -8 \\ 16 & 4 & -16 & 4 \end{bmatrix}\begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix}=\begin{bmatrix} -2 \\ 8 \\ -10 \\ 4 \end{bmatrix},\\
\mathbf{a}_{2}=X^{-1}\left(0\right)\begin{bmatrix} 0 \\ 0 \\ 1 \\ 0 \end{bmatrix}=\begin{bmatrix} -4 & -2 & 5 & -1 \\ 24 & 8 & -24 & 5 \\ -36 & -10 & 36 & -8 \\ 16 & 4 & -16 & 4 \end{bmatrix}\begin{bmatrix} 0 \\ 0 \\ 1 \\ 0 \end{bmatrix}=\begin{bmatrix} 5 \\ -24 \\ 36 \\ -16 \end{bmatrix},\quad\mathbf{a}_{3}=X^{-1}\left(0\right)\begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \end{bmatrix}=\begin{bmatrix} -4 & -2 & 5 & -1 \\ 24 & 8 & -24 & 5 \\ -36 & -10 & 36 & -8 \\ 16 & 4 & -16 & 4 \end{bmatrix}\begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \end{bmatrix}=\begin{bmatrix} -1 \\ 5 \\ -8 \\ 4 \end{bmatrix}.\quad(16)
$$

Thus, by $(16)$, our four basis functions for $K_{2}$ (without other constraints) would be

$$
\begin{matrix} \phi_{0}(x)=-4+24x-36x^{2}+16x^{3}, & \phi_{1}(x)=-2+8x-10x^{2}+4x^{3}, \\ \phi_{2}(x)=5-24x+36x^{2}-16x^{3}, & \phi_{3}(x)=-1+5x-8x^{2}+4x^{3} \end{matrix}\Rightarrow Span\left\{-4+24x-36x^{2}+16x^{3},-2+8x-10x^{2}+4x^{3},5-24x+36x^{2}-16x^{3},-1+5x-8x^{2}+4x^{3}\right\}.
$$

However, our boundary conditions enforce that $v(1)=v'(1)=0\hspace{0.25cm}\forall v\in V_{h}$. Thus, we must remove $\phi_{2}$ and $\phi_{3}$ (as defined above) from our basis, as they are nonzero at this boundary. Therefore, enforcing further constraints, an *actual* basis for $K_{2}$ is 

$$
\boxed{Span\left\{-4+24x-36x^{2}+16x^{3},-2+8x-10x^{2}+4x^{3}\right\}}.\quad(17)
$$

Putting our results $(14)$ and $(17)$ for $K_{1}$ and $K_{2}$ together, a possible basis $\beta:=\{\varphi_{i}\}$ for all of $V_{h}$ is

$$
\boxed{\beta=Span\left\{\phi_{1}(x),\phi_{2}(x)\right\}},\quad(17)
$$

where

$$
\boxed{\phi_{1}(x)=\left\{\begin{matrix} 12x^{2}-16x^{3} \\ -4+24x-36x^{2}+16x^{3} \end{matrix}\quad\begin{matrix} 0\leq x\leq \frac{1}{2}  \\ \frac{1}{2}\leq x\leq1 \end{matrix}\right.,\quad\phi_{2}(x)=\left\{\begin{matrix} -2x^{2}+4x^{3} \\ -2+8x-10x^{2}+4x^{3}\end{matrix}\quad\begin{matrix} 0\leq x\leq \frac{1}{2} \\ \frac{1}{2}\leq x\leq1 \end{matrix}\right.}.\quad(18)
$$

**Part (c)**

First off, we solve the BVP given by $(1)$ and $(2)$ for the true solution. We do this by using an ansatz (i.e. guessing) that $u$ takes the form of a fifth-order polynomial $u(x)=a_{0}+a_{1}x+a_{2}x^{2}+a_{3}x^{3}+a_{4}x^{4}+a_{5}x^{5}$. Why? Because taking four derivatives of a fifth-order term will give something that is first-order (linear), and $f(x)$ is first-order (linear). 

We can match coefficients from here. At first, we substitute our ansatz into $(1)$.

$$
\frac{d^{4}}{dx^{4}}\left(a_{0}+a_{1}x+a_{2}x^{2}+a_{3}x^{3}+a_{4}x^{4}+a_{5}x^{5}\right)=4!a_{4}+5!a_{5}x=24a_{4}+120a_{5}=480x-120\Rightarrow a_{4}=-\frac{120}{24}=-5,\quad a_{5}=\frac{480}{120}=4.
$$

Therefore, we so far have that $u(x)=a_{0}+a_{1}x+a_{2}x^{2}+a_{3}x^{3}-5x^{4}+4x^{5}$. We now match the boundary conditions.

$$
u(0)=0\Rightarrow u(0)=a_{0}=0,\quad u'(0)=0\Rightarrow u'(0)=a_{1}=0,\\
u(1)=0\Rightarrow u(1)=a_{2}+a_{3}-5+4=0\Rightarrow a_{2}+a_{3}=1,\quad u'(1)=0\Rightarrow u'(1)=2a_{2}+3a_{3}-20+20=0\Rightarrow2a_{2}+3a_{3}=0.\quad(19)
$$

$(19)$ gives us a system of linear equations for the remaining two coefficients, the solution of which is $a_{2}=3$ and $a_{3}=-2$. Therefore, the exact solution to our BVP is

$$
\boxed{u(x)=3x^{2}-2x^{3}-5x^{4}+4x^{5}}.\quad(20)
$$

Next, we use the basis derived in part (b) to find an approximate solution to $(1),(2)$ via the Galerkin formula derived in part 9a). Firstly, note that $(9)$ should hold for all $v=\phi_{i}\in V_{h},\hspace{0.25cm}i=1,2$. Thus,

$$
\int_{0}^{1}{u_{h}''(x)\phi_{i}''(x)dx}=\int_{0}^{1}{f(x)\phi_{i}(x)dx}.\quad(21)
$$

The approximate solution $u_{h}(x)$ to $u(x)$ can be written as a linear combination of elements in $V_{h}$, or 

$$
u_{h}(x)=u_{1}\phi_{1}(x)+u_{2}\phi_{2}(x)\Rightarrow u_{h}''(x)=\frac{d^{2}}{dx^{2}}\left[u_{1}\phi_{1}(x)+u_{2}\phi_{2}(x)\right]=u_{1}\phi_{1}''(x)+u_{2}\phi_{2}''(x).\quad(22)
$$

Then, substituting $(22)$ into $(21)$,

$$
\int_{0}^{1}{\left[u_{1}\phi_{1}''(x)+u_{2}\phi_{2}''(x)\right]\phi_{i}''(x)dx}=u_{1}\int_{0}^{1}{\phi_{1}''(x)\phi_{i}''(x)dx}+u_{2}\int_{0}^{1}{\phi_{2}''(x)\phi_{i}''(x)dx}=\int_{0}^{1}{f(x)\phi_{i}(x)dx},\quad(23)
$$

which must be solved for each $i=1,2$. Now, observe that $(23)$ can be recast as a matrix equation of the form $A\mathbf{u}=\mathbf{f}$, where

$$
A_{ij}=\int_{0}^{1}{\phi_{i}''(x)\phi_{j}''(x)dx},\quad\mathbf{u}=\begin{pmatrix} u_{1} \\ u_{2} \end{pmatrix},\quad f_{i}=\int_{0}^{1}{f(x)\phi_{i}(x)dx}=\int_{0}^{1}{\left(480x-120\right)\phi_{i}(x)dx}.\quad(24)
$$

Also, note that 

$$
\phi_{1}''(x)=\left\{\begin{matrix} 24-96x \\ 96x-72 \end{matrix}\quad\begin{matrix} 0\leq x\leq \frac{1}{2}  \\ \frac{1}{2}\leq x\leq1 \end{matrix}\right.,\quad\phi_{2}''(x)=\left\{\begin{matrix} 24x-4 \\ 24x-20\end{matrix}\quad\begin{matrix} 0\leq x\leq \frac{1}{2} \\ \frac{1}{2}\leq x\leq1 \end{matrix}\right.\quad(25)
$$

We now solve for each of components of this matrix equation, since we have enough information to construct $A$ and $\mathbf{f}$ using $(18)$ which gave us $(25)$, and $(1)$. 

$$
A_{11}=\int_{0}{1}{\phi_{1}''(x)\phi_{1}''(x)dx}=\int_{0}^{\frac{1}{2}}{\left(24-96x\right)^{2}dx}+\int_{\frac{1}{2}}^{1}{\left(96x-72\right)^{2}dx}=96+96=192.\quad(26)\\
A_{12}=A_{21}=\int_{0}{1}{\phi_{1}''(x)\phi_{2}''(x)dx}=\int_{0}^{\frac{1}{2}}{\left(24-96x\right)\left(24x-4\right)dx}+\int_{\frac{1}{2}}^{1}{\left(96x-72\right)\left(24x-20\right)dx}=-24+24=0.\quad(27)\\
A_{22}=\int_{0}{1}{\phi_{2}''(x)\phi_{2}''(x)dx}=\int_{0}^{\frac{1}{2}}{\left(24x-4\right)^{2}dx}+\int_{\frac{1}{2}}^{1}{\left(24x-20\right)^{2}dx}=8+8=16.\quad(28)\\
f_{1}=\int_{0}^{1}{f(x)\phi_{1}(x)dx}=\int_{0}^{\frac{1}{2}}{\left(480x-120\right)\left(12x^{2}-16x^{3}\right)dx}+\int_{\frac{1}{2}}^{1}{\left(480x-120\right)\left(-4+24x-36x^{2}+16x^{3}\right)dx}=12+48=60.\quad(29)\\
f_{2}=\int_{0}^{1}{f(x)\phi_{2}(x)dx}=\int_{0}^{\frac{1}{2}}{\left(480x-120\right)\left(-2x^{2}+4x^{3}\right)dx}+\int_{\frac{1}{2}}^{1}{\left(480x-120\right)\left(-2+8x-10x^{2}+4x^{3}\right)dx}=-\frac{1}{2}+\frac{9}{2}=4.\quad(30)
$$

Substituting $(26)-(30)$ into $(24)$,

$$
A=\begin{bmatrix} 192 & 0 \\ 0 & 16 \end{bmatrix}\Rightarrow A^{-1}=\begin{bmatrix} \frac{1}{192} & 0 \\ 0 & \frac{1}{16} \end{bmatrix},\quad\mathbf{f}=\begin{bmatrix} 60 \\ 4 \end{bmatrix}\\
\Rightarrow\mathbf{u}=A^{-1}\mathbf{f}=\begin{bmatrix} \frac{1}{192} & 0 \\ 0 & \frac{1}{16} \end{bmatrix}\begin{bmatrix} 60 \\ 4 \end{bmatrix}=\begin{bmatrix} \frac{60}{192} \\ \frac{4}{16} \end{bmatrix}=\begin{bmatrix} \frac{5}{16} \\ \frac{1}{4} \end{bmatrix}.\quad(31)
$$

Therefore, by $(31)$, the numerical (approximate) solution $u_{h}$ to $(1),(2)$ using the Galerkin method we derived here is

$$
\boxed{u_{h}(x)=\left\{\begin{matrix} \frac{5}{16}\left(12x^{2}-16x^{3}\right)+\frac{1}{4}\left(-2x^{2}+4x^{3}\right) \\ \frac{5}{16}\left(-4+24x-36x^{2}+16x^{3}\right)+\frac{1}{4}\left(-2+8x-10x^{2}+4x^{3}\right) \end{matrix}\quad\begin{matrix} 0\leq x\leq \frac{1}{2}  \\ \frac{1}{2}\leq x\leq1 \end{matrix}\right.}.
$$

or, with each term put in a nicer, more compact form,

$$
\boxed{u_{h}(x)=\left\{\begin{matrix} \frac{1}{4}x^{2}\left(13-16x\right) \\ \frac{1}{4}\left(x-1\right)^{2}\left(24x-7\right) \end{matrix}\quad\begin{matrix} 0\leq x\leq \frac{1}{2}  \\ \frac{1}{2}\leq x\leq1 \end{matrix}\right.}.\quad(32)
$$

We now plot our numerical solution $(32)$ alongside the exact solution (which we solved for in $(20)$) and compare them.

```julia
using PyPlot

#set up point grid
h = 0.01
xq = 0 : h : 1
first = 0 : h : 1 / 2
second = 1 / 2 + h : h : 1

#evaluate exact and approximate solutions on the grid
uexact = @. 3 * xq ^ 2 - 2 * xq ^ 3 - 5 * xq ^ 4 + 4 * xq ^ 5
ugalerkinfirst = @. 0.25 * first ^ 2 * (13 - 16 * first)
ugalerkinsecond = @. 0.25 * (second - 1) ^ 2 * (24 * second - 7)
ugalerkin = vcat(ugalerkinfirst, ugalerkinsecond)

#calculate the mean squared error
mse = h * sum((uexact - ugalerkin) .^ 2)
println("The mean squared error (MSE) between the exact solution and our approximate solution is ", mse, ".")

#plot exact and approximate solutions
figure(1, figsize = (5, 5))
plot(xq, uexact, "--", label = L"u(x)")
plot(xq, ugalerkin, "green", label = L"u_{h}(x)")
legend()
```

As observable on the graph, the MSE was only $\approx 3.106\times 10^{-4}$, which is quite small, even relative to the magnitude of $u$ in the first place. We even see qualitatively how $u_{h}$ tracks $u$ virtually perfectly up to around the middle of the domain, then begins to deviate, but not by a very large margin. It still meets the exact solution quite nicely towards the right boundary of the domain. What is especially interesting to note is how $u_{h}$ hardly satisfies our original PDE at all if we just naively substituted it in, but *still* manages to predict its behavior well. Of course though, it exactly matches the boundary conditions. That is the glory of weak (variational) solutions to PDEs, which form the theoretical basis for finite element methods. $\boxed{}$

## Problem 2

We start by calling in various mesh utility functions, as supplied by Professor Per-Olof Persson and the GSI, Frank (Jingyi) Wang. These will be essential in defining `fempoi`, and later for `poiconv` as well.

```julia
# Various mesh utilities, mostly for unstructured triangular meshes
#
# UC Berkeley Math 228B, Per-Olof Persson <persson@berkeley.edu>

using PyPlot, PyCall

"""
    t = delaunay(p)

Delaunay triangulation `t` of N x 2 node array `p`.
"""
function delaunay(p)
    tri = pyimport("matplotlib.tri")
    t = tri[:Triangulation](p[:,1], p[:,2])
    return Int64.(t[:triangles] .+ 1)
end

"""
    edges, boundary_indices = all_edges(t)

Find all unique edges in the triangulation `t` (ne x 2 array)
Second output is indices to the boundary edges.
"""
function all_edges(t)
    edges = vcat(t[:,[1,2]], t[:,[2,3]], t[:,[3,1]])
    edges = sortslices(sort(edges, dims=2), dims=1)
    dup = all(edges[2:end,:] - edges[1:end-1,:] .== 0, dims=2)[:]
    keep = .![false;dup]
    edges = edges[keep,:]
    dup = [dup;false]
    dup = dup[keep]
    return edges, findall(.!dup)
end

"""
    e = boundary_nodes(t)

Find all boundary nodes in the triangulation `t`.
"""
function boundary_nodes(t)
    edges, boundary_indices = all_edges(t)
    return unique(edges[boundary_indices,:][:])
end

"""
    tplot(p, t, u=nothing)

If `u` == nothing: Plot triangular mesh with nodes `p` and triangles `t`.
If `u` == solution vector: Plot filled contour color plot of solution `u`.
"""
function tplot(p, t, u=nothing)
    clf()
    axis("equal")
    if u == nothing
        tripcolor(p[:,1], p[:,2], t .- 1, 0*t[:,1],
                  cmap="Set3", edgecolors="k", linewidth=1)
    else
        tricontourf(p[:,1], p[:,2], t .- 1, u, 20)
    end
    draw()
end

"""
    inside = inpolygon(p, pv)

Determine if each point in the N x 2 node array `p` is inside the polygon
described by the NE x 2 node array `pv`.
"""
function inpolygon(p::Array{Float64,2}, pv::Array{Float64,2})
    path = pyimport("matplotlib.path")
    poly = path[:Path](pv)
    inside = [poly[:contains_point](p[ip,:]) for ip = 1:size(p,1)]
end

function test_inpolygon()
    pv = Float64[0 0; 1 0; 1 1; 0 1; 0.5 0.6; 0 0]
    x = collect(-.15:0.1:1.15)
    p = [x 0*x.+0.5]
    inside = inpolygon(p, pv)
    clf()
    plot(pv[:,1], pv[:,2],
         p[inside,1], p[inside,2], "*",
         p[.!inside,1], p[.!inside,2], "o")
    return
end

# Delaunay refinement mesh generator.
#
# UC Berkeley Math 228B, Per-Olof Persson <persson@berkeley.edu>

function remove_outside_tris(p, t, pv)
    pmid = dropdims(sum(p[t,:], dims=2), dims=2) / 3
    is_inside = inpolygon(pmid, pv)
    t = t[is_inside,:]
end

function triarea(p, t)
    d12 = @. p[t[:,2],:] - p[t[:,1],:]
    d13 = @. p[t[:,3],:] - p[t[:,1],:]
    @. abs(d12[:,1] * d13[:,2] - d12[:,2] * d13[:,1]) / 2
end

function remove_tiny_tris(p, t)
    t = t[triarea(p,t) .> 1e-14,:]
end

function circumcenter(p)
    dp1 = @. p[2,:] - p[1,:]
    dp2 = @. p[3,:] - p[1,:]

    mid1 = @. (p[1,:] + p[2,:]) / 2
    mid2 = @. (p[1,:] + p[3,:]) / 2

    s = [ -dp1[2] dp2[2]; dp1[1] -dp2[1]] \ (-mid1 .+ mid2)
    pc = @. mid1' + s[1] * [-dp1[2] dp1[1]]
end

function edge_midpoints(p, t)
    pmid = reshape(p[t,:] + p[t[:,[2,3,1]],:], :, 2) / 2
    pmid = unique(pmid, dims=1)
end

function pmesh(pv, hmax, nref)
    p = zeros(Float64, 0, 2)
    for i = 1:size(pv,1) - 1
        pp = pv[i:i+1,:]
        L = sqrt(sum(diff(pp, dims=1).^2, dims=2))[1]
        if L > hmax
            n = ceil(Int, L / hmax)
            ss = (0:n) / n
            pp = [1 .- ss ss] * pp
        end
        p = [p; pp[1:end-1,:]]
    end

    t = zeros(Int64, 0, 3)
    while true
        t = delaunay(p)
        t = remove_tiny_tris(p, t)
        t = remove_outside_tris(p, t, pv)
        # tplot(p,t), pause(1e-3)
        
        area = triarea(p, t)
        maxarea, ix = findmax(area)
        if maxarea < hmax^2 / 2
            break
        end
        pc = circumcenter(p[t[ix,:],:])
        p = [p; pc]
    end

    for iref = 1:nref
        p = [p; edge_midpoints(p, t)]
        t = delaunay(p)
        t = remove_tiny_tris(p, t)
        t = remove_outside_tris(p, t, pv)
        # tplot(p, t), pause(1e-3)
    end
    
    e = boundary_nodes(t)
    p, t, e
end

function pmesh_test()
    pv = [0 0; 1 0; .5 .5; 1 1; 0 1; 0 0]
    p,t,e = pmesh(pv, 0.2, 1);
    tplot(p,t)
end
```

Next, we want to consider the Poisson problem in two dimensions. Namely, we intend to numerically solve

$$
-\nabla^{2}u(x,y)=1\quad \forall(x,y)\in\Omega\quad(33)\\
n\cdot\nabla u=0\quad \forall(x,y)\in\Gamma_{1}\quad(34)\\
u=0\quad \forall(x,y)\in\Gamma_{2},\quad(35)
$$

where $\Gamma_{1}\cup\Gamma_{2}=\partial\Omega$ and in our code, all points on $\Gamma_{2}$ are contained in a given array `e`. Now, we want to eventually code a finite element method which is robust, and therefore works for any *reasonable* arbitrary domain $\Omega$, and we want to be able to mix up boundary conditions as desired. 

We define the space of piecewise linear continuous functions on some arbitrary, given mesh $T_{h}$:

$$
V_{h}:=\left\{v\in C^{0}\left(\Omega\right):\left.v\right|_{K}\in\mathbb{P}_{1}\left(K\right)\hspace{0.1cm}\forall K\in T_{h}\right\}.
$$

Suppose we want some numerical grid solution solution $u_{h}$ which is a linear combination of functions in $V_{h}$. Then, multiplying both sides of $(33)$ by a test function $v\in V_{h}$ and integrating both sides over the region on which our BVP is defined,

$$
\iint_{\Omega}{-\left(\nabla^{2}u_{h}\right)vdA}=\iint_{\Omega}{vdA}.\quad(36)
$$

Applying Green's first identity to the lefthand side of $(36)$, we derive the Galerkin formulation:

$$
\iint_{\Omega}{\nabla u_{h}\cdot \nabla vdA}=\iint_{\Omega}{vdA}+\oint_{\partial\Omega}{v\frac{\partial u_{h}}{\partial n}ds}=\iint_{\Omega}{vdA}+\oint_{\partial\Omega}{vn\cdot\nabla uds}=\iint_{\Omega}{vdA}+0\Rightarrow\boxed{\iint_{\Omega}{\nabla u_{h}\cdot\nabla vdA}=\iint_{\Omega}{vdA}}.\quad(37)
$$

Next, suppose that we have chosen a basis for $V_{h}$, i.e. $\left\{\varphi_{i}(x,y)\right\}$. Then, since $u_{h}\in V_{h}$, $u_{h}$ can be written as a linear combination of elements in this basis, or $u_{h}=\sum_{i}{u_{hi}\varphi_{i}}$. Furthermore, note that $(37)$ must apply to any element of $V_{h}$, so consider $(37)$ for a particular test function $v_{j}\in V_{h}$. Then, substituting these back into $(37)$,

$$
\iint_{\Omega}{\nabla u_{h}\cdot\nabla vdA}=\iint_{\Omega}{\nabla\left(\sum_{i}{u_{hi}\varphi_{i}}\right)\cdot\nabla v_{j}dA},
$$

and switching the integral and sum,

$$
\iint_{\Omega}{\nabla u_{h}\cdot\nabla vdA}=\sum_{i}{u_{hi}\iint_{\Omega}{\nabla\varphi_{i}\cdot\nabla v_{j}dA}}=\boxed{\sum_{i}{u_{hi}\iint_{\Omega}{\nabla\varphi_{i}\cdot\nabla\varphi_{j}dA}}=\iint_{\Omega}{vdA}}.\quad(38)
$$

$(38)$ is in the form of a finite element formulation, and can be turned into a matrix equation of the form $A\mathbf{u}=\mathbf{b}$, where

$$
A_{ij}=u_{hi}\iint_{\Omega}{\nabla\varphi_{i}\cdot\nabla\varphi_{j}dA},\quad\mathbf{u}=\begin{bmatrix} u_{1} \\ \vdots \\ u_{n}\end{bmatrix},\quad b_{j}=\iint_{\Omega}{vdA}.\quad(39)
$$

Furthermore, since our functoin space $V_{h}$ consists solely of piecewise linear functions, $\nabla\varphi_{i}=\left<c_{i,x},c_{i,y}\right>\forall\varphi_{i}\in V_{h}$ for some constants $c_{i,x},c_{i,y}$. Therefore, $\nabla\varphi_{i}\cdot\nabla\varphi_{j}=\left<c_{i,x},c_{i,y}\right>\cdot\left<c_{j,x},c_{j,y}\right>=c_{i,x}c_{j,x}+c_{i,y}c_{j,y}$ (so, a dot product between constant vectors).

From here, we can define a sets of local basis functions (in this case, linear functions) on each triangular element, where triangulations are generated using the usual `pmesh` function. I would use my own, but as suggested, this might run into problems with certain, overly concave geometries using the somewhat crude method we devised on the previous problem sets, so I used the GSI's that he supplied. Anyways, thus, we can construct an elementary matrix and elementary load, and use these to construct a global matrix $A$ and global load $\mathbf{b}$ to solve the matrix equation described in $(39)$ via the stamping method. I would elaborate upon this, but it is all described in Professor Per-Olof Persson's FEM notes, as provided on bCourses. I will, however, put comments in my code to indicate how my code follows the method outlined there.

```julia
using LinearAlgebra
using PyPlot

function fempoi(p, t, e) 
    areas = triarea(p, t)
    n = length(p[:, 1]) #length of u. 
    #Since u is defined on each element, we just need the number of elements (triangles) in our triangulation, which is this.
    A = zeros(n, n) #preallocate a global matrix to be stamped
    b = zeros(n) #preallocate a global load vector to be stamped
    e1 = [1; 0; 0] #define the first standard basis vector
    e2 = [0; 1; 0] #"" second ""
    e3 = [0; 0; 1] #"" third ""
    for k = 1 : length(t[:, 1])
        #locate vertices of each triangle in our triangulation
        indices = t[k, :]
        x1 = p[indices[1], 1] #extract x-coordinate of first vertex
        x2 = p[indices[2], 1] #"" second vertex
        x3 = p[indices[3], 1] #"" third vertex
        y1 = p[indices[1], 2] #extract y-coordinate of first vertex
        y2 = p[indices[2], 2] #"" second vertex
        y3 = p[indices[3], 2] #"" third vertex
        #Vandermonde matrix to be inverted
        V = [1 x1 y1; 1 x2 y2; 1 x3 y3]
        area = areas[k] 
        coeff1 = V \ e1 #get coefficient for constant part of linear function on local element
        coeff2 = V \ e2 #get coefficient for x term of ""
        coeff3 = V \ e3 #get coefficient for y term of ""
        #Calculate the entries of the elementary matrix
        A_11 = dot(coeff1[2 : 3], coeff1[2 : 3]) 
        A_12 = dot(coeff1[2 : 3], coeff2[2 : 3]) 
        A_13 = dot(coeff1[2 : 3], coeff3[2 : 3]) 
        A_22 = dot(coeff2[2 : 3], coeff2[2 : 3]) 
        A_23 = dot(coeff2[2 : 3], coeff3[2 : 3]) 
        A_33 = dot(coeff3[2 : 3], coeff3[2 : 3]) 
        #create elementary matrix
        A_k = Symmetric(area * [A_11 A_12 A_13; A_12 A_22 A_23; A_13 A_23 A_33])
        #create elementary load
        b_k = ones(3) * area / 3 
        #stamp elementary matrix to global matrix
        A[indices, indices] += A_k
        #stamp elementary load to global load vector
        b[indices] += b_k
    end
    #impose Dirichlet conditions on nodes in `e`
    for i = 1 : length(e)
        index = e[i]
        A[index, :] = zeros(n)
        A[index, index] = 1
        b[index] = 0
    end
    #calculate global values of u
    u = A \ b
    return u
end
```

We proceed by testing out our `fempoi` solver on the sample cases provided.

```julia
# Square, Dirichlet left/bottom
pv = Float64[0 0; 1 0; 1 1; 0 1; 0 0]
p, t, e = pmesh(pv, 0.15, 0)
e = e[@. (p[e,1] < 1e-6) | (p[e,2] < 1e-6)]
u = fempoi(p, t, e)
figure(1, figsize = (4, 4))
tplot(p, t, u)
colorbar()

# Circle, all Dirichlet
n = 32; phi = 2pi*(0:n)/n
pv = [cos.(phi) sin.(phi)]
p, t, e = pmesh(pv, 2pi/n, 0)
u = fempoi(p, t, e)
figure(2, figsize = (4, 4))
tplot(p, t, u)
colorbar()

# Generic polygon geometry, mixed Dirichlet/Neumann
x = 0:.1:1
y = 0.1*(-1).^(0:10)
pv = [x y; .5 .6; 0 .1]
p, t, e = pmesh(pv, 0.04, 0)
e = e[@. p[e,2] > (.6 - abs(p[e,1] - 0.5) - 1e-6)]
u = fempoi(p, t, e)
figure(3, figsize = (4, 4))
tplot(p, t, u)
colorbar()
```

We find that the solver recreates the provided plots identically (or at least from a visual standpoint), as desired. $\boxed{}$

## Problem 3

We modify our `fempoi` solver slightly for this problem and define a new function `alldirichlet`. `alldirichlet` instead has `pmesh` built in, such that `pv` (the vertices of our initial polygon), `hmax` (the maximum allowed stepsize for nodes to be placed), and `nref` (the number of refinements) can be taken as inputs for `alldirichlet`. This will allow us to modify the number of refinements, `nref`, in a for loop where we only will need to call `alldirichlet` each time. Plus, since we are imposing Dirichlet conditions on *all* boundary nodes of our mesh (which is what motivated the title of our new function), we can use `e` directly, which gives all the boundary nodes.

From here, we also write a function `poiconv` which returns a vector of max norm errors between our numerical solution, $u_{h}$ and the exact solution, $u$ as the number of mesh refinements goes from $0$ to `nrefmax` (which is given). Since we may not have the exact solution, we take the "exact solution" to approximately be the numerical solution with the maximum possible value for `hmax`. 

```julia
using LinearAlgebra
using PyPlot

function alldirichlet(pv, hmax, nref) 
    p, t, e = pmesh(pv, hmax, nref)
    areas = triarea(p, t)
    n = length(p[:, 1])
    A = zeros(n, n)
    b = zeros(n)
    e1 = [1; 0; 0]
    e2 = [0; 1; 0]
    e3 = [0; 0; 1]
    for k = 1 : length(t[:, 1])
        indices = t[k, :]
        x1 = p[indices[1], 1]
        y1 = p[indices[1], 2]
        x2 = p[indices[2], 1]
        y2 = p[indices[2], 2]
        x3 = p[indices[3], 1]
        y3 = p[indices[3], 2]
        V = [1 x1 y1; 1 x2 y2; 1 x3 y3]
        area = areas[k] 
        coeff1 = V \ e1
        coeff2 = V \ e2
        coeff3 = V \ e3
        A_11 = dot(coeff1[2 : 3], coeff1[2 : 3]) 
        A_12 = dot(coeff1[2 : 3], coeff2[2 : 3]) 
        A_13 = dot(coeff1[2 : 3], coeff3[2 : 3]) 
        A_22 = dot(coeff2[2 : 3], coeff2[2 : 3]) 
        A_23 = dot(coeff2[2 : 3], coeff3[2 : 3]) 
        A_33 = dot(coeff3[2 : 3], coeff3[2 : 3]) 
        A_k = Symmetric(area * [A_11 A_12 A_13; A_12 A_22 A_23; A_13 A_23 A_33])
        b_k = ones(3) * area / 3 
        A[indices, indices] += A_k
        b[indices] += b_k
    end
    #impose homogeneous Dirichlet conditions on ALL boundary nodes (which are contained in `e`)
    for i = 1 : length(e)
        index = e[i]
        A[index, :] = zeros(n)
        A[index, index] = 1
        b[index] = 0
    end
    u = A \ b
    return u
end

function poiconv(pv, hmax, nrefmax)
    nref = 0 : nrefmax
    #calculate finest solution
    ufinest = alldirichlet(pv, hmax, nrefmax)
    #calculate first solution to start us off with no grid refinements 
    ublurry = alldirichlet(pv, hmax, 0)
    #set the cutoff index. 
    #Regardless of how many times we have refined, up to this index, the u values will correspond to the same grid points.
    cutoff = length(ublurry)
    #truncate the finest solution for u (approximately the exact solution) for the grid points that are shared with the
    #solution corresponding to no refinements
    ufinestforcomparison = ufinest[1 : cutoff]
    #preallocate vector, in which we will store our errors
    errors = zeros(nrefmax)
    #calculate the first error between the grid solution with no refinements and the "exact" solution
    errors[1] = norm(ufinestforcomparison - ublurry, Inf)
    
    #calculate a numerical solution for successively more grid refinements, storing the error between the exact solution
    #and this one each time
    for i = 2 : length(nref) - 1
        u = alldirichlet(pv, hmax, nref[i])
        uforcomparison = u[1 : cutoff]
        errors[i] = norm(ufinestforcomparison - uforcomparison, Inf) 
    end
    #return a vector of errors
    return errors
end
```

Next, we assume that the global error in our finite element formulation depends on the stepsize in some kind of power law, i.e. $\left||u-u_{h}|\right|_{max}\approx Ch^{p}$, where $u$ is the exact solution, $u_{h}$ is the approximate (numerical) solution, $h$ is some one-dimensional spatial quantity (say, the average spacing between neighboring nodes) and $C,p$ constants greater than zero. This is the usual error analysis for these numerical methods. Then, by plotting this max norm error as $h$ varies in a log-log plot, the slope of the resulting graph will give us $p$, because 

$$
\left||u-u_{h}|\right|_{max}\approx Ch^{p}\Rightarrow\log\left||u-u_{h}|\right|_{max}\approx \log\left(Ch^{p}\right)=\log C+\log h^{p}=\log C+p\log h.
$$

```julia
hmax = 0.15
pv_square = Float64[0 0; 1 0; 1 1; 0 1; 0 0]
pv_polygon = Float64[0 0; 1 0; .5 .5; 1 1; 0 1; 0 0]

errors_square = poiconv(pv_square, hmax, 3)
errors_polygon = poiconv(pv_polygon, hmax, 3)
errors = [errors_square errors_polygon]

clf()
loglog(hmax ./ [1,2,4], errors)
rates = @. log2(errors[end-1,:]) - log2(errors[end,:])
```

In both cases, the estimated slope of the log-log plot for our max norm errors is roughly two ($\approx 1.9307881504307804$ for the square and $\approx 1.8378056929714415$ for the polygon given in the third problem set), so this means that our FEM formulation used for our Poisson problem with mixed-Dirichlet/Neumann boundary conditions is roughly second-order. It is actually a little bit under two, which may have something to do with how we used an approximate solution (albeit, the one with the largest number of grid refinements) as our "exact" solution for comparison. $\boxed{}$
