**Part (b)**

We can (hopefully) assume that the matrix `pv` is in the correct order and format. Then, we can insert nodes along each polygon segment by looking at consecutive points $(x_{i},y_{i}),(x_{i+1},y_{i+1})\in$ `pv`and how to insert equally-spaced subnodes $(x_{i+\frac{1}{n_{i}}},y_{i+\frac{1}{n_{i}}},\dots,x_{i+\frac{n_{i}-1}{n_{i}}},y_{i+\frac{n_{i}-1}{n_{i}}})$ between them. These should be chosen such that the spacing between them is $h_{i}=\max\{h:h\leq h_{max},\frac{d_{i}}{h}=n_{i}\}$ (the latter condition ensures that the spacing is actually uniform along each segment), where $d_{i}$ is the distance between $(x_{i},y_{i})$ and $(x_{i+1},y_{i+1})$, or $d_{i}=\sqrt{\left(x_{i+1}-x_{i}\right)^{2}+\left(y_{i+1}-y_{i}\right)^{2}}\hspace{0.1cm}\forall i$.

Therefore, we should have that $\frac{d_{i}}{n_{i}}\leq h_{max}\Rightarrow$, and since $h_{i}=\frac{d_{i}}{n_{i}}\leq h_{max}$ should be as great as possible, we require that $\boxed{n_{i}=\lceil \frac{d_{i}}{h_{max}} \rceil}$. Thus, $\boxed{h_{i}=\frac{d_{i}}{\lceil \frac{d_{i}}{h_{max}} \rceil}}$, and the coordinates of each subnode should be

$$
x_{i+\frac{k}{n_{i}}}=x_{i}+\frac{k}{n_{i}}\left(x_{i+1}-x_{i}\right),\hspace{0.25cm}k\in\{0,1,\dots,n_{i}-1,n_{i}\} \\
y_{i+\frac{k}{n_{i}}}=y_{i}+\frac{k}{n_{i}}\left(y_{i+1}-y_{i}\right),\hspace{0.25cm}k\in\{0,1,\dots,n_{i}-1,n_{i}\}.
$$

Of course, we can use `unique` to remove any duplicate nodes.

```julia
using LinearAlgebra
using PyPlot

function nodecreator(pv, hmax)
    #set up initial point
    pwnodes = reshape(pv[1, :], 1, 2)
    for i = 1 : length(pv[:, 1]) - 1 #loop through each polygon segment
        ri = pv[i, :]
        rf = pv[i + 1, :]
        diff = rf - ri #vector difference between successive points
        L = norm(diff) #the distance between successive points
        k = convert(Int64, ceil(L / hmax)) #the number of node points needed along this given segment
        diffs = reshape(repeat(diff, k), 2, k)' #create a vector of vector differences
        newnodes = reshape(repeat(ri, k), 2, k)' + repeat(1 : k, 1, 2) .* diffs / k #calculate subnodes along segment
        pwnodes = vcat(pwnodes, newnodes) #add subnodes 
    end
    pwnodes = unique(pwnodes, dims = 1) #no duplicates
    return pwnodes
end

pwnodes = nodecreator(pv, 0.2)
figure(1, figsize = (4, 4))
scatter(pwnodes[:, 1], pwnodes[:, 2])
```

**Part (c)**

We load the `delaunay` function from the mesh utilities document (provided by Professor Persson himself) to help us triangulate the domain. Also, we write the `tplot` function, which will make it easier when we want to plot our triangulation.

```julia
using PyPlot, PyCall

"""
    t = delaunay(p)

Delaunay triangulation `t` of N x 2 node array `p`.
"""
function delaunay(p)
    tri = pyimport("matplotlib.tri")
    t = tri[:Triangulation](p[:,1], p[:,2])
    return Int64.(t[:triangles] .+ 1)
end

"""
    tplot(p, t, u=nothing)

If `u` == nothing: Plot triangular mesh with nodes `p` and triangles `t`.
If `u` == solution vector: Plot filled contour color plot of solution `u`.
"""
function tplot(p, t, u=nothing)
    clf()
    axis("equal")
    if u == nothing
        tripcolor(p[:,1], p[:,2], t .- 1, 0*t[:,1],
                  cmap="Set3", edgecolors="k", linewidth=1)
    else
        tricontourf(p[:,1], p[:,2], t .- 1, u, 20)
    end
    draw()
end

using PyPlot

pv = [0 0; 1 0; .5 .5; 1 1; 0 1; 0 0]
pwnodes = nodecreator(pv, 0.2)
t = delaunay(pwnodes)
figure(1, figsize = (4, 4))
tplot(pwnodes, t)
```

**Part (d)**

Unfortunately, the `delaunay` function also creates triangles in areas outside of the polygon as well. To remedy this, we calculate the centroids of each of the triangles created through this triangulation using the `centroid` function defined below. Then, using the `inpolygon` function from mesh utilities, we can compare with the original 2-column matrix `pv` to check if these centroids lie inside of the polygon. If they do not, then these correspond to triangles outside of the polygon's domain, which must be removed. Removing these is pretty simple, because `inpolygon` returns the row indices in `t` with corresponding row vectors that contain the vertex indices for each triangle *inside* the polygon.

```julia
using PyPlot

"""
    inside = inpolygon(p, pv)

Determine if each point in the N x 2 node array `p` is inside the polygon
described by the NE x 2 node array `pv`.
"""
function inpolygon(p::Array{Float64,2}, pv::Array{Float64,2})
    path = pyimport("matplotlib.path")
    poly = path[:Path](pv)
    inside = [poly[:contains_point](p[ip,:]) for ip = 1:size(p,1)]
end
```

```julia
using PyPlot

pv = [0 0; 1 0; .5 .5; 1 1; 0 1; 0 0]
pwnodes = nodecreator(pv, 0.2)
t = delaunay(pwnodes)

function centroid(t, pwnodes)
    centroids = zeros(length(t[:, 1]), 2)
    for i = 1 : length(t[:, 1])
        #locate vertices of each triangle in our triangulation
        vertices = pwnodes[t[i, :], :]
        #calculate each centroid
        centroids[i, :] = [(vertices[1, 1] + vertices[2, 1] + vertices[3, 1]) / 3 (vertices[1, 2] + vertices[2, 2] + vertices[3, 2]) / 3]
    end
    return centroids
end

function filteroutside(t, pwnodes, pv)
    centroids = centroid(t, pwnodes) #locate centroids
    inside = inpolygon(centroids, pv) #check triangles with centroids inside polygon
    tfiltered = t[inside, :] #filter out outside triangles
    return tfiltered
end

t = filteroutside(t, pwnodes, pv)
figure(1, figsize = (4, 4))
tplot(pwnodes, t)
```

**Part (e)**

Below, we write code which calculates the area for each triangle with vertices represented in each row vector for `t`, given `t` and `pwnodes`, which gives the spatial coordinates for each node. Then, we follow the steps as given: check if the triangle with largest area $A$ has an $A$ such that $A>\frac{h_{max}^{2}}{2}$. If so, we add its circumcenter to the list of node points. We also write code which can calculate the circumcenter given a vector of areas which correspond to each row vector of `t`, as well as `pwnodes` for how these relate to the spatial coordinates of the nodes.

**Part (f), Part (g)**

We retriangulate and remove outside triangles after each extra circumcenter node we add. This repeats until no more triangles have an area $>\frac{h_{max}^{2}}{2}$.

```julia
using LinearAlgebra
using PyPlot

pv = [0 0; 1 0; .5 .5; 1 1; 0 1; 0 0]
hmax = 0.2
pwnodes = nodecreator(pv, hmax)
t = filteroutside(delaunay(pwnodes), pwnodes, pv) 

function area(t, pwnodes)
    areas = zeros(length(t[:, 1]))
    for i = 1 : length(t[:, 1])
        #locate vertices of each triangle in our triangulation
        vertices = pwnodes[t[i, :], :]
        #calculate each area
        a = norm(vertices[2, :] - vertices[1, :])
        b = norm(vertices[3, :] - vertices[2, :])
        c = norm(vertices[1, :] - vertices[3, :])
        p = (a + b + c) / 2
        areas[i] = (p * (p - a) * (p - b) * (p - c)) ^ (1 / 2)
    end
    return areas
end

for i = 1 : 200 #set some arbitrarily high number for for loop, which should break before then
    areas = area(t, pwnodes)
    imax = findmax(areas) #find triangle with largest area
    if imax[1] > 0.5 * hmax ^ 2 #if so, add circumcenter of triangle with largest area to list of node points
        vertices = pwnodes[t[imax[2], :], :] #recall vertex of triangle with largest area
        #find midpoints and slopes (only two of each are needed, but the slope cannot be zero!)
        midpoint1 = (vertices[1, :] + vertices[2, :]) / 2 
        slope1 = (vertices[2, 2] - vertices[1, 2]) / (vertices[2, 1] - vertices[1, 1])
        midpoint2 = (vertices[2, :] + vertices[3, :]) / 2 
        slope2 = (vertices[3, 2] - vertices[2, 2]) / (vertices[3, 1] - vertices[2, 1])
        midpoint3 = (vertices[3, :] + vertices[1, :]) / 2 
        slope3 = (vertices[1, 2] - vertices[3, 2]) / (vertices[1, 1] - vertices[3, 1])
        #solve linear system of equations for the circumcenter coordinate, with special treatment for cases when slopes are zero
        if slope1 == 0
            A = [1 / slope3 1; 1 / slope2 1]
            b = [midpoint3[2] + (midpoint3[1] / slope3) ; midpoint2[2] + (midpoint2[1] / slope2)]
        elseif slope2 == 0
            A = [1 / slope1 1; 1 / slope3 1]
            b = [midpoint1[2] + (midpoint1[1] / slope1) ; midpoint3[2] + (midpoint3[1] / slope3)]   
        else
            A = [1 / slope1 1; 1 / slope2 1]
            b = [midpoint1[2] + (midpoint1[1] / slope1) ; midpoint2[2] + (midpoint2[1] / slope2)]
        end
        circumcenter = reshape(A \ b, 1, 2) #solve for circumcenter
        pwnodes = vcat(pwnodes, circumcenter) #add circumcenter
        t = filteroutside(delaunay(pwnodes), pwnodes, pv) #refilter outside
    else #if no triangle area is larger than \frac{h^{2}_{max}}{2}, we are finished
        break
    end
end
figure(1, figsize = (4, 4))
tplot(pwnodes, t)
```

**Part (h)**

Finally, after the steps above, we write code which will refine the mesh $n_{ref}$ times. For each time, we accomplish this by adding a node at the center of each mesh edge, then retriangulate and remove outside triangles each time. To help with this, we also load `all_edges`, which will return the indices of every neighboring pair of nodes with a unique edge between them. After doing this, we can just loop through the vector containing these indices, `edges`, and simply add the central node between them to our evergrowing list of nodes, `pwnodes`. (Naturally, for two neighboring nodes $(x_{i},y_{i})$ and $(x_{j},y_{j})$, the center of the edge between them is located at $(\frac{x_{i}+x_{j}}{2},\frac{y_{i}+y_{j}}{2})$.) As earlier, we can also use `unique` again to remove any duplicates.

```julia
"""
    edges, boundary_indices = all_edges(t)

Find all unique edges in the triangulation `t` (ne x 2 array)
Second output is indices to the boundary edges.
"""
function all_edges(t)
    edges = vcat(t[:,[1,2]], t[:,[2,3]], t[:,[3,1]])
    edges = sortslices(sort(edges, dims=2), dims=1)
    dup = all(edges[2:end,:] - edges[1:end-1,:] .== 0, dims=2)[:]
    keep = .![false;dup]
    edges = edges[keep,:]
    dup = [dup;false]
    dup = dup[keep]
    return edges, findall(.!dup)
end
```

```julia
using LinearAlgebra
using PyPlot

function refine(pwnodes, nref)
    for i = 1 : nref #repeat nref time(s)
        edges, boundary_indices = all_edges(delaunay(pwnodes)) #locate all edges
        for i = 1 : length(edges[:, 1])
            center = (pwnodes[edges[i, 1], :] + pwnodes[edges[i, 2], :]) / 2 #find center of each edge
            pwnodes = vcat(pwnodes, center') #add center to node list
        end
        pwnodes = unique(pwnodes, dims = 1)
    end
    return pwnodes
end

pv = [0 0; 1 0; .5 .5; 1 1; 0 1; 0 0]
hmax = 0.2
pwnodes = nodecreator(pv, hmax)
t = filteroutside(delaunay(pwnodes), pwnodes, pv)

for i = 1 : 200 #set some arbitrarily high number for for loop, which should break before then
    areas = area(t, pwnodes)
    imax = findmax(areas) #find triangle with largest area
    if imax[1] > 0.5 * hmax ^ 2 #if so, add circumcenter of triangle with largest area to list of node points
        vertices = pwnodes[t[imax[2], :], :] #recall vertex of triangle with largest area
        #find midpoints and slopes (only two of each are needed, but the slope cannot be zero!)
        midpoint1 = (vertices[1, :] + vertices[2, :]) / 2 
        slope1 = (vertices[2, 2] - vertices[1, 2]) / (vertices[2, 1] - vertices[1, 1])
        midpoint2 = (vertices[2, :] + vertices[3, :]) / 2 
        slope2 = (vertices[3, 2] - vertices[2, 2]) / (vertices[3, 1] - vertices[2, 1])
        midpoint3 = (vertices[3, :] + vertices[1, :]) / 2 
        slope3 = (vertices[1, 2] - vertices[3, 2]) / (vertices[1, 1] - vertices[3, 1])
        #solve linear system of equations for the circumcenter coordinate, with special treatment for cases when slopes are zero
        if slope1 == 0
            A = [1 / slope3 1; 1 / slope2 1]
            b = [midpoint3[2] + (midpoint3[1] / slope3) ; midpoint2[2] + (midpoint2[1] / slope2)]
        elseif slope2 == 0
            A = [1 / slope1 1; 1 / slope3 1]
            b = [midpoint1[2] + (midpoint1[1] / slope1) ; midpoint3[2] + (midpoint3[1] / slope3)]   
        else
            A = [1 / slope1 1; 1 / slope2 1]
            b = [midpoint1[2] + (midpoint1[1] / slope1) ; midpoint2[2] + (midpoint2[1] / slope2)]
        end
        circumcenter = reshape(A \ b, 1, 2) #solve for circumcenter
        pwnodes = vcat(pwnodes, circumcenter) #add circumcenter
        t = filteroutside(delaunay(pwnodes), pwnodes, pv) #retriangulate, refilter outside
    else #if no triangle area is larger than \frac{h^{2}_{max}}{2}, we are finished
        break
    end
end

nref = 1
pwnodes = refine(pwnodes, nref) #refine mesh
t = filteroutside(delaunay(pwnodes), pwnodes, pv) #retriangulate, refilter outside
figure(1, figsize = (4, 4))
tplot(pwnodes, t)
```

**Putting it all together**

Finally, we can use load in `boundary_nodes` to return the nodes `e` on the boundary of our mesh. Along with our defined functions in the code blocks above, we use this to construct our final function, `pmesh`, of the form `p, t, e, = pmesh(pv, hmax, nref)`.

We start by testing this out on the example in the figures of the problem set.

```julia
"""
    e = boundary_nodes(t)

Find all boundary nodes in the triangulation `t`.
"""
function boundary_nodes(t)
    edges, boundary_indices = all_edges(t)
    return unique(edges[boundary_indices,:][:])
end
```

```julia
using LinearAlgebra
using PyPlot

function pmesh(pv, hmax, nref)
    pwnodes = nodecreator(pv, hmax) #create subnodes along polygon segments
    t = filteroutside(delaunay(pwnodes), pwnodes, pv) #retriangulate, refilter outside
    
    for i = 1 : 200 #set some arbitrarily high number for for loop, which should break before then
        areas = area(t, pwnodes)
        imax = findmax(areas) #find triangle with largest area
        if imax[1] > 0.5 * hmax ^ 2 #if so, add circumcenter of triangle with largest area to list of node points
            vertices = pwnodes[t[imax[2], :], :] #recall vertex of triangle with largest area
            #find midpoints and slopes (only two of each are needed, but the slope cannot be zero!)
            midpoint1 = (vertices[1, :] + vertices[2, :]) / 2 
            slope1 = (vertices[2, 2] - vertices[1, 2]) / (vertices[2, 1] - vertices[1, 1])
            midpoint2 = (vertices[2, :] + vertices[3, :]) / 2 
            slope2 = (vertices[3, 2] - vertices[2, 2]) / (vertices[3, 1] - vertices[2, 1])
            midpoint3 = (vertices[3, :] + vertices[1, :]) / 2 
            slope3 = (vertices[1, 2] - vertices[3, 2]) / (vertices[1, 1] - vertices[3, 1])
            #solve linear system of equations for the circumcenter coordinate, with special treatment for cases when slopes are zero
            if slope1 == 0
                A = [1 / slope3 1; 1 / slope2 1]
                b = [midpoint3[2] + (midpoint3[1] / slope3) ; midpoint2[2] + (midpoint2[1] / slope2)]
            elseif slope2 == 0
                A = [1 / slope1 1; 1 / slope3 1]
                b = [midpoint1[2] + (midpoint1[1] / slope1) ; midpoint3[2] + (midpoint3[1] / slope3)]   
            else
                A = [1 / slope1 1; 1 / slope2 1]
                b = [midpoint1[2] + (midpoint1[1] / slope1) ; midpoint2[2] + (midpoint2[1] / slope2)]
            end
            circumcenter = reshape(A \ b, 1, 2) #solve for circumcenter
            pwnodes = vcat(pwnodes, circumcenter) #add circumcenter
            t = filteroutside(delaunay(pwnodes), pwnodes, pv) #retriangulate, refilter outside
        else #if no triangle area is larger than \frac{h^{2}_{max}}{2}, we are finished
            break
        end
    end
    p = refine(pwnodes, nref) #refine mesh nref times
    t = filteroutside(delaunay(p), p, pv) #retriangulate, refilter outside
    e = boundary_nodes(t)
    return p, t, e
end

pv = [0 0; 1 0; .5 .5; 1 1; 0 1; 0 0]
hmax = 0.2
nref = 1
p, t, e = pmesh(pv, hmax, nref)
tplot(p, t)
```

This works *exactly* as expected, so as desired. Now, we test it out on another arbitrary polygon. (No joke, I just typed out some ad hoc, random points, and them ordered them in a way which would form a valid polygon.) We use $h_{max}=n_{ref}=1$.

## Problem 2
**Part (a)**

We desire a numerical solution to the following PDE, called the Eikonal equation:

$$
F(x,y)\left|\nabla\phi\left(x,y\right)\right|=1,\hspace{0.5cm}\phi\left(x_{D},y_{D}\right)=0. \quad (1)
$$

We could apply the much more efficient fast marching method, but this might be very complicated to do in Julia without knowledge of heaping (which I unfortunately could not learn with my other classes within two weeks). Therefore, we will instead add an extra variable, $s$ and try to solve the equation

$$
\phi_{s}+F\left|\nabla\phi\right|=1,\hspace{0.5cm}\phi\left(x_{D},y_{D}\right)=0. \quad (2)
$$

arbitrarily forward in time. $(2)$ acts as a method for us to "timestep" in $s$ towards the exact solution for $(1)$, since by observation, the steady-state solution of $(2)$ as $s\rightarrow\infty$ is the solution to $(1)$. We just need to impose $\phi\left(x_{D},y_{D}\right)=0$ at each "timestep."

We start by writing a function `nablas` which returns the norm of the gradient field of a grid scalar field function `phi`=$\phi(x,y)$, i.e. $\left|\nabla\phi\right|$, given a permutation matrix

$$
\begin{bmatrix} 0 & 1 & 0 & \cdots  & \cdots \\ 0 & 0 & 1  & 0 & \ddots  \\ \vdots  & \ddots  & \ddots  & \ddots & \vdots \\ \ddots & \cdots & \cdots  & 0 & 1  \\ 1  & 0  & \cdots & \cdots & 0 \end{bmatrix}
$$

to aid us in calculating the finite difference, a spatial stepsize $h$, and a number of gridpoints $n$ (such that $nh=1$). This will be used to solve $(2)$, and in turn $(1)$.

(In particular, `nablas` actually returns the gradient norm using "backwards" and "forwards" gradient norms, or 

$$
\nabla_{ij}^{\pm}=\sqrt{\max\left(D^{\mp x}\phi_{ij}^{n},0\right)^{2}+\min\left(D^{\pm x}\phi_{ij}^{n},0\right)^{2}+\max\left(D^{\mp y}\phi_{ij}^{n},0\right)^{2}+\min\left(D^{\pm y}\phi_{ij}^{n},0\right)^{2}}
$$

where $D^{-x}$/$D^{+x}$ is the first-order backward/forward difference operator in the $x$-direction and $D^{-y}$/$D^{+y}$ is the first-order backward/forward difference operator in the $y$-direction . This will be used to numerically calculate $\left|\nabla\phi\left(x_{i},y_{j}\right)\right|\approx \max\left(F,0\right)\nabla_{ij}^{+}+\min\left(F,0\right)\nabla_{ij}^{-}$.)

We treat calculating the gradient on the boundaries of the unit square separately by applying a second-order forward (3 -4 1) or backward stencil (-3 4 -1), depending on which edge we are at.

```julia
function nablas(phi, permutation, h, n)  
    invh = Int64(1 / h)
    phixminus = permutation' * phi
    phixplus = permutation * phi
    phiyminus = phi * permutation
    phiyplus = phi * permutation'
    Dminusx = -invh * (phixminus - phi) #calculate finite differences
    Dplusx = invh * (phixplus - phi)
    Dminusy = -invh * (phiyminus - phi)
    Dplusy = invh * (phiyplus - phi)
    #impose special treatment of boundaries (replace backward/forward schemes with forward/backward schemes)
    Dminusx[1, :] = 0.5 * invh * (-3 * phi[1, :] + 4 * phi[2, :] - phi[3, :])
    Dplusx[n + 1, :] = 0.5 * invh * (3 * phi[n + 1, :] - 4 * phi[n, :] + phi[n - 1, :])
    Dminusy[:, 1] = 0.5 * invh * (-3 * phi[:, 1] + 4 * phi[:, 2] - phi[:, 3]) 
    Dplusy[:, n + 1] = 0.5 * invh * (3 * phi[:, n + 1] - 4 * phi[:, n] + phi[:, n - 1])
    nablaplus = sqrt.(max.(Dminusx, zeros(n + 1, n + 1)) .^ 2 + min.(Dplusx, zeros(n + 1, n + 1)) .^ 2 + max.(Dminusy, zeros(n + 1, n + 1)) .^ 2 + min.(Dplusy, zeros(n + 1, n + 1)) .^ 2)
    nablaminus = sqrt.(max.(Dplusx, zeros(n + 1, n + 1)) .^ 2 + min.(Dminusx, zeros(n + 1, n + 1)) .^ 2 + max.(Dplusy, zeros(n + 1, n + 1)) .^ 2 + min.(Dminusy, zeros(n + 1, n + 1)) .^ 2)
    return nablaplus, nablaminus
end
```

Using `nablas`, we can construct a function, `eikonal`, that will provide us with a numerical solution to the Eikonal equation, $(2)$, provided a spatial stepsize $h$, a matrix grid function `fij`, a "timestep" `dt`$\equiv s_{i+1}-s{i}$, a tolerance `tol` (to check if we have reached a steady-state for $\phi$), a maximum number of iterations `maxit` (to prevent blowups if our scheme does not converge properly), and an initial point `(xd, yd)`$\equiv\left(x_{D},y_{D}\right)$ at which we will impose that $\phi\left(x_{D},y_{D}\right)=0$ for each "timestep." We want to use this to solve $(2)$ using an explicit linear multistep method. Professor Persson suggested using forward Euler, but we will use RK4 (i.e. "the" Runge-Kutta method) instead for accuracy.

Without stating how the RK4 scheme works again, we note that we are simply looking for a numerical solution at arbitrarily large $s$ to the matrix differential equation $\frac{\partial\phi}{\partial s}=\eta\left(x,y,\nabla\phi(x,y)\right)$, where $\eta\left(x,y,\nabla\phi(x,y)\right)=1-F(x,y)\left|\nabla\phi(x,y)\right|$. Of course, we can see why defining `nablas` as a function will be helpful for our solver, because we need to evaluate $\eta$ four times for each timestep.

```julia
import LinearAlgebra

function eikonal(h, fij, dt, tol, maxit, xd, yd) #gives a solution to the Eikonal equation
    n = Int64(1.0 / h)
    invh = 1 / h
    x = h * (0 : n)
    y = x
    phi = 0.5 * ones(n + 1, n + 1)  #use halves as an initial guess for phi
   
    #set up permutation matrix
    permutation = zeros(n + 1, n + 1)
    for i = 1 : n 
        permutation[i, i + 1] = 1
    end
    permutation[n + 1, 1] = 1
    
    #index at (x_{D}, y_{D})
    xdi = findall(x -> x == xd, x)[1, 1]
    ydi = findall(y -> y == yd, y)[1, 1]
    phi[xdi, ydi] = 0 #set phi at (x_{D}, y_{D}) to zero 
    
    iter = 0
    for i = 1 : maxit #loop via RK4
        iter += 1 #count iterations (for posterity's sake)
        nablaplus1, nablaminus1 = nablas(phi, permutation, h, n)  
        k1 = dt * (1 .- (max.(fij, zeros(n + 1, n + 1)) .* nablaplus1 + min.(fij, zeros(n + 1, n + 1)) .* nablaminus1))
        nablaplus2, nablaminus2 = nablas(phi + 0.5 * k1, permutation, h, n)  
        k2 = dt * (1 .- (max.(fij, zeros(n + 1, n + 1)) .* nablaplus2 + min.(fij, zeros(n + 1, n + 1)) .* nablaminus2)) 
        nablaplus3, nablaminus3 = nablas(phi + 0.5 * k2, permutation, h, n) 
        k3 = dt * (1 .- (max.(fij, zeros(n + 1, n + 1)) .* nablaplus3 + min.(fij, zeros(n + 1, n + 1)) .* nablaminus3))
        nablaplus4, nablaminus4 = nablas(phi + k3, permutation, h, n) 
        k4 = dt * (1 .- (max.(fij, zeros(n + 1, n + 1)) .* nablaplus4 + min.(fij, zeros(n + 1, n + 1)) .* nablaminus4))
        phinew = phi + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        phinew[xdi, ydi] = 0 #impose that \phi(x_{D}, y_{D}) is zero for each step
        if norm(phi .- phinew, Inf) < tol #if we have reached an adequate steady-state, stop and use it
            phi = phinew
            break
            else #otherwise, keep iterating
            phi = phinew
        end
    end
    return x, y, phi, iter
end
```

Finally, we test our `eikonal` on $F$ as given in case 1 of problem 2(c) (which we will revisit later for calculating our optimal path, of course), or $F=1$ with $\phi\left(x_{D},y_{D}\right)=0$, and see if the resultant level curves for $\phi$ make sense.

```julia
import PyPlot

tol = 1e-8
maxit = 6000
h = 1 / 100
n = Int64(1.0 / h)
x = h * (0 : n)
y = x

#fill up function grid
fij = ones(n + 1, n + 1)

dt = 1 / 1000
xd = 0.2
yd = 0.2
x, y, phi, iter = eikonal(h, fij, dt, tol, maxit, xd, yd)
println("This solver took ", iter, " iterations to converge.")

clf()
contourf(x, y, phi, 25)
axis("equal")
title(L"\phi(x, y)")
colorbar()
```

As expected, our computed level curves appear to be circles emanating from $\phi\left(x_{D},y_{D}\right)=0$. $F=1$ should give a distance function (i.e. the level curves travel at constant speed $1$ from $\left(x_{D},y_{D}\right)$), so this numerical solution looks correct. 

**Part (b)**

Next, we know that the optimal path from our departure point $\left(x_{D},y_{D}\right)$ to our arrival point $\left(x_{A},y_{A}\right)$ obeys the ODE $\frac{d\overrightarrow{r}}{dt}=\overrightarrow{n}$, or equivalently,

$$
\frac{d\overrightarrow{r}}{dt}=\frac{1}{\sqrt{\left(\frac{\partial\phi}{\partial x}\right)^{2}+\left(\frac{\partial\phi}{\partial y}\right)^{2}}}\left<\frac{\partial\phi}{\partial x},\frac{\partial\phi}{\partial y}\right> \\
\Rightarrow\frac{dx}{dt}=\frac{\frac{\partial\phi}{\partial x}}{\sqrt{\left(\frac{\partial\phi}{\partial x}\right)^{2}+\left(\frac{\partial\phi}{\partial y}\right)^{2}}}:=n_{x},\quad\frac{dy}{dt}=\frac{\frac{\partial\phi}{\partial y}}{\sqrt{\left(\frac{\partial\phi}{\partial x}\right)^{2}+\left(\frac{\partial\phi}{\partial y}\right)^{2}}}:=n_{y}. \quad(3)
$$

We define a function `normal` which, when given a matrix scalar field `phi` and stepsize `h` (assuming a square grid) will return the $x$- and $y$-components of the normal vector field $\overrightarrow{n}(x,y)$ of `phi` evaluated on this grid, or $n_{x}$ and $n_{y}$, respectively.

```julia
function normal(phi, h) #given a scalar field phi and stepsize h (assume square grid), returns the field of normal vectors
    invh = Int64(1 / h)
    N = length(phi[:, 1])
    
    #set up permutation matrix
    permutation = zeros(N, N)
    for i = 1 : N - 1
        permutation[i, i + 1] = 1
    end
    permutation[N, 1] = 1
    
    phixminus = permutation' * phi
    phixplus = permutation * phi
    phiyminus = phi * permutation
    phiyplus = phi * permutation'
    dphidx = 0.5 * invh * (phixplus - phixminus) #calculate derivatives using 2nd-order central differences
    dphidy = 0.5 * invh * (phiyplus - phiyminus)
    #impose special treatment of boundaries (replace central schemes with forward/backward schemes)
    dphidx[1, :] = 0.5 * invh * (-3 * phi[1, :] + 4 * phi[2, :] - phi[3, :])
    dphidx[end, :] = 0.5 * invh * (3 * phi[end, :] - 4 * phi[end - 1, :] + phi[end - 2, :])
    dphidy[:, 1] = 0.5 * invh * (-3 * phi[:, 1] + 4 * phi[:, 2] - phi[:, 3])
    dphidy[:, end] = 0.5 * invh * (3 * phi[:, end] - 4 * phi[:, end - 1] + phi[:, end - 2])
    invnormgrad = @. 1 / sqrt(dphidx ^ 2 + dphidy ^ 2)
    nx = invnormgrad .* dphidx 
    ny = invnormgrad .* dphidy
    return nx, ny
end
```

However, we need to numerically integrate $(3)$, which requires us to evaluate $\overrightarrow{n}(x,y)$ at an arbitrary point $(x,y)$ within the unit square. Since `normal` only returns $\overrightarrow{n}$ on a grid, we need to use interpolation to evaluate $\overrightarrow{n}(x,y)$ for $(x,y)$ off the grid. 

There are many interpolation methods that we could use here, but we choose bilinear interpolation. Suppose that we are given an arbitrary grid function $f_{i,j}=f\left(x_{i},y_{j}\right)$ which we want to evaluate at $(x,y)$. Then, given grid function values at $f_{i,j}=f\left(x_{i},y_{j}\right)$, $f_{i,l}=f\left(x_{i},y_{l}\right)$, $f_{k,j}=f\left(x_{k},y_{j}\right)$, and $f_{k,l}=f\left(x_{k},y_{l}\right)$, our approximate value for $f(x,y)$ using bilinear interpolation is

$$
f(x,y)\approx\frac{1}{\left(x_{k}-x_{i}\right)\left(y_{l}-y_{j}\right)}\begin{bmatrix} x_{k}-x & x-x_{i} \end{bmatrix}\begin{bmatrix} f_{i,j} & f_{i,l} \\ f_{k,j} & f_{k,l} \end{bmatrix}\begin{bmatrix} y_{l}-y \\ y-y_{j} \end{bmatrix} \quad(4)
$$

In particular, we want to minimize the local truncation error (LTE) introduced via bilinear interpolation, which scales quadratically with the distance between $(x,y)$ & $\left(x_{i},y_{j}\right)$, $(x,y)$ & $\left(x_{i},y_{l}\right)$, $(x,y)$ & $\left(x_{k},y_{j}\right)$, and $(x,y)$ & $\left(x_{k},y_{l}\right)$. Therefore, we enforce that $k=i+1$, $l=j+1$, as well as that $(x,y)\in D_{i,j}$, where $D_{i,j}$ is the square with side lengths $h=x_{i+1}-x_{i}=y_{j+1}-y_{j}\hspace{0.1cm}\forall i,j$ with vertices $\left(x_{i},y_{j}\right)$, $\left(x_{i},y_{j+1}\right)$, $\left(x_{i+1},y_{j}\right)$, and $\left(x_{i+1},y_{j+1}\right)$. Then, $(4)$ becomes

$$
f(x,y)\approx\frac{1}{\left(x_{i+1}-x_{i}\right)\left(y_{j+1}-y_{j}\right)}\begin{bmatrix} x_{i+1}-x & x-x_{i} \end{bmatrix}\begin{bmatrix} f_{i,j} & f_{i,j+1} \\ f_{i+1,j} & f_{i+1,j+1} \end{bmatrix}\begin{bmatrix} y_{j+1}-y \\ y-y_{j} \end{bmatrix} \quad(4)
$$

This choice of interpolating points will minimize the distance between these points and the query point $(x,y)$, and thereby the local error introduced by using $(5)$ to approximate $f(x,y)$. `bilinear` does this below, given a grid function `f`, $x$-values `x`, $y$-values `y`, and query point `rq`$=(x,y)$.

```julia
function bilinear(f, x, y, rq) #uses bilinear interpolation to approximate the value of f(x, y) at query point rq=(xq, yq)
    xq = rq[1]
    yq = rq[2]
    distxq = @. abs(xq - x)
    distyq = @. abs(yq - y)
    mindistxqval, mindistxqind = findmin(distxq) #find index of first closest x value
    #we want to make sure that xq lies between x_{i} and x_{i+1}
    if xq > x[mindistxqind] 
        xless = x[mindistxqind] 
        xmore = x[mindistxqind + 1] 
        i = mindistxqind
    elseif xq <= x[mindistxqind]
        xless = x[mindistxqind - 1] 
        xmore = x[mindistxqind]
        i = mindistxqind - 1
    end
    mindistyqval, mindistyqind = findmin(distyq) #find index of first closest y value
    #we want to make sure that yq lies between y_{i} and y_{i+1}
    if yq > y[mindistyqind]
        yless = y[mindistyqind] 
        ymore = y[mindistyqind + 1] 
        j = mindistyqind
    elseif yq <= y[mindistyqind]
        yless = y[mindistyqind - 1] 
        ymore = y[mindistyqind] 
        j = mindistyqind - 1
    end
    first = 1 / ((xmore - xless) * (ymore - yless))
    second = [xmore - xq xq - xless]
    third = [f[i, j] f[i, j + 1]; f[i + 1, j] f[i + 1, j + 1]]
    fourth = [ymore - yq; yq - yless]
    fxqyq = first * second * third * fourth
end
```

Finally, using functions `normal` and `bilinear`, we construct a function `optimalpath` which will return the two-column array `r` that contains the coordinates of the optimal path, as well as the time `t` that it will take to complete this path (by evaluating `phi` at $\left(x_{A},y_{A}\right)$, since $\phi\left(x_{A},y_{A}\right)=t$). `optimalpath` takes in level set function `phi` (i.e. the solution to the Eikonal equation), spatial stepsize `h`, timestep `k`, $x$ grid values `x`, $y$ grid values `y`, arrival point `(xi, yi)`$\equiv\left(x_{A},y_{A}\right)$, a tolerance value `tol` close to zero to halt our timestepping once the level set value at a timestep is below it (since $\phi\left(x_{D},y_{D}\right)=0$), and a maximum iterations count `maxit` in case our numerical solution fails and misses $\left(x_{D},y_{D}\right)$ for some reason.

More specifically, `optimalpath` actually integrates the system of ODEs given by $(3)$ *backwards* in time via the RK4 scheme. Why backwards? Well, regardless of our arrival point, all optimal paths should lead back to the departure point, $\left(x_{D},y_{D}\right)$, where the zero level set is located. Therefore, we would expect a singularity to be at $\left(x_{D},y_{D}\right)$, since the uniqueness of our solution breaks down there. Sure enough, since $\overrightarrow{n}=\frac{1}{\sqrt{\left(\frac{\partial\phi}{\partial x}\right)^{2}+\left(\frac{\partial\phi}{\partial y}\right)^{2}}}\left<\frac{\partial\phi}{\partial x},\frac{\partial\phi}{\partial y}\right>$, we see that $\overrightarrow{n}$ approaches $\frac{0}{0}$ near $\left(x_{D},y_{D}\right)$, so our ODE is not well-defined there. Anyways, integrating backwards is no problem for RK4, so we will be fine.

```julia
function optimalpath(phi, h, k, x, y, xi, yi, tol, maxit) #returns the optimal path via RK4 integration
    #h is spatial stepsize
    #k is timestep
    nx, ny = normal(phi, h)
    r = [xi yi]
    phif = bilinear(phi, x, y, r) 
    t = phif[1]
    i = 0 #set counter to 0
    while norm(phif) > tol #if we have reached the desired level set, stop. It should approach zero.
        i += 1 #count index
        if i > maxit #prevent blowup 
            break
        end
        k1 = k * [bilinear(nx, x, y, r[i, :]) bilinear(ny, x, y, r[i, :])]
        k2 = k * [bilinear(nx, x, y, r[i, :]' + 0.5 * k1) bilinear(ny, x, y, r[i, :]' + 0.5 * k1)]
        k3 = k * [bilinear(nx, x, y, r[i, :]' + 0.5 * k2) bilinear(ny, x, y, r[i, :]' + 0.5 * k2)]
        k4 = k * [bilinear(nx, x, y, r[i, :]' + k3) bilinear(ny, x, y, r[i, :]' + k3)]
        rf = r[i, :]' + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        phif = bilinear(phi, x, y, rf) 
        r = vcat(r, rf)
    end
    return r, t
end
```

We now use `eikonal` and `optimalpath` to find solutions to the Eikonal equation and the corresponding optimal paths, respectively, for the following initial value problems.

***Case 1***
$$
\left(x_{D},y_{D}\right)=\left(0.2,0.2\right),\quad\left(x_{A},y_{A}\right)=\left(0.8,0.8\right),\quad F(x,y)=1
$$

```julia
phitol = 1e-8
maxit = 3000
h = 1 / 100
n = Int64(1.0 / h)

#fill up function grid
fij = ones(n + 1, n + 1)

dt = 1 / 1000
xd = 0.2
yd = 0.2
x, y, phi, iter = eikonal(h, fij, dt, phitol, maxit, xd, yd)

tol = 2e-2
k = -1 / 1000 #negative timestep, because we are integrating backwards in time
xa = 0.8
ya = 0.8
r, t = optimalpath(phi, h, k, x, y, xa, ya, tol, 1000)

figure()
title("Eikonal equation solution and the optimal path from (0.2, 0.2) to (0.8, 0.8)")
contourf(x, y, phi, 25)
colorbar()
scatter(r[:, 1], r[:, 2], s = 1, c = "w")
show()

println("This path will take ", t, " seconds.")
```

As expected, the optimal path is just a straight line from $\left(x_{D},y_{D}\right)=(0.2,0.2)$ to $\left(x_{A},y_{A}\right)=(0.8,0.8)$, since the level sets are circles radiating out from $(0.2,0.2)$.

***Case 2***

$\left(x_{D},y_{D}\right)=\left(0.2,0.2\right),\left(x_{A},y_{A}\right)=\left(0.8,0.45\right),$ and

$$

This optimal path was a little bit surprising and unintuitive, especially with how it appeared to leave normal to the circles surrounding $(0.2,0.2)$, almost immediately cut to the right, then back down again diagonal-right. It probably has a lot to do with the discontinuity in $F$.
