## Problem 1

Show that the region of absolute stability of the 2-stage Milne's method,

$$u_{n+1}=u_{n-1}+h\left(\frac{1}{3}f_{n+1}+\frac{4}{3}f_n+\frac{1}{3}f_{n-1}\right),$$

is only along the imaginary axis.

___
**Proof:** 
For sake of consistency with the reference text, *Finite Difference Methods for Ordinary and Partial Differential Equations* by Randall J. LeVeque, we should shift indices and redefine the 2-stage Milne's method as

$$
u_{n+2}=u_{n}+h\left(\frac{1}{3}f_{n+2}+\frac{4}{3}f_{n+1}+\frac{1}{3}f_{n}\right) \hspace{0.5cm} (1)
$$

To find the region of absolute stability (RAS) for the 2-stage Milne method, we apply $(1)$ to the test problem $u^{'}(t)=\lambda u(t) \equiv f(u, t)$, which means that $f_{i} \equiv \lambda u_{i}$. With this, we get the following recurrence relation:

$$
u_{n+2}=u_{n}+h\left(\frac{\lambda}{3}u_{n+2}+\frac{4\lambda}{3}u_{n+1}+\frac{\lambda}{3}u_{n}\right)=u_{n}+\frac{h\lambda}{3}\left(u_{n+2}+4u_{n+1}+u_{n}\right) \hspace{0.5cm} (2)
$$

Defining $z:=h\lambda$, $(2)$ becomes

$$
u_{n+2}=u_{n}+\frac{z}{3}\left(u_{n+2}+4u_{n+1}+u_{n}\right) \Rightarrow u_{n+2}(1-\frac{z}{3})=\frac{4z}{3}u_{n+1}+u_{n}(1+\frac{z}{3}) \\ \Rightarrow (1-\frac{z}{3})u_{n+2}-\frac{4z}{3}u_{n+1}-(1+\frac{z}{3})u_{n}=0 \hspace{0.5cm} (3)
$$

Now, we want to study the characteristic polynomial $\pi(\zeta ; z)$ corresponding to $(3)$, which has two roots: $\zeta_{1}$ and $\zeta_{2}$. Here it is:

$$
\pi(\zeta_{12} ; z):=(1-\frac{z}{3})\zeta_{12}^{2}-\frac{4z}{3}\zeta_{12}-1-\frac{z}{3}=0 \hspace{0.5cm} (4)
$$

Our goal is to find the set of all points $z$, *in the complex plane*, such that $\pi(\zeta ; z)$ satisfies the root condition, which is the RAS. Namely, we want that $\left|\zeta_{i}\right|\leq 1$, $\forall i=1,2$, and if $\zeta_{1}=\zeta_{2}$, then $\left|\zeta_{1}\right|=\left|\zeta_{2}\right|<1$. Thus, since we are considering $z\in \mathbb{C}$, we can solve for the boundary of this set by substituting $e^{i\theta}$ for $\zeta_{12}$ in $(4)$, giving us an equation for the boundary. This gives us

$$
(1-\frac{z}{3})e^{2i\theta}-\frac{4z}{3}e^{i\theta}-1-\frac{z}{3}=0 \hspace{0.5cm} (5)
$$

Our last step is to use $(5)$ to solve for $z$ along this boundary.

$$
\Rightarrow e^{2i\theta}-1=\frac{z}{3}e^{2i\theta}+\frac{4z}{3}e^{i\theta}+\frac{z}{3}=z\left(\frac{1}{3}e^{2i\theta}+\frac{4}{3}e^{i\theta}+\frac{1}{3}\right) \\ \Rightarrow z=\frac{3\left(e^{2i\theta}-1\right)}{e^{2i\theta}+4e^{i\theta}+1}=\frac{3\left(e^{2i\theta}-1\right)}{e^{2i\theta}+4e^{i\theta}+1}\cdot\frac{e^{-i\theta}}{e^{-i\theta}}=\frac{3\left(e^{i\theta}-e^{-i\theta}\right)}{e^{i\theta}+e^{-i\theta}+4}=\frac{3\cdot2i\left(\frac{e^{i\theta}-e^{-i\theta}}{2i}\right)}{2\cdot\left(\frac{e^{i\theta}+e^{-i\theta}}{2}+2\right)} \hspace{0.5cm} (6)
$$

Finally, by definition, $\cos\theta\equiv\frac{e^{i\theta}+e^{-i\theta}}{2}$ and $\sin\theta\equiv\frac{e^{i\theta}-e^{-i\theta}}{2i}$. Therefore, $(6)$ simplifies to

$$
z=\frac{3i\sin\theta}{\cos\theta+2} \hspace{0.5cm} (7)
$$

Therefore, since, regardless of $\theta$, the denominator of $(7)$ is real and its numerator is always imaginary, the boundary for our RAS lies entirely on the imaginary axis. Furthermore, even the boundary along the imaginary axis is bounded (i.e. the boundary does not extend to the entire imaginary axis). To solve for these bounds, we find the extrema for $(7)$ on the interval $[0,2\pi]$, since these can be generalized $\forall\theta\in\mathbb{R}$.

$$
\frac{dz}{d\theta}=\frac{3i(1+2\cos\theta_{c})}{\left(2+\cos\theta_{c}\right)^{2}}=0, \hspace{0.25cm} \theta_{c}\in[0,2\pi] \hspace{0.5cm} \\ \Rightarrow \theta_{c}=\frac{2\pi}{3}, \frac{4\pi}{3}
$$

A quick calculation reveals that $\theta_{c}=\frac{2\pi}{3}$ attains the maximum, $i\sqrt{3}$, and $\theta_{c}=\frac{4\pi}{3}$ attains the minimum, $-i\sqrt{3}$, so the boundary along the imaginary axis is bounded by $\left[-i\sqrt{3},i\sqrt{3}\right]$. Outside of this, along the imaginary axis, we can say that our scheme is no longer absolutely stable. Furthermore, after another quick calculation, we find that $z=\pm i\sqrt{3}$ both give us repeated roots for $\zeta$ (and in such a case, $\pi(\zeta ; \pm i\sqrt{3})$ would violate the root condition), so along the imaginary axis, our scheme is only absolutely stable $\forall \operatorname{Im}(z)\in\left(-i\sqrt{3},i\sqrt{3}\right)$.

To test both sides of the imaginary axis, we try out $z=1$ and $z=-1$ in our characteristic equation for the upper and lower half of the complex plane, respectively, and solve for the resultant roots to see if they allow $\pi$ to satisfy the root condition.

**z=1:**

$$
\pi(\zeta_{12} ; 1)=\frac{2}{3}\zeta_{12}^{2}-\frac{4}{3}\zeta_{12}-\frac{4}{3}=0 \Rightarrow \zeta_{12}^{2}-2\zeta_{12}-2=0 \\ \Rightarrow \zeta_{12}=1-\sqrt3,1+\sqrt3, \hspace{0.25cm} 1+\sqrt3 > 1 \hspace{0.25cm} \times 
$$

**z=-1:**

$$
\pi(\zeta_{12} ; -1)=\frac{4}{3}\zeta_{12}^{2}+\frac{4}{3}\zeta_{12}-\frac{2}{3}=0 \Rightarrow \zeta_{12}^{2}+\zeta_{12}-\frac{1}{2}=0 \\ \Rightarrow \zeta_{12}=-\frac{\sqrt3+1}{2},\frac{\sqrt3-1}{2}, \hspace{0.25cm} \left|-\frac{\sqrt3+1}{2}\right|=\frac{\sqrt3+1}{2} > 1 \hspace{0.25cm} \times 
$$

$\therefore$ Since both values of $z$ on either side of the RAS boundary cause $\pi$ to fail the root condition, we conclude that the RAS for the 2-stage Milne method lies only along the imaginary axis, $\forall \operatorname{Im}(z)\in\left(-i\sqrt{3},i\sqrt{3}\right)$. Q.E.D.

## Problem 2


Plot  the boundary of the stability region of the BDF4 method 

$$u_{n+4} -\frac{48}{25} u_{n+3} + \frac{36}{25} u_{n+2} - \frac{16}{25} u_{n+1} + \frac{3}{25} u_n = h \frac{12}{25} f_{n+4}.$$

Which side of the region is stable?

___
As above, to find the RAS for the BDF4 method, we apply the method to the test problem $u^{'}(t)=\lambda u(t) \equiv f(u, t)$, which means that $f_{i} \equiv \lambda u_{i}$. With this, we get the following recurrence relation:

$$
u_{n+4}-\frac{48}{25}u_{n+3}+\frac{36}{25}u_{n+2}-\frac{16}{25}u_{n+1}+\frac{3}{25}u_{n}=\frac{12h\lambda}{25}u_{n+4} \hspace{0.5cm} (1)
$$

Defining $z:=h\lambda$, $(1)$ becomes

$$
\left(1-\frac{12z}{25}\right)u_{n+4}-\frac{48}{25}u_{n+3}+\frac{36}{25}u_{n+2}-\frac{16}{25}u_{n+1}+\frac{3}{25}u_{n}=0 \hspace{0.5cm} (2)
$$

Now, we want to study the characteristic polynomial $\pi(\zeta ; z)$ corresponding to $(2)$, which has four roots: $\zeta_{1}$, $\zeta_{2}$, $\zeta_{3}$, and $\zeta_{4}$. Here it is:

$$
\pi(\zeta_{1234} ; z)=\left(1-\frac{12z}{25}\right)\zeta_{1234}^{4}-\frac{48}{25}\zeta_{1234}^{3}+\frac{36}{25}\zeta_{1234}^{2}-\frac{16}{25}\zeta_{1234}+\frac{3}{25}=0 \hspace{0.5cm} (3)
$$

Also, as before, we want to find the set of all points $z$, *in the complex plane*, such that $\pi(\zeta ; z)$ satisfies the root condition. Thus, we solve for the boundary of this set by substituting $e^{i\theta}$ for $\zeta_{1234}$ in $(3)$, giving us

$$
\left(1-\frac{12z}{25}\right)e^{4i\theta}-\frac{48}{25}e^{3i\theta}+\frac{36}{25}e^{2i\theta}-\frac{16}{25}e^{i\theta}+\frac{3}{25}=0 \hspace{0.5cm} \Rightarrow \frac{12z}{25}e^{4i\theta}=e^{4i\theta}-\frac{48}{25}e^{3i\theta}+\frac{36}{25}e^{2i\theta}-\frac{16}{25}e^{i\theta}+\frac{3}{25} \\ \Rightarrow z=\frac{e^{4i\theta}-\frac{48}{25}e^{3i\theta}+\frac{36}{25}e^{2i\theta}-\frac{16}{25}e^{i\theta}+\frac{3}{25}}{\frac{12}{25}e^{4i\theta}}=\frac{25e^{4i\theta}-48e^{3i\theta}+36e^{2i\theta}-16e^{i\theta}+3}{12e^{4i\theta}} \\ \boxed{\therefore  z=\frac{25-48e^{-i\theta}+36e^{-2i\theta}-16e^{-3i\theta}+3e^{-4i\theta}}{12}} \hspace{0.5cm} (4)
$$

Using $(4)$, we can now plot the boundary in the complex plane via a mesh in $\theta$, where $\theta\in[0,2\pi]$.

```julia
using PyPlot

N = 2018
ts = range(0, stop = 2 * pi, length = N)
zeta = exp.(im * ts)
z = (1 / 12) * (25 .- 48 * zeta .^ (-1) .+ 36 * zeta .^ (-2) .- 16 * zeta .^ (-3) .+ 3 * zeta .^ (-4)) 

figure(1, figsize = (10, 10))
plot(real(z), imag(z), "r")
xlabel(L"\operatorname{Re}(h\lambda)")
ylabel(L"\operatorname{Im}(h\lambda)")
title("Boundary of the region of absolute stability for the BDF4 method")
```

$\therefore$ As we will soon show, the RAS is actually outside of this boundary (i.e. the outer region consists of stable points, and the inner region of unstable ones).

To test which side of the boundary is stable (*if* any), we try out $z=-1$ and $z=1$ for the outer and inner region of the boundary locus, respectively. In both cases, we solve for the roots to see if they allow $\pi$ to satisfy the root condition. However, since this is a quartic equation whose analytical solutions could be quite messy, we use Newton's method to solve for the roots. As a reminder, *all* four roots need to be located within the unit circle in the complex plane (i.e. $\left|\zeta_{i}\right|\leq 1$, $\forall i=1,2,3,4$).

**z=-1:**

$$
\pi(\zeta ; -1)=\frac{37}{25}\zeta_{1234}^{4}-\frac{48}{25}\zeta_{1234}^{3}+\frac{36}{25}\zeta_{1234}^{2}-\frac{16}{25}\zeta_{1234}+\frac{3}{25}
$$

Therefore, Newton's method gives the following iterative scheme:

$$
\Rightarrow \pi^{'}(\zeta ; -1)=\frac{148}{25}\zeta_{1234}^{3}-\frac{144}{25}\zeta_{1234}^{2}+\frac{72}{25}\zeta_{1234}-\frac{16}{25} \\
x_{n+1}=x_{n}-\frac{\pi\left(x_{n};-1\right)}{\pi^{'}\left(x_{n};-1\right)} \Rightarrow x_{n+1}=x_{n}-\frac{\frac{37}{25}x_{n}^{4}-\frac{48}{25}x_{n}^{3}+\frac{36}{25}x_{n}^{2}-\frac{16}{25}x_{n}+\frac{3}{25}}{\frac{148}{25}x_{n}^{3}-\frac{144}{25}x_{n}^{2}+\frac{72}{25}x_{n}-\frac{16}{25}} \\ \Rightarrow \boxed{x_{n+1}=x_{n}-\frac{37x_{n}^{4}-48x_{n}^{3}+36x_{n}^{2}-16x_{n}+3}{148x_{n}^{3}-144x_{n}^{2}+72x_{n}-16}}
$$

If we use a variety of initial guesses, we can eventually hope to find all of the roots. Thus, we set up a grid of initial conditions over the complex plane and test Newton's method on each sample point in the grid.

```julia
#Set up array of initial guesses
N = 10
s = range(-5, stop = 5, length = N)
e = ones(ComplexF64, N, 1)
tr = s * e'
ti = tr'
z = tr .+ im * ti #Grid of complex numbers
zs = z[:]

#Newton's method
tol = 1e-12 #Set the error tolerance 
rounding = 1e-12 #Set the rounding amount at the end for saving each computed root
maxit = 5000 #Set iteration limit
i = 0
roots = zeros(ComplexF64, length(zs))
for j = 1 : N ^ 2 #Iterate through all of the initial conditions on the grid
x = copy(zs[j])
xnew = copy(zs[j]) + 1
    while (abs(x - xnew) > tol) && (i < maxit)
        x = copy(xnew)
        xnew = x - ((37 * x ^ 4 - 48 * x ^ 3 + 36 * x ^ 2 - 16 * x + 3) / (148 * x ^ 3 - 144 * x ^ 2 + 72 * x - 16))
        i = i + 1
    end
    roots[j] = rounding * round(xnew / rounding)
end
#Filter out redundant roots
unique(roots)
```

```julia
#Print amplitudes
unique(abs.(roots))
```

$\therefore$ For $z=-1$, we found all four roots: $\zeta_{1}\approx0.213784188815+0.592476368456i$, $\zeta_{2}\approx0.213784188815-0.592476368456i$, $\zeta_{3}\approx0.434864459834+0.123553085032i$, and $\zeta_{4}\approx0.434864459834-0.123553085032i$. These roots all lie within the unit circle in the complex plane (as their amplitudes are below 1), so the outer region of the boundary locus consists of absolutely stable points.

**z=1:**

$$
\pi(\zeta ; 1)=\frac{13}{25}\zeta_{1234}^{4}-\frac{48}{25}\zeta_{1234}^{3}+\frac{36}{25}\zeta_{1234}^{2}-\frac{16}{25}\zeta_{1234}+\frac{3}{25}
$$

Therefore, Newton's method gives the following iterative scheme:

$$
\Rightarrow \pi^{'}(\zeta ; 1)=\frac{52}{25}\zeta_{1234}^{3}-\frac{144}{25}\zeta_{1234}^{2}+\frac{72}{25}\zeta_{1234}-\frac{16}{25} \\
x_{n+1}=x_{n}-\frac{\pi\left(x_{n};1\right)}{\pi^{'}\left(x_{n};1\right)} \Rightarrow x_{n+1}=x_{n}-\frac{\frac{13}{25}x_{n}^{4}-\frac{48}{25}x_{n}^{3}+\frac{36}{25}x_{n}^{2}-\frac{16}{25}x_{n}+\frac{3}{25}}{\frac{52}{25}x_{n}^{3}-\frac{144}{25}x_{n}^{2}+\frac{72}{25}x_{n}-\frac{16}{25}} \\ \Rightarrow \boxed{x_{n+1}=x_{n}-\frac{13x_{n}^{4}-48x_{n}^{3}+36x_{n}^{2}-16x_{n}+3}{52x_{n}^{3}-144x_{n}^{2}+72x_{n}-16}}
$$

Again, hoping to find all the roots via Newton's method, we set up a test grid of initial conditions over the complex plane.

```julia
N = 10
s = range(-5, stop = 5, length = N)
e = ones(ComplexF64, N, 1)
tr = s * e'
ti = tr'
z = tr .+ im * ti 
zs = z[:]

tol = 1e-12  
rounding = 1e-12
maxit = 5000 
i = 0
roots = zeros(ComplexF64, length(zs))
for j = 1 : N ^ 2
x = copy(zs[j])
xnew = copy(zs[j]) + 1
    while (abs(x - xnew) > tol) && (i < maxit)
        x = copy(xnew)
        xnew = x - ((13 * x ^ 4 - 48 * x ^ 3 + 36 * x ^ 2 - 16 * x + 3) / (52 * x ^ 3 - 144 * x ^ 2 + 72 * x - 16))
        i = i + 1
    end
    roots[j] = rounding * round(xnew / rounding)
end
unique(roots)
```

```julia
unique(abs.(roots))
```

$\therefore$ For $z=1$, we still found all four roots: $\zeta_{1}\approx2.866142980578$, $\zeta_{2}\approx0.341424819296$, $\zeta_{3}\approx0.242369946217+0.420807745765i$, and $\zeta_{4}\approx0.242369946217-0.420807745765i$, but unfortunately, $\zeta_{1}>1$. Thus, this implies that the inner region of the boundary locus consists of points which are *not* absolutely stable.

## Problem 3

For the inhomogeneous linear equation

$$u'(t)=\lambda u(t) + a,\quad a,u(t)\in\mathbb{R}.$$

consider the Runge-Kutta method with Butcher tableau $A\in\mathbb{R}^{r\times r},b\in\mathbb{R}^{r},Re\lambda<0$

a) Prove that the numerical solution can be written as

$$u_n = R(\lambda h)^n u_0 + \left[\frac{R(\lambda h)^n-1}{R(\lambda h)-1}\right] h b^T(I-zA)^{-1}e a.$$

Here $e=(1,\ldots,1)^T\in\mathbb{R}^{r}$,

$$R(z)=1+zb^T(I-zA)^{-1}e.$$

b) Assume $|R(z)|<1$, prove that 

$$\lim_{n\to\infty} u_n=-\frac{a}{\lambda}.$$

Verify that this limit is indeed the limit of the true solution i.e.

$$\lim_{t\to+\infty} u(t)=-\frac{a}{\lambda}.$$

___
**Part a:**

**Proof:** 
With the given Butcher tableau, a general $r$-stage Runge-Kutta (RK) method can be written in the following form, with $e\equiv \overrightarrow{e}$, $A$, and $b\equiv \overrightarrow{b}$ defined as above, and weight vector $\overrightarrow{k}$ defined through recursion:

$$
\overrightarrow{k}=f\left(u_{n}\overrightarrow{e}+hA\overrightarrow{k},t_{n}\overrightarrow{e}+\overrightarrow{c}h\right) \\ u_{n+1}=u_{n}+h\overrightarrow{b}^{T}\overrightarrow{k} \hspace{0.5cm} (1)
$$

For the given inhomogeneous linear equation, i.e. $u^{'}(t)=\lambda u(t)+a:=f\left(u(t),t\right)$, $(1)$ becomes:

$$
\overrightarrow{k}=\lambda\left(u_{n}\overrightarrow{e}+hA\overrightarrow{k}\right)+a\overrightarrow{e}=\lambda u_{n}\overrightarrow{e}+h\lambda A\overrightarrow{k}+a\overrightarrow{e} \Rightarrow \left(I-h\lambda A\right)\overrightarrow{k}=\lambda u_{n}\overrightarrow{e}+a\overrightarrow{e} \hspace{0.5cm} (2)
$$

Note that $I\in\mathbb{R}^{r\times r}$ is the $r$-dimensional identity matrix. Now, since our choice of stepsize, $h$, is arbitrary, assume that we can pick an $h$ such that $I-h\lambda A$ is invertible. Then, we can muliply both sides of $(2)$ by $\left(I-h\lambda A\right)^{-1}$, which gives us, noting that $\lambda$ and $u_{n}$ are scalars in $\mathbb{R}$,

$$
\overrightarrow{k}=\lambda u_{n}\left(I-h\lambda A\right)^{-1}\overrightarrow{e}+a\left(I-h\lambda A\right)^{-1}\overrightarrow{e} \\ (1) \Rightarrow u_{n+1}=u_{n}+h\lambda u_{n}\overrightarrow{b}^{T}\left(I-h\lambda A\right)^{-1}\overrightarrow{e}+ha\overrightarrow{b}^{T}\left(I-h\lambda A\right)^{-1}\overrightarrow{e} \\ u_{n+1}=\left(1+h\lambda \overrightarrow{b}^{T}\left(I-h\lambda A\right)^{-1}\overrightarrow{e}\right)u_{n}+ha\overrightarrow{b}^{T}\left(I-h\lambda A\right)^{-1}\overrightarrow{e} \hspace{0.5cm} (3)
$$

As suggested by what we are trying to show above, we can define $z:=h\lambda$ and $R(z):=1+z\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}$. Furthermore, we should shift indices from $n \rightarrow n-1$. From these, $(3)$ simplifies to

$$
u_{n}=R(z)u_{n-1}+ha\overrightarrow{b}^{T}\left(I-z A\right)^{-1}\overrightarrow{e} \hspace{0.5cm} (4)
$$

As $(4)$ suggests, we should claim the following to finish our proof:

**Lemma:** $u_{n}=R(z)^{k}u_{n-k}+ha\left[\frac{R(z)^{k}-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e} \hspace{0.2cm} \forall k\in 0, 1, \dots, n$

We prove this lemma by induction. 

*Base case:* Suppose $k=0$. Then,

$$
u_{n}=R(z)^{0}u_{n-0}+ha\left[\frac{R(z)^{0}-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}=u_{n}+ha\left[\frac{1-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}=u_{n} \Rightarrow \therefore u_{n}=u_{n}
$$

*Inductive hypothesis:* Now, suppose that $u_{n}=R(z)^{i}u_{n-i}+ha\left[\frac{R(z)^{i}-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e} \hspace{0.2cm} \forall i\in 0, 1, \dots, n-1$.

*Inductive step:* By shifting the indices such that $n \rightarrow n-i$, $(4)$ gives us that

$$
u_{n-i}=R(z)u_{n-(i+1)}+ha\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e} \hspace{0.5cm} (5)
$$

Then, plugging $(5)$ into our inductive hypothesis, we get that

$$
u_{n}=R(z)^{i}\left(R(z)u_{n-(i+1)}+ha\overrightarrow{b}^{T}\left(I-z A\right)^{-1}\overrightarrow{e}\right)+ha\left[\frac{R(z)^{i}-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e} \\ u_{n}=R(z)^{i+1}u_{n-(i+1)}+haR(z)^{i}\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}+ha\left[\frac{R(z)^{i}-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e} \\ u_{n}=R(z)^{i+1}u_{n-(i+1)}+\left[R(z)^{i}+\frac{R(z)^{i}-1}{R(z)-1}\right]ha\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e} \hspace{0.5cm} (6)
$$

Now, note that $R(z)^{i}+\frac{R(z)^{i}-1}{R(z)-1}=\frac{R(z)^{i}\left(R(z)-1\right)+R(z)^{i}-1}{R(z)-1}=\frac{R(z)^{i+1}-R(z)^{i}+R(z)^{i}-1}{R(z)-1}=\frac{R(z)^{i+1}-1}{R(z)-1}$. Then, $(6)$ becomes

$$
\therefore u_{n}=R(z)^{i+1}u_{n-(i+1)}+
\left[\frac{R(z)^{i+1}-1}{R(z)-1}\right]ha\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e} 
$$

Therefore, by induction, our lemma is proven. In particular, if we let $k \rightarrow n$ in our lemma, we get that $\boxed{\therefore u_{n}=R(z)^{n}u_{0}+ha\left[\frac{R(z)^{n}-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}}$. Q.E.D.

**Part b:**

**Proof:** We start off by proving a lemma which will be useful later.

**Lemma:** As long as $\left|R(z)\right|<1$, $R(z)^{n}\xrightarrow{n\rightarrow\infty}0$.

*Proof of lemma:* Let $\varepsilon>0$. Suppose that $\left|R(z)\right|=\frac{1}{1+r}\Rightarrow r=\frac{1}{|R(z)|}-1$ and $N=\frac{\left|R(z)\right|}{\varepsilon\left(1-\left|R(z)\right|\right)}=\frac{1}{\varepsilon r}$. Then, $\forall n>N$, we have that

$$
\left|R(z)^{n}\right|=\left|R(z)\right|^{n}=\left|\frac{1}{1+r}\right|^{n}=\frac{1}{\left|1+r\right|^{n}} \hspace{0.5cm} (*)
$$

Furthermore, by the Bernoulli inequality (which only works because $\left|R(z)\right|<1$), we can bound $(*)$:

$$
\left|R(z)^{n}\right|=\frac{1}{\left|1+r\right|^{n}}\leq \frac{1}{1+nr}<\frac{1}{nr}<\frac{1}{Nr}=\frac{\varepsilon\left(1-\left|R(z)\right|\right)}{\left|R(z)\right|\left(\frac{1}{|R(z)|}-1\right)}=\frac{\varepsilon\left(1-\left|R(z)\right|\right)}{\left(1-\left|R(z)\right|\right)}=\varepsilon \\ \Rightarrow \therefore \left|R(z)^{n}\right|<\varepsilon
$$

Therefore, by the definition of limit, our lemma is proven. Now, going back to the main proof, we note that $u_{n}:= v_{n}+w_{n}$, where $v_{n}:=R(z)^{n}u_{0}$ and $w_{n}:=ha\left[\frac{R(z)^{n}-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}$. Thus, by algebraic properties of limits, if $v_{n}\xrightarrow{n\rightarrow\infty}v$ and $w_{n}\xrightarrow{n\rightarrow\infty}w$ for some $v,w\in\mathbb{R}$, then $u_{n}\xrightarrow{n\rightarrow\infty}v+w \hspace{0.15cm} (1)$. Thus, we need to check if these two limits exist, and if so, what they are.

**v:**

$$
\lim _{n\rightarrow\infty}{v_{n}}=\lim _{n\rightarrow\infty}{R(z)^{n}u_{0}}=u_{0}\lim _{n\rightarrow\infty}{R(z)^{n}} \hspace{0.5cm} (2)
$$

Now, since $\left|R(z)\right|<1$, $(2)$ must go to zero, so $\therefore \lim _{n\rightarrow\infty}{v_{n}}=v=0 \hspace{0.15cm} (3)$.

**w:**

This limit is a little trickier, but not by much.

$$
\lim _{n\rightarrow\infty}{w_{n}}=\lim _{n\rightarrow\infty}{ha\left[\frac{R(z)^{n}-1}{R(z)-1}\right]\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}}=ha\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}\lim _{n\rightarrow\infty}{\left[\frac{R(z)^{n}-1}{R(z)-1}\right]} \\ \lim _{n\rightarrow\infty}{w_{n}}=ha\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}\left(\frac{1}{R(z)-1}\lim _{n\rightarrow\infty}{R(z)^{n}}-\frac{1}{R(z)-1}\right) \hspace{0.5cm} (4)
$$

Again, since $\left|R(z)\right|<1$, $\lim _{n\rightarrow\infty}{R(z)^{n}}=0$, so $(4)$ simplifies to

$$
\lim _{n\rightarrow\infty}{w_{n}}=-\frac{ha\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}}{R(z)-1}=-\frac{ha\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}}{z\overrightarrow{b}^{T}\left(I-zA\right)^{-1}\overrightarrow{e}}=-\frac{ha}{z}=-\frac{ha}{h\lambda}=-\frac{a}{\lambda} \\ \Rightarrow \therefore \lim _{n\rightarrow\infty}{w_{n}}=-\frac{a}{\lambda} \hspace{0.5cm} (5)
$$

Therefore, by $(1)$, $(3)$, and $(5)$, $\boxed{\therefore u_{n}\xrightarrow{n\rightarrow\infty}-\frac{a}{\lambda}}$. Q.E.D.

Now, we check that this is also the limit of the actual solution, i.e. that $\lim _{t\rightarrow\infty}{u(t)}=-\frac{a}{\lambda}$. Firstly, we calculate what $u(t)$ is, noting that $u(t)=u_{h}(t)+u_{p}(t)$, where $u_{h}(t)$ is the solution to the equivalent homogeneous system $u'(t)=\lambda u(t)$, and $u_{p}(t)$ is some particular solution to $u'(t)=\lambda u(t)+a$.

**Homogeneous solution:** For $u_{h}'(t)=\lambda u_{h}(t)$, we guess a solution of the form $u_{h}(t)=Ce^{rt}$, where C is an arbitrary constant, $\lambda$ to be determined, and plug it into the ODE. Thus, we solve the resulting characteristic equation and get that $r=\lambda$. $\therefore u_{h}(t)=Ce^{\lambda t}$.

**Particular solution:** For $u_{p}'(t)=\lambda u_{p}(t)+a$, we guess a solution of the form $u_{p}(t)=A+Bt$, where $A, B$ are constants to be determined. Plugging it into the ODE, we get

$$
B=A\lambda + B\lambda t + a \Rightarrow B=0 \Rightarrow \\ A\lambda=-a \Rightarrow A=-\frac{a}{\lambda}
$$

$\therefore u_{p}(t)=-\frac{a}{\lambda}$. From here, our complete solution is $u(t)=u_{h}(t)+u_{p}(t)=Ce^{\lambda t}-\frac{a}{\lambda}$. Solving for $C$ by matching this with the initial condition $u(0)=u_{0}$, we get that $u(t)=u_{0}e^{\lambda t}+\frac{a}{\lambda}\left(e^{\lambda t}-1\right)$. Furthermore, since $\operatorname{Re}(\lambda)<0$, we have the fact that $\lim _{t\rightarrow\infty}{e^{\lambda t}}=0$. Thus,

$$
\lim _{t\rightarrow\infty}{u(t)}=\lim _{t\rightarrow\infty}{\left[u_{0}e^{\lambda t}+\frac{a}{\lambda}\left(e^{\lambda t}-1\right)\right]}=u_{0}\lim _{t\rightarrow\infty}{e^{\lambda t}}+\frac{a}{\lambda}\lim _{t\rightarrow\infty}{e^{\lambda t}}-\frac{a}{\lambda}=-\frac{a}{\lambda}
$$

$\boxed{\therefore \lim _{t\rightarrow\infty}{u(t)}=-\frac{a}{\lambda}}$. Q.E.D.

**Part a:**

First off, since this is a second-order nonlinear ordinary differential equation, we transform the system into an equivalent two-dimensional system by introducing a new variable (think of this as a velocity), $v(t):=\dot{u}(t)$. Then, defining $\overrightarrow{u}:=\begin{pmatrix} u \\ v \end{pmatrix}$ and $\frac{\partial \overrightarrow{u}}{\partial t}:=\dot{\overrightarrow{u}}=\begin{pmatrix} \dot{u} \\ \dot{v} \end{pmatrix}:=\overrightarrow{f}\left(t, \overrightarrow{u}\right)$, our dynamical system can be written as the following equivalent two-dimensional system:

$$
\dot{\overrightarrow{u}}=\begin{pmatrix} v \\ \frac{1}{\varepsilon}\left(\left(1-u^{2}\right)v-u\right) \end{pmatrix}=\overrightarrow{f}\left(t, \overrightarrow{u}\right)
$$

Before applying RK4 (i.e. *the* Runge-Kutta method), let us define a function (in Julia) which evaluates $\overrightarrow{f}\left(t, \overrightarrow{u}\right)$ for given values of $\overrightarrow{u}$ and $\varepsilon$.

```julia
using LinearAlgebra

function vdp(u, eps)
    return [u[2], (1 / eps) * ((u[2] * (1 - u[1]^2)) - u[1])]
end
```

The Julia function we defined here will be very useful later. Next, given an IVP of the form $\dot{\overrightarrow{u}}=\overrightarrow{f}\left(t, \overrightarrow{u}\right)$ with initial condition $\overrightarrow{u(0)}:=\overrightarrow{u_{0}}$, a general RK4 scheme is defined as the following, with stepsize $h:=t_{n+1}-t_{n} \forall n$ chosen beforehand:

$$
\overrightarrow{u_{n+1}}=\overrightarrow{u_{n}}+\frac{h}{6}\left(\overrightarrow{k_{n,1}}+2\overrightarrow{k_{n,2}}+2\overrightarrow{k_{n,3}}+\overrightarrow{k_{n,4}}\right)
$$

where each $\overrightarrow{k_{n,j}}$ is defined $\forall j \in \{1, 2, 3, 4\}$ as:

$$
\overrightarrow{k_{n,1}}=\overrightarrow{f}\left(t_{n}, \overrightarrow{u_{n}}\right) \\
\overrightarrow{k_{n,2}}=\overrightarrow{f}\left(t_{n}+\frac{h}{2}, \overrightarrow{u_{n}}+\frac{h}{2}\overrightarrow{k_{n,1}}\right) \\
\overrightarrow{k_{n,3}}=\overrightarrow{f}\left(t_{n}+\frac{h}{2}, \overrightarrow{u_{n}}+\frac{h}{2}\overrightarrow{k_{n,2}}\right) \\
\overrightarrow{k_{n,4}}=\overrightarrow{f}\left(t_{n}+h, \overrightarrow{u_{n}}+h\overrightarrow{k_{n,3}}\right) 
$$

Thus, accordingly, we can use the RK4 method to numerically solve our IVP in question, given that we have already put it in the desired form of $\dot{\overrightarrow{u}}=\overrightarrow{f}\left(t, \overrightarrow{u}\right)$ above. We will do this below and plot the numerical solution:

```julia
using LinearAlgebra

# RK4
function rk4(N, T, u0, eps)
    h = T / N
    t = h * collect(0 : N)
    u = zeros(2, N + 1)
    u[:, 1] = u0
    for i = 1 : N
        ui = u[:, i]
        ti = t[i]
        k1 = vdp(ui, eps)
        k2 = vdp(ui + 0.5 * h * k1, eps)
        k3 = vdp(ui + 0.5 * h * k2, eps)
        k4 = vdp(ui + h * k3, eps)
        u[:, i + 1] = ui + (h / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
    end
    return u
end

using PyPlot

##############################
T = 6.0
h = 0.001
N = round(Int64, T / h)
t = collect(0 : N) * h
eps = 0.1
u0 = [2.0, 0.0]

#RK4
urk4 = rk4(N, T, u0, eps)

figure(1, figsize = (5, 5))
plot(t, vec(urk4[1, :]), "r-")
xlabel(L"t")
ylabel(L"u(t)")
```

**Part b:**

As stated earlier in matrix form, an arbitrary $r$-stage Runge-Kutta (RK) method is of the form:

$$
\overrightarrow{k_{1}}=f\left(\overrightarrow{u_{n}}+h\left(a_{11}\overrightarrow{k_{1}}+\dots+a_{1r}\overrightarrow{k_{r}}\right),t_{n}+c_{1}h\right) \\ \vdots \\ \overrightarrow{k_{r}}=f\left(\overrightarrow{u_{n}}+h\left(a_{r1}\overrightarrow{k_{1}}+\dots+a_{rr}\overrightarrow{k_{r}}\right),t_{n}+c_{r}h\right) \\ \overrightarrow{u_{n+1}}=\overrightarrow{u_{n}}+h\left(b_{1}\overrightarrow{k_{1}}+\dots+b_{r}\overrightarrow{k_{r}}\right)
$$

Now, the given Butcher tableau for the two-step Gauss-Legendre method (GL2) describes a two-step method, so in our case, $r=2$, and $a_{11}=\frac{1}{4}$, $a_{12}=\frac{1}{4}-\frac{\sqrt{3}}{6}$, $a_{21}=\frac{1}{4}+\frac{\sqrt{3}}{6}$, $a_{22}=\frac{1}{4}$, and $b_{1}=b_{2}=\frac{1}{2}$. (We can neglect $c_{1}$ and $c_{2}$ because the van der Pol equation is autonomous.) Therefore, our GL2 scheme is the following, where $f$ is defined as in part a:

$$
\overrightarrow{k_{1}}=f\left(\overrightarrow{u_{n}}+h\left(a_{11}\overrightarrow{k_{1}}+a_{12}\overrightarrow{k_{2}}\right)\right)=f\left(\overrightarrow{u_{n}}+h\left(\frac{1}{4}\overrightarrow{k_{1}}+\left(\frac{1}{4}-\frac{\sqrt{3}}{6}\right)\overrightarrow{k_{2}}\right)\right) \\ \overrightarrow{k_{2}}=f\left(\overrightarrow{u_{n}}+h\left(a_{21}\overrightarrow{k_{1}}+a_{22}\overrightarrow{k_{2}}\right)\right)=f\left(\overrightarrow{u_{n}}+h\left(\left(\frac{1}{4}+\frac{\sqrt{3}}{6}\right)\overrightarrow{k_{1}}+\frac{1}{4}\overrightarrow{k_{2}}\right)\right) \\ \overrightarrow{u_{n+1}}=\overrightarrow{u_{n}}+h\left(b_{1}\overrightarrow{k_{1}}+b_{2}\overrightarrow{k_{2}}\right)=\overrightarrow{u_{n}}+\frac{h}{2}\left(\overrightarrow{k_{1}}+\overrightarrow{k_{2}}\right) \hspace{0.5cm} (1)
$$

Solving $(1)$ at each timestep requires solving a nonlinear system of equations for $\overrightarrow{k_{1}}$ and $\overrightarrow{k_{2}}$, which is not explicitly solvable. Thus, we implement the following fixed-point iterative solver to approximate $\overrightarrow{u_{n+1}}$ for each timestep, halting the iterations in $i$ once the absolute difference (i.e. the error) between successive iterations falls below a given tolerance, which we set to be $10^{-12}$ in the code below. According to the theory, given that $h$ is small enough, $\overrightarrow{k_{1}^{i}}\xrightarrow{i\rightarrow\infty}\overrightarrow{k_{1}}$ and $\overrightarrow{k_{2}^{i}}\xrightarrow{i\rightarrow\infty}\overrightarrow{k_{2}}$, but of course, it is not feasible to implement an infinite number of iterations.

$$
\overrightarrow{k_{1}^{i+1}}=f\left(\overrightarrow{u_{n}}+h\left(\frac{1}{4}\overrightarrow{k_{1}^{i}}+\left(\frac{1}{4}-\frac{\sqrt{3}}{6}\right)\overrightarrow{k_{2}^{i}}\right)\right) \\ \overrightarrow{k_{2}^{i+1}}=f\left(\overrightarrow{u_{n}}+h\left(\left(\frac{1}{4}+\frac{\sqrt{3}}{6}\right)\overrightarrow{k_{1}^{i}}+\frac{1}{4}\overrightarrow{k_{2}^{i}}\right)\right) \hspace{0.5cm} (2)
$$

Thus, once $\left\|\begin{pmatrix} \overrightarrow{k_{1}^{N}}-\overrightarrow{k_{1}^{N-1}} \\ \overrightarrow{k_{2}^{N}}-\overrightarrow{k_{1}^{N-1}} \end{pmatrix}\right\|_{2}<10^{-12}$ for some $N$, then we can approximate $\overrightarrow{u_{n+1}}\approx\overrightarrow{u_{n}}+\frac{h}{2}\left(\overrightarrow{k_{1}^{N}}+\overrightarrow{k_{2}^{N}}\right)$. To make an initial guess for our iterative solver, we will let $\overrightarrow{k_{1}^{i=0}}=\overrightarrow{k_{2}^{i=0}}=f\left(\overrightarrow{u_{n}},t_{n}\right)$ for each $n$th timestep. We expect that this iterative method should converge at each step, provided that our timestep, $h$, is small enough. First off, we try $h=0.01$.

```julia
function gl2fixpoint(N, T, u0, eps, tol, maxiterations)
    #setup and coefficients from Butcher tableau
    h = T / N
    A = [0.25 0.25 - (sqrt(3) / 6); 0.25 + (sqrt(3) / 6) 0.25]
    b = [0.5 0.5]
    u = zeros(2, N + 1)
    u[:, 1] = u0
    fu = zeros(2, N + 1)
    iterations = zeros(N)
    #numerical scheme
    for i = 1 : N
        ui = u[:, i]
        #nonlinear iterative solver for k1 and k2
        iter = 0
        k1 = vdp(ui, eps)
        k2 = k1
        while iter < maxiterations
            k1m = vdp(ui + h * (A[1, 1] * k1 + A[1, 2] * k2), eps)
            k2m = vdp(ui + h * (A[2, 1] * k1 + A[2, 2] * k2), eps)
            if(norm([k1 - k1m; k2 - k2m]) < tol)
                break
            end
            k1 = k1m
            k2 = k2m
            iter += 1
        end
        iterations[i] = iter
        u[:, i + 1] = ui + h * (b[1] * k1 + b[2] * k2)
    end
    return u, iterations
end

using PyPlot

##############################
T = 6.0
h = 0.01
N = round(Int64, T / h)
t = collect(0 : N) * h
eps = 0.1
tol = 1e-12
maxiterations = 100
u0 = [2.0, 0.0]

#GL2
ugl2, iterations = gl2fixpoint(N, T, u0, eps, tol, maxiterations)

figure(1, figsize = (5, 5))
plot(t, vec(ugl2[1, :]), "r-")
xlabel(L"t")
ylabel(L"u(t)")

figure(2, figsize = (5, 5))
plot(t[2: N + 1], iterations, "g-")
xlabel(L"t")
ylabel(L"N_{iterations}")
```

An interesting observation to note is that the fixed-point iterative solver required more iterations whenever $u(t)$ was increasing or decreasing (i.e. changing) a lot. This is expected, as our initial guess uses the previous function value as an approximation, so whenever $u(t)$ is changing the most, we should expect the difference between successive function values to be the greatest. Thus, we should expect that more iterations are required for convergence, since the initial error is the greatest. Anyways, we now try $h=0.1$ and see if our numerical solver works as well as above.

```julia
using PyPlot
T = 6.0
h = 0.1
N = round(Int64, T / h)
t = collect(0 : N) * h
eps = 0.1
tol = 1e-12
maxiterations = 300
u0 = [2.0, 0.0]

#GL2
ugl2, iterations = gl2fixpoint(N, T, u0, eps, tol, maxiterations)

figure(1, figsize = (5, 5))
plot(t, vec(ugl2[1, :]), "r-")
xlabel(L"t")
ylabel(L"u(t)")

figure(2, figsize = (5, 5))
plot(t[2: N + 1], iterations, "g-")
xlabel(L"t")
ylabel(L"N_{iterations}")
```

Evidently, our GL2 numerical solver did not perform nearly as well this time, and we observe that the number of iterations required for our nonlinear solver at each step grew tremendously, especially near where $u(t)$ changed the most. Again, this is as expected.

**Part c:**

Now, we want to test the convergence of the two-step Gauss-Legendre method (GL2) as we decrease the stepsize, $h$. We do this by increasing the number of steps for our scheme, $N$, without changing the size of our interval, $[0,T]$, testing GL2 at $N_q$ samples of different $N$ values, ranging from $2^{6}$ to $2^{11}$. Afterwards, we make a log-log plot of the error $||u_{GL2}(T)-u_{RK4}(T)||\approx||u_{GL2}(T)-u(T)||$ vs. the number of steps used, $N$. We approximate the error in the GL2 numerical solution by using the RK4 solution in place of the actual solution. 

The slope of this plot should be roughly equal to the order of convergence for our numerical scheme. Why? Well, for $C$ a constant, and $u(T)$ the exact solution at $t=T$,

$$
u(T)=u_{GL2}(T)+Ch^{p}+O\left(h^{p+1}\right) \Rightarrow ||u_{GL2}(T)-u(T)||\approx Ch^{p} \Rightarrow ln||u_{GL2}(T)-u(T)||\approx ln\left(Ch^{p}\right) \\ \Rightarrow\therefore ln||u_{GL2}(T)-u(T)||=ln(C)+pln(h)  
$$

```julia
T = 6.0
eps = 0.1
tol = 1e-12
maxiterations = 300
u0 = [2.0, 0.0]

Nq = 10
stepnumber = round.(Int, 2 .^ (6 : (5 / (Nq - 1)) : 11))
errors = zeros(Nq)
for i = 1 : Nq 
    local i
    local ugl2
    local iterations
    N = stepnumber[i]
    ugl2, iterations = gl2fixpoint(N, T, u0, eps, tol, maxiterations)
    errors[i] = norm(urk4[1, end] - ugl2[1, end])
end

using PyPlot
figure(1, figsize = (5, 5))
loglog(stepnumber, errors, "b-^", label="GL2")
xlabel(L"ln(N)")
xscale("log")
ylabel(L"$ln||u_{GL2}(T)-u_{RK4}(T)||$")
yscale("log")
legend()

errorslope = -(log(errors[end]) - log(errors[1])) / (log(stepnumber[end]) - log(stepnumber[1]))
println("Slope for two-step Gauss-Legendre method = ", errorslope)
```

$\therefore$ Based off of the slope in the loglog plot above between number of timesteps and absolute error, we can say that the convergence rate for the two-step Gauss-Legendre method is roughly quartic (i.e. with $p \approx 4$ for an error bound of $Ch^{p}$, $C$ a constant), which is what we wanted to show. Actually, it is quite a bit less than 4, but we might expect that this slope will not be super accurate, since we used RK4 (*the* Runge-Kutta method) to approximate the actual solution.
