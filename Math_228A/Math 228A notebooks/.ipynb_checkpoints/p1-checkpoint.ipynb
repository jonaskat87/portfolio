## Problem 1

Consider $u'(t)=f(u(t)),u(0)=u_0$, and $f:\mathbb{R}^{d}\to \mathbb{R}$ 
is Lipschitz continuous with a
Lipschitz constant $L$. Let $\max \Vert u''(t)\Vert$ be bounded by a constant $M$ for $0\le t\le T$. Prove that the backward Euler method converges as the step size $h\to 0$.

**Proof:** 

First, consider the backward Euler method:

$$
u_{n+1} = u_{n} + h f(u_{n+1}) \hspace{0.5cm} (1)
$$

Next, we consider a similar equation which involves the exact solution for this IVP at $t=t_n$, i.e. $u(t_n)$:

$$ 
u(t_{n+1}) = u(t_{n}) + h f(u(t_{n+1})) +\tau_n \hspace{0.5cm} (2)
$$

$\tau_n$ is the LTE (local truncation error). Now, define $e_{n+1}:=u\left(t_{n+1}\right)-u_{n+1}$ and $e_{n}:=u\left(t_{n}\right)-u_{n}$, which are the errors between the exact and approximate solutions at $t=t_{n+1}$ and $t=t_n$, respectively. From here, after subtracting $(1)$ from $(2)$, we get the following:

$$
e_{n+1} = e_{n} + h (f(u(t_{n+1})) - f(u_{n+1})) +\tau_n \hspace{0.5cm} (3) 
$$

We now take the norm of both sides of $(3)$, then use the triangle inequality to separate each term out.

$$
\|e_{n+1}\|=\|e_{n} + h(f(u(t_{n+1})) - f(u_{n+1})) + \tau _{n} \| \le \|e_{n}\| + h\|f(u(t_{n+1})) - f(u_{n+1})\| + \|\tau _{n}\|  \hspace{0.5cm} (4) 
$$

Since $f$ is Lipschitz with Lipschitz constant $L$, we can bound $\| f(u(t_{n+1})) - f(u_{n+1})\| $ as such:

$$
\|f(u(t_{n+1})) - f(u_{n+1})\| \le L\|u(t_{n+1}) - u_{n+1}\| = L\|e_{n+1}\| \hspace{0.5cm} \\ (4) \Rightarrow \|e_{n+1}\| \le \|e_{n}\| + hL\|e_{n+1}\| + \|\tau _{n}\| \hspace{0.5cm} (5)
$$

Thus, subtracting to the other side:

$$
(1-hL) \|e_{n+1}\| \le \|e_{n}\| + \|\tau _{n}\| \hspace{0.5cm} (6)
$$

Now, for us to divide both sides of $(6)$ by $(1-hL)$ without having to switch the inequality, we must enforce the necessary condition that $hL<1$. With this, $(1-hL)$ is nonnegative, so we can divide without switching the inequality. We want this, because a switched inequality would indicate an unbounded error between our exact and approximate solutions, which we cannot have for a convergent scheme. Anyways, enforcing this condition gives the following:

$$
\|e_{n+1}\| \le \frac{1}{1-hL} (\|e_{n}\| + \|\tau _{n}\|) \hspace{0.5cm} (7) 
$$

Now, let $n+1=N$, where $N=\frac{T}{h}$. Hence, with this given inequality, we can relate successive timesteps backwards until we can relate the error at the $N$th timestep to the initial absolute error $\|e_{0}\|$ and the LTE at each timestep up to the $(N-1)$th one:

$$
\|e_{N}\| \le {\left(\frac{1}{1-hL}\right)}^{N}\|e_{0}\| + \frac { \|\tau_{N-1}\| }{1-hL} +\frac {\|\tau_{N-2} \|}{{(1-hL)}^{2}} +\dots +\frac {\|\tau_{ 0 }\|}{{(1-hL)}^{N}} \hspace{0.5cm} (8)
$$

Next, we proceed by bounding the LTE at each timestep. We first solve for the LTE using $(2)$:

$$
\tau_n = u(t_{n+1}) - u(t_{n}) - h f(u(t_{n+1})) \hspace{0.5cm} (9)
$$

We note that the antiderivative of $f(u(t))$ with respect to time is just the actual solution at time $t$. Therefore, by the first fundamental theorem of calculus:

$$
u(t_{n+1}) - u(t_{n}) = \int_{t_{n}}^{t_{n+1}}{f(u(s))}ds \hspace{0.5cm} (10)
$$

From here, we integrate by parts once, using the substitutions $m=f(u(s))$, $dm=f'(u(s))ds$, $n=s-t_{n}$, and $dn=ds$:

$$
u(t_{n+1}) - u(t_{n}) = f(u(s)) (s-t_{n}){|}_{{t}_{n}}^{{t}_{n+1}} - \int_{t_{n}}^{t_{n+1}}{f'(u(s))(s-t_{n})}ds \hspace{0.5cm} (11)
$$

Note that the original differential equation gives that $u'(t)=f(u(t)) \Rightarrow u''(t)=f'(u(t))$. Also, we again assume a uniform step size, so after we evaluate the left expression on the right side of $(12)$, we get

$$
u(t_{n+1}) - u(t_{n}) = h f(u(t_{n+1})) - \int_{t_{n}}^{t_{n+1}}{u''(s)(s-t_{n})}ds \\ \Rightarrow h f(u(t_{n+1})) = u(t_{n+1}) - u(t_{n}) + \int_{t_{n}}^{t_{n+1}}{u''(s)(s-t_{n})}ds \hspace{0.5cm} (12)    
$$

Plugging $(12)$ into $(9) gives the following:

$$
\tau_n = u(t_{n+1}) - u(t_{n}) - \left(u(t_{n+1}) - u(t_{n}) + \int_{t_{n}}^{t_{n+1}}{u''(s)(s-t_{n})}ds\right) = -\int_{t_{n}}^{t_{n+1}}{u''(s)(s-t_{n})}ds \hspace{0.5cm} (13)  
$$

Finally, as stated in the problem, assume that $\|u''(t)\|$ is uniformly bounded $\forall t \in [0, T]$, i.e. $\max _{0 \le t \le T}{\|u''(t)\|}=M$. Then, we take the norm of $(13)$, apply the triangle inequality for integrals, then finally use the Cauchy-Schwarz inequality to get that

$$
\|\tau_n\| = \|\int_{t_{n}}^{t_{n+1}}{u''(s)}(s-t_{n})ds\| \le \int_{t_{n}}^{t_{n+1}}{\|u''(s)(s-t_{n})\|}ds \le \int_{t_{n}}^{t_{n+1}}{\|u''(s)\| \|s-t_{n}\|}ds \\ \Rightarrow \|\tau_n\| \le M \int_{t_{n}}^{t_{n+1}}{\|s-t_{n}\|}ds = M \frac{(t_{n+1} - t_{n})^{2}}{2} \\ \Rightarrow \|\tau_n\| = \frac{Mh^{2}}{2} \hspace{0.5cm} (14)
$$

Now that $(14)$ gives a bound for the LTE at every timestep, we substitute $(14)$ for each LTE found in $(8)$, which gives that

$$
\|e_{N}\| \le {\left(\frac{1}{1-hL}\right)}^{N}\|e_{0}\| + \frac{Mh^{2}}{2} \left(\frac{1}{1-hL}\right) + \frac{Mh^{2}}{2} {\left(\frac{1}{1-hL}\right)}^{2} +\dots + \frac{Mh^{2}}{2} {\left(\frac{1}{1-hL}\right)}^{N} \\ \Rightarrow \|e_{N}\| \le {\left(\frac{1}{1-hL}\right)}^{N}\|e_{0}\| + \frac{Mh^{2}}{2} \left[\left(\frac{1}{1-hL}\right) + {\left(\frac{1}{1-hL}\right)}^{2} +\dots + {\left(\frac{1}{1-hL}\right)}^{N}\right] \hspace{0.5cm} (15)
$$

Note that the terms in the brackets form a geometric series with ratio $r = \frac{1}{1-hL}$ and initial term $a_{0} = \frac{1}{1-hL}$. The formula for a general geometric series with $n$ terms, ratio $r$, and initial term $a_{0}$ is $S_{n} = \frac{a_{0} (r^{n}-1)}{r-1}$, so in this case, these bracketed terms can be simplified as follows:

$$
\left(\frac{1}{1-hL}\right) + {\left(\frac{1}{1-hL}\right)}^{2} +\dots + {\left(\frac{1}{1-hL}\right)}^{N} = \frac{1}{1-hL} \left[\frac{\left(\frac{1}{1-hL}\right)^{N}-1}{\frac{1}{1-hL}-1}\right] = \frac{\left(\frac{1}{1-hL}\right)^{N}-1}{hL} \hspace{0.5cm} (16)
$$

Now substitute $(16)$ into $(15)$:

$$
\|e_{N}\| \le {\left(\frac{1}{1-hL}\right)}^{N}\|e_{0}\| + \frac{Mh^{2}}{2} \left[\frac{\left(\frac{1}{1-hL}\right)^{N}-1}{hL}\right] = {\left(\frac{1}{1-hL}\right)}^{N}\|e_{0}\| + \frac{Mh}{2L} \left[\left(\frac{1}{1-hL}\right)^{N}-1\right] \\ \Rightarrow \|e_{N}\| \le {\left(1 + \frac{hL}{1-hL}\right)}^{N}\|e_{0}\| + \frac{Mh}{2L} \left[\left(1 + \frac{hL}{1-hL}\right)^{N}-1\right] \hspace{0.5cm} (17)
$$

We rewrote the exponential terms in $(17)$ because we want to simplify this expression even further, using the lemma which we are about to prove below.

*Lemma:* $1+x \le e^{x} \hspace{0.2cm} \forall x \in \mathbb{R}$

*Proof of lemma:* Consider the function $g : \mathbb{R} \rightarrow \mathbb{R}$, $g(x)=e^{x}-x-1 \Rightarrow g'(x)=e^{x}-1$. First, we start by checking the sign of $g'(x)$ for different values of $x$. Readily, one can see that $g'(x)<0$ when $x<0$, $g'(x)=0$ when $x=0$, and $g'(x)>0$ when $x>0$. Furthermore, $g(0)=e^{0}-0-1=1-1=0$.

Now first, let us consider any $x<0$. By the mean value theorem, and using the fact that $g'(x)<0 \hspace{0.2cm} \forall x<0$, $\exists a,b \in [x,0], a<b$ such that $g'(x)=\frac{g(b)-g(a)}{b-a}<0 \Rightarrow g(b)<g(a)$. In particular, let $b=0$. Then, $g(a)>g(0)=0$, and since $x$ was arbitrary, it follows that $g(x)>0 \hspace{0.2cm} \forall x<0$. Next, consider any $x>0$. Again, by the MVT, and using the fact that $g'(x)>0 \hspace{0.2cm} \forall x>0$, $\exists a,b \in [0,x], a<b$ such that $g'(x)=\frac{g(b)-g(a)}{b-a}>0$. In particular, take $a=0$. Then, $g(b)>g(0)=0$, and since $x$ was arbitrary, it follows that $g(x)>0 \hspace{0.2cm} \forall x>0$. 

Putting these two facts together, along with how $g(0)=0$, we have that, $\forall x \in \mathbb{R}, g(x) \ge 0 \Rightarrow e^{x}-x-1 \ge 0 \Rightarrow \boxed {\therefore e^{x}\ge x+1}$.

Applying this lemma with $x=\frac{hL}{1-hL}$, we have that

$$
1+\frac{hL}{1-hL} \le e^{\frac{hL}{1-hL}}
\Rightarrow (1+\frac{hL}{1-hL})^{N} \le e^{\frac{NhL}{1-hL}}=e^{\frac{LT}{1-hL}} \hspace{0.5cm} (18)
$$

In the last step, note that we used that $T=Nh$. Finally, substituting $(18)$ into $(17)$, we get that:

$$
\boxed {\therefore \|e_{N}\| \le e^{\frac{LT}{1-hL}}\|e_{0}\| + \frac{Mh}{2L} \left[e^{\frac{LT}{1-hL}}-1\right]} 
$$

This concludes our proof of convergence for the backward Euler method, given that the first derivative of $f$ is uniformly bounded and $f$ is Lipschitz continuous for $0\le t\le T$. 

## Problem 2

Consider the initial value problem
$$
\begin{split}
u'_{1} &= u_{1},\\
u'_{2} &= u_{1} - u_{2},\\
\end{split}
$$
with initial condition $(u_{1}(0),u_{2}(0))=(1,1)$. 

1. Find a way to solve the problem above analytically, and write down the answer of u(T=2.0).

2. Solve the problem numerically using the trapezoidal rule. Verify that the solution agrees with the exact solution at time $T=2.0$. Use a number of step sizes $h=2^{-i},i=5,\ldots,10.$ Plot the error $e=\Vert u(1)-u^*(1)\Vert_2$ v.s. step size $h$ in a loglog plot, and estimate the convergence rate for each scheme.

**Note**: the routine for computing the error, plotting and computing the slope is provided. 

**Part 1** 

First, we write this IVP as a two-dimensional system of differential equations:

$$
\begin{pmatrix} \dot{u_{1}}(t) \\ \dot{u_{2}}(t) \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} u_{1}(t) \\ u_{2}(t) \end{pmatrix}
$$

We define $A:=\begin{pmatrix} 1 & 0 \\ 1 & -1 \end{pmatrix}$ and $\overrightarrow{u(t)}:=\begin{pmatrix} u_{1}(t) \\ u_{2}(t) \end{pmatrix} \Rightarrow \dot{\overrightarrow{u(t)}}=\begin{pmatrix} \dot{u_{1}}(t) \\ \dot{u_{2}}(t) \end{pmatrix}$. From ODE theory, we know that each solution to this system, given by the phase space defined with the fundamental matrix for this system, is of the form $Ce^{\lambda t}$, with $C$ a constant determined by initial conditions and $\lambda$ an eigenvalue for $A$. Thus, we start by calculating the eigenvalues of $A$. 

$$
A-\lambda I = \begin{pmatrix} 1-\lambda & 0 \\ 1 & -1-\lambda \end{pmatrix} \Rightarrow det(A-\lambda I) = (1-\lambda)(-1-\lambda) = 0 \\ \Rightarrow \lambda = 1, -1
$$

Next, we find the eigenspace corresponding to each of these eigenvalues.

$\underline{\lambda_{1} = 1}$
$$
A-\lambda_{1} I = \begin{pmatrix} 1-1 & 0 \\ 1 & -1-1 \end{pmatrix} = \begin{pmatrix} 0 & 0 \\ 1 & -2 \end{pmatrix} \Rightarrow Nul(A-I) = Span\left\{ \begin{pmatrix} 2 \\ 1 \end{pmatrix} \right\} 
$$

$\underline{\lambda_{2} = -1}$
$$
A-\lambda_{2} I = \begin{pmatrix} 1+1 & 0 \\ 1 & -1+1 \end{pmatrix} = \begin{pmatrix} 2 & 0 \\ 1 & 0 \end{pmatrix} \Rightarrow Nul(A+I) = Span\left\{ \begin{pmatrix} 0 \\ 1 \end{pmatrix} \right\} 
$$

Thus, we can write the general solution to this system of ODEs, where $c_{1}$ and $c_{2}$ are arbitrary constants, to be determined by initial conditions:

$$
\overrightarrow{u(t)} = c_{1} e^{\lambda_{1} t} \begin{pmatrix} 2 \\ 1 \end{pmatrix} + c_{2} e^{\lambda_{2} t} \begin{pmatrix} 0 \\ 1 \end{pmatrix} = c_{1} e^{t} \begin{pmatrix} 2 \\ 1 \end{pmatrix} + c_{2} e^{-t} \begin{pmatrix} 0 \\ 1 \end{pmatrix}
$$

We determine $c_{1}$ and $c_{2}$:

$$
\begin{pmatrix} u_{1}(0) \\ u_{2}(0) \end{pmatrix} = \begin{pmatrix} 1 \\ 1 \end{pmatrix} = \begin{pmatrix} 2c_{1} \\ c_{1}+c_{2} \end{pmatrix} \Rightarrow c_{1} = \frac{1}{2} \Rightarrow c_{2} = \frac{1}{2}
$$

Therefore, given the initial conditions, the exact solution to this IVP, using the identity $\frac{e^{t}+e^{-t}}{2}=cosh(t)$, is:

$$
\overrightarrow{u(t)} = e^{t} \begin{pmatrix} 1 \\ \frac{1}{2} \end{pmatrix} + e^{-t} \begin{pmatrix} 0 \\ \frac{1}{2} \end{pmatrix} \\ \Rightarrow \boxed {\therefore \overrightarrow{u(t)} = \begin{pmatrix} e^{t} \\ cosh(t) \end{pmatrix}}
$$

Now, the problem asks for us to write the answer at $t=2.0$, which is:

$$
\boxed {\therefore \overrightarrow{u(2)} = \begin{pmatrix} e^{2} \\ cosh(2) \end{pmatrix} \approx \begin{pmatrix} 7.389056 \dots \\ 3.762196 \dots \end{pmatrix}}
$$

**Part 2** 

First, we write the equation for trapezoidal rule, generally:

$$
\overrightarrow{u_{n+1}} = \overrightarrow{u_{n}} + \frac{h}{2} \left( \overrightarrow{f}\left(t_{n}, \overrightarrow{u_{n}}\right) + \overrightarrow{f}\left(t_{n+1}, \overrightarrow{u_{n+1}}\right)\right) 
$$

In particular, note that $\overrightarrow{f}\left(t_{n}, \overrightarrow{u_{n}}\right) = A \overrightarrow{u_{n}}$, where $A$ is defined in part 1. Then, substituting this into the above, we get that

$$
\overrightarrow{u_{n+1}} = \overrightarrow{u_{n}} + \frac{h}{2} \left( A \overrightarrow{u_{n}} + A \overrightarrow{u_{n+1}}\right) \hspace{0.5cm} \\ \Rightarrow \overrightarrow{u_{n+1}} - \frac{h}{2} A \overrightarrow{u_{n+1}} = \overrightarrow{u_{n}} + \frac{h}{2} A \overrightarrow{u_{n}} \\ \Rightarrow \left(I - \frac{h}{2} A\right) \overrightarrow{u_{n+1}} = \left(I + \frac{h}{2} A\right) \overrightarrow{u_{n}} 
$$

Assuming that $I - \frac{h}{2} A$ is invertible (and it will be for appropriate values of h), our equation for $u_{n+1}$ becomes:

$$
\overrightarrow{u_{n+1}} = \left(I-\frac{h}{2}A\right)^{-1}\left(I+\frac{h}{2}A\right)\overrightarrow{u_{n}}
$$

Skipping some tedious computational algebra, after simplifying, $\left(I-\frac{h}{2}A\right)^{-1}\left(I+\frac{h}{2}A\right)=\begin{pmatrix} \frac{2+h}{2-h} & 0 \\ \frac{4h}{4-h^{2}} & \frac{2-h}{2+h} \end{pmatrix}$. Thus, we get our final computational scheme:

$$
\boxed {\therefore \overrightarrow{u_{n+1}} = \begin{pmatrix} \frac{2+h}{2-h} & 0 \\ \frac{4h}{4-h^{2}} & \frac{2-h}{2+h} \end{pmatrix} \overrightarrow{u_{n}}}
$$

Using this scheme, we proceed to use it in our numerical solution to this ODE system. 

```julia
using LinearAlgebra
using PyPlot

function Trapezoid(N,T,u0)
    ui = copy(u0)
    uf = copy(u0)
    h = T / N
    M = [(2 + h) / (2 - h) 0; 4 * h / (4 - h ^ 2)  (2 - h) / (2 + h)]
    for i = 1 : N
        uf = M * ui
        ui = uf
    end
    return ui
end

u0 = [1.0,1.0]
T = 2.0
# You need to compute uexact at T=2.0
uexact = [exp(T); cosh(T)]


# Plotting routine provided below
NList = 2 .^(5:10)
errTrapeList = zeros(size(NList))
for i = 1 : length(NList)
    N = NList[i]
    utrape = Trapezoid(N,T,u0)
    errTrapeList[i] = norm(utrape-uexact)
end

figure(1,figsize=(4,4))
loglog(T./NList, errTrapeList, "r-d", label="Trapezoid")
xlabel(L"h")
ylabel(L"||u-u^*||_2")
legend()

slopeTrape = -(log(errTrapeList[end])-log(errTrapeList[1])) / 
(log(NList[end]) - log(NList[1]))

println("Slope for Trapezoid = ", slopeTrape )
```

Since the absolute error here is rather small between the numerical solution and the exact solution (something on order $10^{-2}$ to $10^{-5}$), we can say with good certainty that our numerical scheme works for finding $\overrightarrow{u(2)}$. Qualitatively, we also see that it indeed converges to the exact solution as $h \rightarrow 0$. Furthermore, based off of the slope in the loglog plot above between stepsize and absolute error, we can say that the convergence rate for the trapezoidal rule is roughly quadratic (i.e. with $p \approx 2$ for an error bound of $Kh^{p}$, $K$ a constant), which we would expect from the theory.

## Problem 3

Consider a simple pendulum of length $l$, forming an angle $\theta(t)$ with the
vertical axis. The equation of motion for the pendulum is given by

$$\theta''(t)+\frac{g}{l} \sin(\theta(t))=0,$$

where $g$ is the gravitational constant, with given initial conditions

$$\theta(0)=\theta_0,\quad \theta'(0)=\omega_0.$$

1. Rewrite the equation as a system of first order equations in $u(t)=(\theta(t),\omega(t))$. Fill the `func` function below. 

2. Write a forward Euler method. Use a sufficiently small time step, and plot the solution of $\theta(t)$ with $g=l=1$ and initial condition $(\theta(0),\omega(0))=(2.5,0.0)$ till $T=25$. Since no exact solution is available, measure the convergence rate by means of the solution at T=25 obtained from three time steps h, h/2, and h/4.

**Part 1**

Converting this equation of motion into a two-dimensional system of first-order equations is pretty straightforward. We define the angular velocity as $\omega(t) := \dot{\theta}(t)$, so that the equation of motion becomes $\dot{\omega}(t) = -\frac{g}{l}sin(\theta(t))$. Thus, our two-dimensional system is:

$$
\dot{\overrightarrow{u(t)}}=\begin{pmatrix} \dot{\theta}(t) \\ \ \dot{\omega}(t) \end{pmatrix}=\begin{pmatrix} \omega(t) \\ -\frac{g}{l}sin(\theta(t)) \end{pmatrix}
$$

```julia
function func(u, g, l)
    return [u[2], -(g / l) * sin(u[1])]
end
```

**Part 2**

We start by writing out a general forward Euler method, which looks like the following:

$$
\overrightarrow{u_{n+1}} = \overrightarrow{u_{n}} + h\overrightarrow{f}\left(\overrightarrow{u_{n}}\right)
$$

In our case, $\overrightarrow{u_{n+1}}:=\begin{pmatrix} \theta_{n+1} \\ \omega_{n+1} \end{pmatrix}$, $\overrightarrow{u_{n}}:=\begin{pmatrix} \theta_{n} \\ \omega_{n} \end{pmatrix}$, and $\overrightarrow{f}\left(\overrightarrow{u_{n}}\right):=\begin{pmatrix} \omega_{n} \\ -\frac{g}{l}sin(\theta_{n}) \end{pmatrix}$. Therefore, the forward Euler method applied to this IVP is:

$$
\boxed {\therefore \overrightarrow{u_{n+1}}=\begin{pmatrix} \theta_{n+1} \\ \omega_{n+1} \end{pmatrix}=\begin{pmatrix} \theta_{n} + h\omega_{n} \\ \omega_{n} - \frac{gh}{l}sin(\theta_{n}) \end{pmatrix} =\begin{pmatrix} \theta_{n} \\ \omega_{n} \end{pmatrix} + h \begin{pmatrix} \omega_{n} \\ - \frac{g}{l}sin(\theta_{n}) \end{pmatrix}}
$$

This is the iterative scheme which we will use below.

```julia
function ForwardEulerFull(N, T, u0, g, l)
    # Fill code here. 
    # The output should be the full history of the solution
    # i.e. the output u should be of size 2(N+1)
    # This is of course memory intensive and is not the 
    # commonly used approach
    u = zeros(2,N+1)
    u[:, 1] = u0
    h = T / N
    for i = 1 : N
        u[:, i + 1] = u[:, i] + h * func(u[:, i], g, l)
    end
    return u
end

# This should plot out the solution
g = 1.0
l = 1.0
T = 25.0
N = 100000
u0 = [2.5, 0.0]

uEulerFull = ForwardEulerFull(N, T, u0, g, l)
tList = collect(0 : N) * (T / N)

using PyPlot
figure(1, figsize = (4 , 4))
plot(tList, vec(uEulerFull[1, :]), "b-")
xlabel(L"t")
ylabel(L"\theta(t)")

function ForwardEuler(N, T, u0, g, l)
    # Fill code here. 
    # The output contains only the final solution u(T)
    u = copy(u0)
    h = T / N
    for i = 1 : N
        u = u + h * func(u, g, l)
    end
    return u
end

Eulerlarge = ForwardEuler(N, T, u0, g, l)
Eulermedium = ForwardEuler(2 * N, T, u0, g, l)
Eulersmall = ForwardEuler(4 * N, T, u0, g, l)

# Estimate the order of convergence
slope = log(norm(Eulerlarge - Eulermedium) / norm(Eulermedium - Eulersmall)) / log(2)
println("Slope for Forward Euler = ", slope)
```

Therefore, since $p \approx 1.019614 \approx 1$ in our approximation for $p$, we have roughly linear convergence for the forward Euler method applied to this IVP (i.e. with $p \approx 1$ for an error bound of $Kh^{p}$, $K$ a constant). This is what we would expect from the theory, which states that the forward Euler method tends to exhibit linear convergence globally.

## Problem 4

1. Repeat Problem 3.2 using the trapezoidal rule. Modify your implementation in Problem 2 if needed.

2. Repeat Problem 3.2 using the Adams-Bashforth 3 (AB3) scheme. Obtain the first two steps using forward Euler and trapezoidal rule, respectively. Observe the difference in terms of the convergence rate.



**Part 1**

As before, we write the equation for trapezoidal rule, generally:

$$
\overrightarrow{u_{n+1}} = \overrightarrow{u_{n}} + \frac{h}{2} \left( \overrightarrow{f}\left(\overrightarrow{u_{n}}\right) + \overrightarrow{f}\left(\overrightarrow{u_{n+1}}\right)\right) 
$$

We remind that, in our case, $\overrightarrow{u_{n+1}}=\begin{pmatrix} \theta_{n+1} \\ \omega_{n+1} \end{pmatrix}$, $\overrightarrow{u_{n}}=\begin{pmatrix} \theta_{n} \\ \omega_{n} \end{pmatrix}$, and $\overrightarrow{f}\left(\overrightarrow{u_{n}}\right)=\begin{pmatrix} \omega_{n} \\ -\frac{g}{l}sin(\theta_{n}) \end{pmatrix}$. Then, substituting this into the above, we get that

$$
\begin{pmatrix} \theta_{n+1} \\ \omega_{n+1} \end{pmatrix} = \begin{pmatrix} \theta_{n} \\ \omega_{n} \end{pmatrix} + \frac{h}{2} \left( \begin{pmatrix} \omega_{n} \\ -\frac{g}{l}sin(\theta_{n}) \end{pmatrix} + \begin{pmatrix} \omega_{n+1} \\ -\frac{g}{l}sin(\theta_{n+1}) \end{pmatrix}\right)
$$

As one can see, the issue here is that our equation for $\overrightarrow{u_{n+1}}$ is nonlinear, so we cannot explicitly solve for it. Thus, we implement the following iterative nonlinear solver to approximate $\overrightarrow{u_{n+1}}$ for each timestep, halting the iterations in $i$ once the absolute difference between successive iterations falls below a tolerance of $10^{-12}$:

$$
\begin{pmatrix} \theta_{n+1}^{i+1} \\ \omega_{n+1}^{i+1} \end{pmatrix} = \begin{pmatrix} \theta_{n} \\ \omega_{n} \end{pmatrix} + \frac{h}{2} \left( \begin{pmatrix} \omega_{n} \\ -\frac{g}{l}sin(\theta_{n}) \end{pmatrix} + \begin{pmatrix} \omega_{n+1}^{i} \\ -\frac{g}{l}sin(\theta_{n+1}^{i}) \end{pmatrix}\right) \\ \Rightarrow \overrightarrow{u_{n+1}^{i+1}} = \overrightarrow{u_{n}} + \frac{h}{2} \left( \overrightarrow{f}\left(\overrightarrow{u_{n}}\right) + \overrightarrow{f}\left(\overrightarrow{u_{n+1}^{i}}\right)\right) 
$$

To make an initial guess for our iteration at each timestep, we will let $\overrightarrow{u_{n+1}^{i=0}}=\overrightarrow{u_{n}}$. We expect that this iterative method should converge at each step, provided that our timestep, $h$, is small enough.

```julia
# Need to solve the nonlinear equation in trapezoidal rule.
function TrapezoidFull(N, T, u0, g, l)
    # Fill code here. 
    # The output should be the full history of the solution
    # i.e. the output u should be of size 2(N+1)
    # This is of course memory intensive and is not the 
    # commonly used approach
    u = zeros(2, N + 1)
    u[:, 1] = u0
    ui = copy(u0)
    um = copy(u0)
    h = T / N
    tol = 1e-12
    maxiterations = 5000
    for i = 1 : N + 1
        fni = func(ui, g, l)
        for iter = 1 : maxiterations
            fnf = func(um, g, l)
            um = ui + (h / 2) * (fni + fnf)
            if (norm(ui - um) < tol)
                break
            end
            u[:, i] = um
        end
        ui = um
    end
    return u
end

# This should plot out the solution
g = 1.0
l = 1.0
T = 25.0
N = 2000
u0 = [2.5, 0.0]

uTrapFull = TrapezoidFull(N, T, u0, g, l)
tList = collect(0 : N) * (T / N)

using PyPlot
figure(1, figsize = (4 , 4))
plot(tList, vec(uTrapFull[1, :]), "b-")
xlabel(L"t")
ylabel(L"\theta(t)")

function Trapezoid(N, T, u0, g, l)
    # Fill code here. 
    # The output contains only the final solution u(T)
    ui = copy(u0)
    um = copy(u0)
    u = copy(u0)
    h = T / N
    tol = 1e-12
    maxiterations = 5000
    for i = 1 : N 
        fni = func(ui, g, l)
        for iter = 1 : maxiterations
            fnf = func(um, g, l)
            um = ui + (h / 2) * (fni + fnf)
            if (norm(ui - um) < tol)
                break
            end
            u = um
        end
        ui = um
    end
    return u
end

Traplarge = Trapezoid(N, T, u0, g, l)
Trapmedium = Trapezoid(2 * N, T, u0, g, l)
Trapsmall = Trapezoid(4 * N, T, u0, g, l)

# Estimate the order of convergence
slope = log(norm(Traplarge - Trapmedium) / norm(Trapmedium - Trapsmall)) / log(2)
println("Slope for trapezoidal rule = ", slope)
```

Therefore, since $p \approx 2.000045 \approx 2$ in our approximation for $p$, we have roughly quadratic convergence for the trapezoidal rule applied to this IVP (i.e. with $p \approx 2$ for an error bound of $Kh^{p}$, $K$ a constant). This is what we would expect from the theory, which states that the trapezoidal rule tends to exhibit quadratic convergence globally, which is also what we saw in problem 2.

**Part 2**

We start by writing out the general form of the 3rd-order Adams-Bashforth (AB3) scheme:

$$
\overrightarrow{u_{n+3}} = \overrightarrow{u_{n+2}} + h \left(\frac{23}{12} \overrightarrow{f}\left(\overrightarrow{u_{n+2}}\right) - \frac{4}{3} \overrightarrow{f}\left(\overrightarrow{u_{n+1}}\right) + \frac{5}{12} \overrightarrow{f}\left(\overrightarrow{u_{n}}\right)\right)
$$

In our case, $\overrightarrow{u_{n+3}}:=\begin{pmatrix} \theta_{n+3} \\ \omega_{n+3} \end{pmatrix}$, $\dots$ ,$\overrightarrow{u_{n}}:=\begin{pmatrix} \theta_{n} \\ \omega_{n} \end{pmatrix}$, $\overrightarrow{f}\left(\overrightarrow{u_{n+3}}\right):=\begin{pmatrix} \omega_{n+3} \\ -\frac{g}{l}sin(\theta_{n+3}) \end{pmatrix}$, $\dots$, and $\overrightarrow{f}\left(\overrightarrow{u_{n}}\right):=\begin{pmatrix} \omega_{n} \\ -\frac{g}{l}sin(\theta_{n}) \end{pmatrix}$. Since we do not have explicit values for $\overrightarrow{u_{1}}$ and $\overrightarrow{u_{2}}$, we will at first use the forward Euler method (i.e. AB1) to find $\overrightarrow{u_{1}}$ and $\overrightarrow{u_{2}}$, then continue from there on out. Then, we will do the same with the trapezoidal rule (i.e. AM2). 

```julia
# Need some method to compute u1,u2
function AB3Full(N, T, u0, u1, u2, g, l)
    # Fill code here. 
    # The output contains only the final solution u(T)
    u = zeros(2,N+1)
    u[:, 1] = copy(u0)
    u[:, 2] = copy(u1)
    u[:, 3] = copy(u2)
    fu0 = func(u[:, 1], g, l)
    fu1 = func(u[:, 2], g, l)
    fu2 = func(u[:, 3], g, l) 
    h = T / N
    if N < 3
        print("Congratulations! You do not have enough timesteps to use AB3. Try again.")
    else
        for i = 3 : N
            u[:, i + 1] = u[:, i] + h * ((23 / 12) * fu2 - (4 / 3) * fu1 + (5 / 12) * fu0)
            fu0 = fu1
            fu1 = fu2
            fu2 = func(u[:, i + 1], g, l)
        end
        return u
    end
end

# This should plot out the solution
g = 1.0
l = 1.0
T = 25.0
N = 100000
u0 = [2.5, 0.0]
u1 = ForwardEuler(1, T / N, u0, g, l)
u2 = ForwardEuler(1, T / N, u1, g, l)
uAB3Full = AB3Full(N, T, u0, u1, u2, g, l)
tList = collect(0 : N) * (T / N)

using PyPlot
figure(1, figsize = (4 , 4))
plot(tList, vec(uAB3Full[1, :]), "b-")
xlabel(L"t")
ylabel(L"\theta(t)")

function AB3(N, T, u0, u1, u2, g, l)
    # Fill code here. 
    # The output contains only the final solution u(T)
    ut0 = copy(u0)
    ut1 = copy(u1)
    ut2 = copy(u2)
    u = 0
    fu0 = func(ut0, g, l)
    fu1 = func(ut1, g, l)
    fu2 = func(ut2, g, l) 
    h = T / N
    if N < 3
        print("Congratulations! You do not have enough timesteps to use AB3. Try again.")
    else
        for i = 3 : N
            u = ut2 + h * ((23 / 12) * fu2 - (4 / 3) * fu1 + (5 / 12) * fu0)
            ut0 = ut1
            ut1 = ut2
            ut2 = u
            fu0 = fu1
            fu1 = fu2
            fu2 = func(u, g, l)
        end
        return u
    end
end

AB3large = AB3(N, T, u0, u1, u2, g, l)

u12N = ForwardEuler(1, T / (2 * N), u0, g, l)
u22N = ForwardEuler(1, T / (2 * N), u12N, g, l)
AB3medium = AB3(2 * N, T, u0, u12N, u22N, g, l)

u14N = ForwardEuler(1, T / (4 * N), u0, g, l)
u24N = ForwardEuler(1, T / (4 * N), u14N, g, l)
AB3small = AB3(4 * N, T, u0, u14N, u24N, g, l)

# Estimate the order of convergence
slope = log(norm(AB3large - AB3medium) / norm(AB3medium - AB3small)) / log(2)
println("Slope for AB3 using an AB1 predictor = ", slope)
```

Therefore, since $p \approx 1.996897 \approx 2$ in our approximation for $p$, we have roughly quadratic convergence for the 3rd-order Adams-Bashforth scheme applied to this IVP (i.e. with $p \approx 2$ for an error bound of $Kh^{p}$, $K$ a constant) when using a forward Euler predictor. Now, if our values for $\overrightarrow{u_{1}}$ and $\overrightarrow{u_{2}}$ had been exact, we would have expected cubic convergence (i.e. $p=3$). However, the theory says that, since we used a predictor which has a 2nd-order truncation error (i.e. $O(h^{2})$), our convergence will drop down from cubic to quadratic (i.e. from $p=3$ to $p=2$). This is exactly what we observe, so this result is consistent with the theory.

```julia
# This should plot out the solution
g = 1.0
l = 1.0
T = 25.0
N = 100000
u0 = [2.5, 0.0]
u1 = Trapezoid(1, T / N, u0, g, l)
u2 = Trapezoid(1, T / N, u1, g, l)
uAB3Full = AB3Full(N, T, u0, u1, u2, g, l)
tList = collect(0 : N) * (T / N)

using PyPlot
figure(1, figsize = (4 , 4))
plot(tList, vec(uAB3Full[1, :]), "b-")
xlabel(L"t")
ylabel(L"\theta(t)")

AB3large = AB3(N, T, u0, u1, u2, g, l)

u12N = Trapezoid(1, T / (2 * N), u0, g, l)
u22N = Trapezoid(1, T / (2 * N), u12N, g, l)
AB3medium = AB3(2 * N, T, u0, u12N, u22N, g, l)

u14N = Trapezoid(1, T / (4 * N), u0, g, l)
u24N = Trapezoid(1, T / (4 * N), u14N, g, l)
AB3small = AB3(4 * N, T, u0, u14N, u24N, g, l)

# Estimate the order of convergence
slope = log(norm(AB3large - AB3medium) / norm(AB3medium - AB3small)) / log(2)
println("Slope for AB3 using an AM2 predictor = ", slope)
```

Therefore, since $p \approx 2.996506 \approx 3$ in our approximation for $p$, we have roughly cubic convergence for the 3rd-order Adams-Bashforth scheme applied to this IVP (i.e. with $p \approx 3$ for an error bound of $Kh^{p}$, $K$ a constant) when using a trapezoidal rule predictor. This is exactly what the theory predicts, even though our values for $\overrightarrow{u_{1}}$ and $\overrightarrow{u_{2}}$ were not exact. This is because the trapezoidal rule has a local truncation error (LTE) which is third-order as well (i.e. $O(h^{3})$), so the convergence overall for AB3 will still remain third-order, unlike above, where the convergence order dropped as a result of us using a less precise predicting scheme.
