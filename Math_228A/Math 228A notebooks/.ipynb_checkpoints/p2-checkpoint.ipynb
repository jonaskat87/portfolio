## Problem 1

Derive explicitly the four-step Adams-Moulton method.

Any arbitrary linear multistep method (LMM) can be written in the following form, where each $y_{i} \in \mathbb{R}^{d}$ and $f \in \mathbb{R}^{d}$:

$$
y_{n+r}=\sum_{k=1}^{r} a_{r-k}\hspace{0.1cm} y_{n+r-k}+h \sum_{k=0}^{r} b_{r-k}\hspace{0.1cm} f\left(t_{n+r-k},y_{n+r-k}\right) \hspace{0.5cm} (1) 
$$

For simplicity's sake, we dropped all arrows above our variables. For the family of Adams–Moulton (AM) methods, we enforce that $a_{r-1}=1$ and $a_{r-2}=\dots=a_{0}=0$, which gives us the following from $(1)$:

$$
y_{n+r}=y_{n+r-1}+h \sum_{k=0}^{r} b_{r-k}\hspace{0.1cm} f\left(t_{n+r-k},y_{n+r-k}\right) \hspace{0.5cm} (2) 
$$

Now, each coefficient $b_{r-k}$ is chosen such that this scheme has the highest order of convergence as possible. Thus, using $r$ prior points and the point in question at $y_{n+r}$, we use a $k$th order interpolating polynomial, $p_{k}$, to interpolate values of $f$ for each consecutive set of $r+1$ points. This means that:

$$
b_{r-k} = \frac{1}{t_{n+r+1}-t_{n+r}} \int_{t_{n+r}}^{t_{n+r+1}}p_{k}(s)ds:=\frac{1}{h} \int_{t_{n+r}}^{t_{n+r+1}}p_{k}(s)ds \hspace{0.5cm} (3) 
$$

We could choose any variety of $p_{k}$ to interpolate, but for AM methods, we use a $k$th-order Lagrange polynomial. Thus, $p_{k}(s)=\prod_{\substack{j=0 \\ j \neq k}}^{r} \left(\frac{s-s_{n+j}}{s_{n+k}-s_{n+j}}\right)$, so $(3)$ becomes:

$$
b_{r-k}=\frac{1}{h} \int_{t_{n+r}}^{t_{n+r+1}}\prod_{\substack{j=0 \\ j \neq k}}^{r} \left(\frac{s-s_{n+j}}{s_{n+k}-s_{n+j}}\right)ds \hspace{0.5cm} \\ \begin{align} s=t_{n+r}+\theta h \\ ds=hd\theta \end{align} \Rightarrow b_{r-k}=\int_{0}^{1}\prod_{\substack{j=0 \\ j \neq k}}^{r} \left(\frac{j-1+\theta}{k-j}\right)d\theta \hspace{0.5cm} (4) 
$$

Finally, after making some combinatorial simplifications, $(4)$ yields us the following:

$$
b_{r-k}=\frac{(-1)^{k}}{k!(r-k)!}\int_{0}^{1}\prod_{\substack{j=0 \\ j \neq k}}^{r} \left(j-1+\theta\right)d\theta \hspace{0.5cm} (5) 
$$

In particular, for this problem, we want to derive the four-step AM method (i.e. all of this for $r=4$). Thus, our scheme becomes the following two equations, from $(2)$ and $(5)$ respectively:

$$
y_{n+4}=y_{n+3}+h \sum_{k=0}^{4} b_{4-k}\hspace{0.1cm} f\left(t_{n+4-k},y_{n+4-k}\right) \hspace{0.5cm} (6) \\ 
b_{4-k}=\frac{(-1)^{k}}{k!(4-k)!}\int_{0}^{1}\prod_{\substack{j=0 \\ j \neq k}}^{4} \left(j-1+\theta\right)d\theta \hspace{0.5cm} (7) 
$$

Since $0\le k\le 4$, we use $(7)$ to find all $b_{i}$ required to get our scheme.

**k=0:**
$$
b_{4}=\frac{1}{0!4!}\int_{0}^{1}(\theta)(1+\theta)(2+\theta)(3+\theta)d\theta=\left. \frac{1}{24}\left(\frac{\theta^{5}}{5}+\frac{3\theta^{4}}{2}+\frac{11\theta^{3}}{3}+3\theta^{2}\right) \right|_{0}^{1}=\frac{251}{720} \hspace{0.5cm} (A) 
$$
**k=1:**
$$
b_{3}=\frac{-1}{1!3!}\int_{0}^{1}(\theta-1)(1+\theta)(2+\theta)(3+\theta)d\theta=\left. -\frac{1}{6}\left(\frac{\theta^{5}}{5}+\frac{5\theta^{4}}{4}+\frac{5\theta^{3}}{3}-\frac{5\theta^{2}}{2}-6\theta\right) \right|_{0}^{1}=\frac{323}{360} \hspace{0.5cm} (B)
$$
**k=2:**
$$
b_{2}=\frac{1}{2!2!}\int_{0}^{1}(\theta-1)(\theta)(2+\theta)(3+\theta)d\theta=\left. \frac{1}{4}\left(\frac{\theta^{5}}{5}+\theta^{4}+\frac{\theta^{3}}{3}-3\theta^{2}\right) \right|_{0}^{1}=-\frac{11}{30} \hspace{0.5cm} (C)
$$
**k=3:**
$$
b_{1}=\frac{-1}{3!1!}\int_{0}^{1}(\theta-1)(\theta)(1+\theta)(3+\theta)d\theta=\left. -\frac{1}{6}\left(\frac{\theta^{5}}{5}+\frac{3\theta^{4}}{4}-\frac{\theta^{3}}{3}-\frac{3\theta^{2}}{2}\right) \right|_{0}^{1}=\frac{53}{360} \hspace{0.5cm} (D)
$$
**k=4:**
$$
b_{0}=\frac{1}{4!0!}\int_{0}^{1}(\theta-1)(\theta)(1+\theta)(2+\theta)d\theta=\left. \frac{1}{24}\left(\frac{\theta^{5}}{5}+\frac{\theta^{4}}{2}-\frac{\theta^{3}}{3}-\theta^{2}\right) \right|_{0}^{1}=-\frac{19}{720} \hspace{0.5cm} (E)
$$

Finally, plugging $(A)$ to $(E)$ into $(6)$, we get:

$$
y_{n+4}=y_{n+3}+h \left(b_{4}\hspace{0.1cm} f\left(t_{n+4},y_{n+4}\right)+b_{3}\hspace{0.1cm} f\left(t_{n+3},y_{n+3}\right)+b_{2}\hspace{0.1cm} f\left(t_{n+2},y_{n+2}\right)+b_{1}\hspace{0.1cm} f\left(t_{n+1},y_{n+1}\right)+b_{0}\hspace{0.1cm} f\left(t_{n},y_{n}\right)\right) \\ \Rightarrow \boxed {\therefore y_{n+4}=y_{n+3}+h\left(\frac{251}{720}\hspace{0.1cm}f\left(t_{n+4},y_{n+4}\right) + \frac{323}{360}\hspace{0.1cm}f\left(t_{n+3},y_{n+3}\right)-\frac{11}{30}\hspace{0.1cm}f\left(t_{n+2},y_{n+2}\right)+\frac{53}{360}\hspace{0.1cm}f\left(t_{n+1},y_{n+1}\right)-\frac{19}{720}\hspace{0.1cm}f\left(t_{n},y_{n}\right)\right)}
$$

This is the four-step Adams-Moulton method, as desired.

## Problem 2

The $r$-step Nystrom method is given as follows: Replace $f(u(t),t)$ in the exact formulation below

$$u(t_{n+r})-u(t_{n+r-2}) = \int_{t_{n+r-2}}^{t_{n+r}} f(u(t),t) dt$$

by a Lagrange polynomial interpolating at $f_{n},\ldots,f_{n+r}$, where $f_{n+s}\equiv f(u_{n+s},t_{n+s})$. The explicit scheme similar to Adams-Bashforth method is called the Nystrom method.  The implicit scheme similar to Adams-Moulton method is called the Milne method.

a) Derive the Nystrom method with $r=1$ and $r=2$. Prove that both result in the same method, and prove that this is a second order method.

b) Derive the Milne method with $r=2$, and prove that this is a fourth order method.

**NOTE**: For this problem you only need to show the LTE is of the correct order.

**Part a:**

For $r=1$ with the Nyström method, we have:
$$
u(t_{n+1})-u(t_{n-1})=\int_{t_{n-1}}^{t_{n+1}}f(u(t),t)dt \approx \int_{t_{n-1}}^{t_{n+1}}P(t)dt \hspace{0.5cm} (1)
$$
$P(t)$ is the Lagrange polynomial that just interpolates $(t_{n},f_{n})$, so in this case, $P(t)=f_{n}p_{n}=f_{n}$, because $p_{n}=1$. Thus,
$$
u_{n+1}-u_{n-1}=\int_{t_{n-1}}^{t_{n+1}}f_{n}dt=f_{n}\int_{t_{n-1}}^{t_{n+1}}dt=f_{n}(t_{n+1}-t_{n-1}):=2hf_{n} \\ \Rightarrow \boxed{u_{n+1}=u_{n-1}+2hf_{n}} \hspace{0.5cm} (2)
$$
On the other hand, for $r=2$, we now have:
$$
u(t_{n+2})-u(t_{n})=\int_{t_{n}}^{t_{n+2}}f(u(t),t)dt \approx \int_{t_{n}}^{t_{n+2}}P(t)dt \hspace{0.5cm} (3)
$$
$P(t)$ interpolates $(t_{n},f_{n})$ and $(t_{n+1},f_{n+1})$, so by construction, $P(t)=f_{n}p_{n}(t)+f_{n+1}p_{n+1}(t)=f_{n}\left(\frac{t-t_{n+1}}{t_{n}-t_{n+1}}\right)+f_{n+1}\left(\frac{t-t_{n}}{t_{n+1}-t_{n}}\right)$. Defining $h:=t_{n+1}-t_{n}:=t_{n}-t_{n-1}$ and substituting this all into $(3)$, we get:
$$
u_{n+2}-u_{n}=\frac{1}{h}\int_{t_{n}}^{t_{n+2}}\left(f_{n+1}(t-t_{n})-f_{n}(t-t_{n+1})\right)dt \\ \begin{align} t=t_{n}+2h\theta \\ dt=2hd\theta \end{align} \Rightarrow u_{n+2}-u_{n}=2h\left(\int_{0}^{1}\left(2\theta f_{n+1}-f_{n}(2\theta-1)\right)d\theta\right)=2h(f_{n+1}-f_{n}+f_{n}) \\ \Rightarrow \boxed{u_{n+2}=u_{n}+2hf_{n+1}} \hspace{0.5cm} (4)
$$
$\therefore$ Upon inspection, we see that $(2)$ and $(4)$ are identical if we shift the indices in either method by 1.

To prove that this method is second-order, we calculate the local truncation error (LTE) from $(2)$, $e(u_{n})$, noting that $f_{n}\equiv u^{'}_n$.

$$
e(u_{n})=u_{n+1}-u_{n-1}-2hf_{n}\equiv u_{n+1}-u_{n-1}-2hu^{'}_{n} \hspace{0.5cm} (5)
$$

We now take Taylor series of $u_{n+1}$ and $u_{n-1}$, each about $u_{n}$.

$$
u_{n+1}=u_{n}+u^{'}_{n}(t_{n+1}-t_{n})+\frac{u^{''}_{n}}{2}(t_{n+1}-t_{n})^{2}+\frac{u^{'''}_{n}}{6}(t_{n+1}-t_{n})^{3}+O\left((t_{n+1}-t_{n})^{4}\right)=u_{n}+hu^{'}_{n}+\frac{h^{2}u^{''}_{n}}{2}+\frac{h^{3}u^{'''}_{n}}{6}+O\left(h^{4}\right) \hspace{0.5cm} (6) \\
u_{n-1}=u_{n}+u^{'}_{n}(t_{n-1}-t_{n})+\frac{u^{''}_{n}}{2}(t_{n-1}-t_{n})^{2}+\frac{u^{'''}_{n}}{6}(t_{n-1}-t_{n})^{3}+O\left((t_{n-1}-t_{n})^{4}\right)=u_{n}-hu^{'}_{n}+\frac{h^{2}u^{''}_{n}}{2}-\frac{h^{3}u^{'''}_{n}}{6}+O\left(h^{4}\right) \hspace{0.5cm} (7)
$$
Plugging $(6)$ and $(7)$ into $(5)$, we find that:

$$
e(u_{n})=\left(u_{n}+hu^{'}_{n}+\frac{h^{2}u^{''}_{n}}{2}+\frac{h^{3}u^{'''}_{n}}{6}+O\left(h^{4}\right)\right)-\left(u_{n}-hu^{'}_{n}+\frac{h^{2}u^{''}_{n}}{2}-\frac{h^{3}u^{'''}_{n}}{6}+O\left(h^{4}\right)\right)-2hu^{'}_{n} \\ e(u_{n})=u_{n}+hu^{'}_{n}+\frac{h^{2}u^{''}_{n}}{2}+\frac{h^{3}u^{'''}_{n}}{6}-u_{n}+hu^{'}_{n}-\frac{h^{2}u^{''}_{n}}{2}+\frac{h^{3}u^{'''}_{n}}{6}-2hu^{'}_{n}+O\left(h^{4}\right)=\frac{h^{3}u^{'''}_{n}}{3}+O\left(h^{4}\right)\hspace{0.5cm} (8) 
$$

$\therefore$ By $(8)$, since the LTE is $\frac{h^{3}u^{'''}_{n}}{3}$ (i.e. $O(h^{3})$), this is a second-order method. 

**Part b:**

For $r=2$ with the Milne method, we still have:

$$
u(t_{n+2})-u(t_{n})=\int_{t_{n}}^{t_{n+2}}f(u(t),t)dt \approx \int_{t_{n}}^{t_{n+2}}P(t)dt \hspace{0.5cm} (1)
$$

However, this time, $P(t)$ is the Lagrange polynomial that interpolates $(t_{n},f_{n})$, $(t_{n+1},f_{n+1})$, and $(t_{n+2},f_{n+2})$. Thus, 

$$
P(t)=f_{n}p_{n}(t)+f_{n+1}p_{n+1}(t)+f_{n+2}p_{n+2}(t)=f_{n}\frac{(t-t_{n+1})(t-t_{n+2})}{(t_{n}-t_{n+1})(t_{n}-t_{n+2})}+f_{n+1}\frac{(t-t_{n})(t-t_{n+2})}{(t_{n+1}-t_{n})(t_{n+1}-t_{n+2})}+f_{n+2}\frac{(t-t_{n})(t-t_{n+1})}{(t_{n+2}-t_{n})(t_{n+2}-t_{n+1})} \hspace{0.5cm} (2)
$$ 

Again defining $h:=t_{n+2}-t_{n+1}:=t_{n+1}-t_{n}$ and substituting $(2)$ all into $(1)$, we get:

$$
u_{n+2}-u_{n}=\frac{1}{h^{2}}\int_{t_{n}}^{t_{n+2}}\left(\frac{f_{n}}{2}(t-t_{n+1})(t-t_{n+2})-f_{n+1}(t-t_{n})(t-t_{n+2})+\frac{f_{n+2}}{2}(t-t_{n+1})(t-t_{n})\right)dt \\ \begin{align} t=t_{n}+2h\theta \\ dt=2hd\theta \end{align} \Rightarrow u_{n+2}-u_{n}=2h\int_{0}^{1}\left(\frac{f_{n}}{2}(2\theta -1)(2\theta -2)-f_{n+1}(2\theta)(2\theta -2)+\frac{f_{n+2}}{2}(2\theta -1)(2\theta)\right)d\theta=2h\left(\frac{f_{n}}{6}+\frac{2f_{n+1}}{3}+\frac{f_{n+2}}{6}\right) \\ \Rightarrow \boxed{\therefore u_{n+2}=u_{n}+\frac{h}{3}\left(f_{n}+4f_{n+1}+f_{n+2}\right)} \hspace{0.5cm} (3)
$$

This is the Milne method with $r=2$, as desired. 

Next, we prove that this method is fourth-order. First, we define the LTE from $(3)$, $e(u_{n})$, noting that $f_{n}\equiv u^{'}_{n}$, $f_{n+1}\equiv u^{'}_{n+1}$, and $f_{n+2}\equiv u^{'}_{n+2}$.

$$
e(u_{n})=u_{n+2}-u_{n}-\frac{h}{3}\left(f_{n}+4f_{n+1}+f_{n+2}\right)=u_{n+2}-u_{n}-\frac{hu^{'}_{n}}{3}-\frac{4hu^{'}_{n+1}}{3}-\frac{hu^{'}_{n+2}}{3} \hspace{0.5cm} (4)
$$

We now take Taylor series of $u_{n+2}$ and $u_{n}$, both about $u_{n+1}$.
$$
u_{n+2}=u_{n+1}+u^{'}_{n+1}(t_{n+2}-t_{n+1})+\frac{u^{''}_{n+1}}{2}(t_{n+2}-t_{n+1})^{2}+\frac{u^{(3)}_{n+1}}{6}(t_{n+2}-t_{n+1})^{3}+\frac{u^{(4)}_{n+1}}{24}(t_{n+2}-t_{n+1})^{4}+\frac{u^{(5)}_{n+1}}{120}(t_{n+2}-t_{n+1})^{5}+O\left((t_{n+2}-t_{n+1})^{6}\right) \\ u_{n+2}=u_{n+1}+hu^{'}_{n+1}+\frac{h^{2}u^{''}_{n+1}}{2}+\frac{h^{3}u^{(3)}_{n+1}}{6}+\frac{h^{4}u^{(4)}_{n+1}}{24}+\frac{h^{5}u^{(5)}_{n+1}}{120}+O\left(h^{6}\right) \hspace{0.5cm} (5) \\
u_{n}=u_{n+1}+u^{'}_{n+1}(t_{n}-t_{n+1})+\frac{u^{''}_{n+1}}{2}(t_{n}-t_{n+1})^{2}+\frac{u^{(3)}_{n+1}}{6}(t_{n}-t_{n+1})^{3}+\frac{u^{(4)}_{n+1}}{24}(t_{n}-t_{n+1})^{4}+\frac{u^{(5)}_{n+1}}{120}(t_{n}-t_{n+1})^{5}+O\left((t_{n}-t_{n+1})^{6}\right) \\ u_{n}=u_{n+1}-hu^{'}_{n+1}+\frac{h^{2}u^{''}_{n+1}}{2}-\frac{h^{3}u^{(3)}_{n+1}}{6}+\frac{h^{4}u^{(4)}_{n+1}}{24}-\frac{h^{5}u^{(5)}_{n+1}}{120}+O\left(h^{6}\right) \hspace{0.5cm} (6)
$$

Now, we also need the Taylor series for $u^{'}_{n+2}$ and $u^{'}_{n}$, both about $u^{'}_{n+1}$, so we differentiate $(5)$ and $(6)$, respectively:
$$
u^{'}_{n+2}=u^{'}_{n+1}+hu^{''}_{n+1}+\frac{h^{2}u^{(3)}_{n+1}}{2}+\frac{h^{3}u^{(4)}_{n+1}}{6}+\frac{h^{4}u^{(5)}_{n+1}}{24}+\frac{h^{5}u^{(6)}_{n+1}}{120}+O\left(h^{6}\right) \hspace{0.5cm} (7) \\
u^{'}_{n}=u^{'}_{n+1}-hu^{''}_{n+1}+\frac{h^{2}u^{(3)}_{n+1}}{2}-\frac{h^{3}u^{(4)}_{n+1}}{6}+\frac{h^{4}u^{(5)}_{n+1}}{24}-\frac{h^{5}u^{(6)}_{n+1}}{120}+O\left(h^{6}\right) \hspace{0.5cm} (8)
$$

Finally, for the ultimate step: we substitute $(5)$, $(6)$, $(7)$, and $(8)$ into $(4)$. This gives us the following rather large expression (which will still simplify nicely):
$$
e(u_{n})=\left(u_{n+1}+hu^{'}_{n+1}+\frac{h^{2}u^{''}_{n+1}}{2}+\frac{h^{3}u^{(3)}_{n+1}}{6}+\frac{h^{4}u^{(4)}_{n+1}}{24}+\frac{h^{5}u^{(5)}_{n+1}}{120}+O\left(h^{6}\right)\right)-\left(u_{n+1}-hu^{'}_{n+1}+\frac{h^{2}u^{''}_{n+1}}{2}-\frac{h^{3}u^{(3)}_{n+1}}{6}+\frac{h^{4}u^{(4)}_{n+1}}{24}-\frac{h^{5}u^{(5)}_{n+1}}{120}+O\left(h^{6}\right)\right)-\frac{h}{3}\left(u^{'}_{n+1}-hu^{''}_{n+1}+\frac{h^{2}u^{(3)}_{n+1}}{2}-\frac{h^{3}u^{(4)}_{n+1}}{6}+\frac{h^{4}u^{(5)}_{n+1}}{24}-\frac{h^{5}u^{(6)}_{n+1}}{120}+O\left(h^{6}\right)\right)-\frac{4hu^{'}_{n+1}}{3}-\frac{h}{3}\left(u^{'}_{n+1}+hu^{''}_{n+1}+\frac{h^{2}u^{(3)}_{n+1}}{2}+\frac{h^{3}u^{(4)}_{n+1}}{6}+\frac{h^{4}u^{(5)}_{n+1}}{24}+\frac{h^{5}u^{(6)}_{n+1}}{120}+O\left(h^{6}\right)\right) \\ e(u_{n})=2hu^{'}_{n+1}+\frac{h^{3}u^{(3)}_{n+1}}{3}+\frac{h^{5}u^{(5)}_{n+1}}{60}-2hu^{'}_{n+1}+\frac{h^{2}u^{''}_{n+1}}{3}-\frac{h^{2}u^{''}_{n+1}}{3}-\frac{h^{3}u^{(3)}_{n+1}}{6}-\frac{h^{3}u^{(3)}_{n+1}}{6}+\frac{h^{4}u^{(4)}_{n+1}}{18}-\frac{h^{4}u^{(4)}_{n+1}}{18}-\frac{h^{5}u^{(5)}_{n+1}}{72}-\frac{h^{5}u^{(5)}_{n+1}}{72}+O\left(h^{6}\right) \\ e(u_{n})=\frac{h^{3}u^{(3)}_{n+1}}{3}-\frac{h^{3}u^{(3)}_{n+1}}{3}+\frac{h^{5}u^{(5)}_{n+1}}{60}-\frac{h^{5}u^{(5)}_{n+1}}{36}+O\left(h^{6}\right)=-\frac{h^{5}u^{(5)}_{n+1}}{90}+O\left(h^{6}\right) \hspace{0.5cm} (9)
$$

$\therefore$ By $(9)$, since the LTE is $-\frac{h^{5}u^{(5)}_{n+1}}{90}$ (i.e. $O(h^{5})$), this is a fourth-order method. 

## Problem 3

The Rabi oscillation in quantum physics can be modeled by the following ODEs
$$
\begin{split}
i\partial_t u_1(t) &= \Omega \cos (\omega t) u_2(t),\\
i\partial_t u_2(t) &= \Omega_0 u_2(t) + \Omega \cos (\omega t) u_1(t).
\end{split}
$$

This model describes the transition of an electron from state 1 to state 2 under an external electric field $\mathcal{E}(t)=\Omega \cos(\omega t)$. $u_1(t),u_2(t)\in \mathbb{C}$, and $\Omega_0$ is the energy gap between the two states.


Given Initial condition $u_1(0)=1,u_2(0)=0$, and $\Omega_0=2\pi,\Omega=1.0$.  Write the 4th order Runge-Kutta method (rk4) which returns the full trajectory of the two components of $u$ as $u[1,:]$ and $u[2,:]$.

a) Use RK4 scheme with $\Delta t = 0.1$ to solve this equation and plot out the trajectory of $|u_1(t)|^2$,$|u_2(t)|^2$ on the interval $[0,10\pi]$ for $\omega=\pi$. Do the same calculation for $\omega=2\pi$ and you should observe the resonance behavior (i.e. $|u_2|$ can reach 1.0 at some t).

b) Derive that $A(t)=|u_1(t)|^2+|u_2(t)|^2$ is a conserved quantity and plot $A(t)$ obtained from the RK4 scheme. Observe that you can improve on the accuracy of the conservation of $A(t)$ by choosing a smaller time step.



**Part a:**

It is pretty straightforward for us to write this IVP as a two-dimensional system of differential equations:

$$
\begin{split}
i\partial_{t} u_{1}(t) &= \Omega \cos (\omega t) u_{2}(t),\\
i\partial_{t} u_{2}(t) &= \Omega_0 u_{2}(t) + \Omega \cos (\omega t) u_{1}(t)
\end{split} 
\Rightarrow 
\begin{split}
\partial_{t} u_{1}(t) &= -i \Omega \cos (\omega t) u_{2}(t),\\
\partial_{t} u_{2}(t) &= -i \left(\Omega_0 u_{2}(t) + \Omega \cos (\omega t) u_{1}(t)\right)
\end{split} 
$$

All we used above was the fact that $\frac{1}{i}=-i$. Now, we define $\overrightarrow{u(t)}:=\begin{pmatrix} u_{1}(t) \\ u_{2}(t) \end{pmatrix}$, $M:=-i\begin{pmatrix} 0 & \Omega \cos (\omega t) \\ \Omega \cos (\omega t) & \Omega_0 \end{pmatrix}$, and $\overrightarrow{f}\left(t, \overrightarrow{u(t)}\right):=M\overrightarrow{u(t)}$, such that our system is of the form $\frac{\partial \overrightarrow{u(t)}}{\partial t}:=\dot{\overrightarrow{u(t)}}=\overrightarrow{f}\left(t, \overrightarrow{u(t)}\right)$. Before applying RK4 (i.e. *the* Runge-Kutta method), let us define a function (in Julia) which evaluates $\overrightarrow{f}\left(t, \overrightarrow{u(t)}\right)$ for given values of $\overrightarrow{u(t)}$, $t$, $\Omega$, $\Omega_{0}$, and $\omega$.

```julia
function func(u, t, Omega, Omega0, w)
    return -im * [0 Omega * cos(w * t); Omega * cos(w * t) Omega0] * u
end
```

The Julia function we defined here will be very useful later. Next, given an IVP of the form $\dot{\overrightarrow{u(t)}}=\overrightarrow{f}\left(t, \overrightarrow{u(t)}\right)$ with initial condition $\overrightarrow{u(t_{0})}:=\overrightarrow{u_{0}}$, a general RK4 scheme is defined as the following, with stepsize $h:=t_{n+1}-t_{n} \forall n$ chosen beforehand:

$$
\overrightarrow{u_{n+1}}=\overrightarrow{u_{n}}+\frac{h}{6}\left(\overrightarrow{k_{n,1}}+2\overrightarrow{k_{n,2}}+2\overrightarrow{k_{n,3}}+\overrightarrow{k_{n,4}}\right)
$$

where each $\overrightarrow{k_{n,j}}$ is defined $\forall j \in \{1, 2, 3, 4\}$ as:

$$
\overrightarrow{k_{n,1}}=\overrightarrow{f}\left(t_{n}, \overrightarrow{u_{n}}\right) \\
\overrightarrow{k_{n,2}}=\overrightarrow{f}\left(t_{n}+\frac{h}{2}, \overrightarrow{u_{n}}+\frac{h}{2}\overrightarrow{k_{n,1}}\right) \\
\overrightarrow{k_{n,3}}=\overrightarrow{f}\left(t_{n}+\frac{h}{2}, \overrightarrow{u_{n}}+\frac{h}{2}\overrightarrow{k_{n,2}}\right) \\
\overrightarrow{k_{n,4}}=\overrightarrow{f}\left(t_{n}+h, \overrightarrow{u_{n}}+h\overrightarrow{k_{n,3}}\right) 
$$

Thus, accordingly, we can use the RK4 method to numerically solve our IVP in question, given that we have already put it in the desired form of $\dot{\overrightarrow{u(t)}}=\overrightarrow{f}\left(t, \overrightarrow{u(t)}\right)$ above. We do this below:

```julia
using LinearAlgebra 

function rk4(N, T, u0, Omega, Omega0, w)
    h = T / N
    t = h * collect(0 : N)
    u = zeros(ComplexF32, 2, N + 1)
    u[:, 1] = u0
    for i = 1 : N
        ui = u[:, i]
        ti = t[i]
        k1 = func(ui, ti, Omega, Omega0, w)
        k2 = func(ui + 0.5 * h * k1, ti + 0.5 * h, Omega, Omega0, w)
        k3 = func(ui + 0.5 * h * k2, ti + 0.5 * h, Omega, Omega0, w)
        k4 = func(ui + h * k3, ti + h, Omega, Omega0, w)
        u[:, i + 1] = ui + (h / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
    end
    return u
end
```

We now use this to solve our IVP with the given conditions and produce two plots for $|u_1(t)|^2$ and $|u_2(t)|^2$ on the interval $[0,10\pi]$, the first for $\omega=\pi$, and the second for $\omega=2\pi$ (i.e. resonance).

```julia
using PyPlot

T = 10.0 * pi
dt = 0.1
N = round(Int64, T / dt)
t = collect(0 : N) * dt
Omega = 1.0
Omega0 = 2 * pi
w = pi
u0 = [1.0, 0.0]
upi = rk4(N, T, u0, Omega, Omega0, w)
u2pi = rk4(N, T, u0, Omega, Omega0, 2 * w)

figure(1, figsize = (5, 5))
plot(t, vec(abs.(upi[1, :]) .^ 2), "r-", label = L"|u_{1}|^{2}")
plot(t, vec(abs.(upi[2, :]) .^ 2), "b-", label = L"|u_{2}|^{2}")
xlabel("t")
legend()
title(L"rk4, \Delta t=0.1, \omega=\pi")
figure(2, figsize = (5, 5))
plot(t, vec(abs.(u2pi[1, :]) .^ 2), "r-", label = L"|u_{1}|^{2}")
plot(t, vec(abs.(u2pi[2, :]) .^ 2), "b-", label = L"|u_{2}|^{2}")
xlabel("t")
legend()
title(L"rk4, \Delta t=0.1, \omega=2\pi")
```

For the second case of $\omega=2\pi$, we note that, with our given parameters, resonance is only reached (i.e. $|u_1(t)|^2$ or $|u_2(t)|^2$ reaches 1) at $t=0$. This is rather suspect, since we do not see any damping terms in our original ODE, nor any sources of dissipation that are accounted for in our original physical scenario. Thus, this is evidence that our numerical scheme might be somewhat flawed. However, in fact, it really is not. All we need to do is decrease our step size to $h=0.01$, and we immediately yield a plot which makes physical and mathematical sense. 

```julia
using PyPlot

T = 10.0 * pi
dt = 0.01
N = round(Int64, T / dt)
t = collect(0 : N) * dt
Omega = 1.0
Omega0 = 2 * pi
w = 2 * pi
u0 = [1.0, 0.0]
u = rk4(N, T, u0, Omega, Omega0, w)

figure(3, figsize = (5, 5))
plot(t, vec(abs.(u[1, :]) .^ 2), "r-", label = L"|u_{1}|^{2}")
plot(t, vec(abs.(u[2, :]) .^ 2), "b-", label = L"|u_{2}|^{2}")
xlabel("t")
legend()
title(L"rk4, \Delta t=0.01, \omega=2\pi")
```

**Part b:**

We start off by explicitly showing that the oscillation amplitude, $A(t)=|u_{1}(t)|^{2}+|u_{2}(t)|^{2}$, is conserved $\Leftrightarrow \frac{\partial A}{\partial t}=0$. 

$$
\frac{\partial A}{\partial t}=\frac{\partial}{\partial t}\left(|u_{1}(t)|^{2}+|u_{2}(t)|^{2}\right)=\frac{\partial}{\partial t}\left(u_{1}(t)\overline{u_{1}(t)}+u_{2}(t)\overline{u_{2}(t)}\right)=\partial_{t}u_{1}(t)\overline{u_{1}(t)}+u_{1}(t)\partial_{t}\overline{u_{1}(t)}+\partial_{t}u_{2}(t)\overline{u_{2}(t)}+u_{2}(t)\partial_{t}\overline{u_{2}(t)} \hspace{0.5cm} (1) 
$$

Now, note from part a that $\partial_{t} u_{1}(t) = -i \Omega \cos (\omega t) u_{2}(t)$ and $\partial_{t} u_{2}(t) = -i \Omega_0 u_{2}(t) -i \Omega \cos (\omega t) u_{1}(t)$, which implies by complex conjugation that $\partial_{t}\overline{u_{1}(t)} = i \Omega \cos (\omega t) \overline{u_{2}(t)}$ and $\partial_{t} \overline{u_{2}(t)} = i \Omega_0 \overline{u_{2}(t)} +i \Omega \cos (\omega t) \overline{u_{1}(t)}$. Therefore, plugging these into $(1)$, we have:

$$
\frac{\partial A}{\partial t}=\left(-i \Omega \cos (\omega t) u_{2}(t)\right) \overline{u_{1}(t)}+u_{1}(t)\left(i \Omega \cos (\omega t) \overline{u_{2}(t)}\right)+\left(-i \Omega_0 u_{2}(t) -i \Omega \cos (\omega t) u_{1}(t)\right)\overline{u_{2}(t)}+u_{2}(t)\left(i \Omega_0 \overline{u_{2}(t)} +i \Omega \cos (\omega t) \overline{u_{1}(t)}\right) \\ \Rightarrow 
\frac{\partial A}{\partial t} = -i \Omega \cos (\omega t) u_{2}(t) \overline{u_{1}(t)}+i \Omega \cos (\omega t) u_{1}(t) \overline{u_{2}(t)} -i \Omega_0 u_{2}(t)\overline{u_{2}(t)} -i \Omega \cos (\omega t) u_{1}(t)\overline{u_{2}(t)}+i \Omega_0 u_{2}(t)\overline{u_{2}(t)} +i \Omega \cos (\omega t) u_{2}(t)\overline{u_{1}(t)} \hspace{0.5cm} (2)
$$

Clearly, all of the terms in $(2)$ cancel out each other in pairs, so it follows that $\boxed {\therefore \frac{\partial A}{\partial t} = 0}$. Therefore, the oscillation amplitude, $A(t)=|u_{1}(t)|^{2}+|u_{2}(t)|^{2}$, is conserved in time, so $A(t)=|u_{1}(t)|^{2}+|u_{2}(t)|^{2} \equiv C$ for some constant $C \in \mathbb{R}$. Based off of the plot (the more accurate one) in part a for $|u_1(t)|^2$ and $|u_2(t)|^2$, we can expect $C \approx 1$. 

Next, we plot $A(t)$ as derived from the RK4 scheme used in part a. At first, we do this with $h=0.1$, $\omega=\pi$.

```julia
using PyPlot

T = 10.0 * pi
dt = 0.1
N = round(Int64, T / dt)
t = collect(0 : N) * dt
Omega = 1.0
Omega0 = 2 * pi
w = pi
u0 = [1.0, 0.0]
u = rk4(N, T, u0, Omega, Omega0, w)

figure(4, figsize = (5, 5))
plot(t, vec(abs.(u[1, :]) .^ 2 + abs.(u[2, :]) .^ 2), "g-", label = L"A(t) = |u_{1}|^{2} + |u_{2}|^{2}")
xlabel("t")
legend()
title(L"rk4, \Delta t=0.1, \omega=\pi")
```

The above graph of $A(t)$ should be constant, so the fact that it is not is rather troubling (however, it is pretty close). Now, hoping to get more precise results, we try a stepsize of $h=0.01$.

```julia
using PyPlot

T = 10.0 * pi
dt = 0.01
N = round(Int64, T / dt)
t = collect(0 : N) * dt
Omega = 1.0
Omega0 = 2 * pi
w = pi
u0 = [1.0, 0.0]
u = rk4(N, T, u0, Omega, Omega0, w)
figure(5, figsize = (5, 5))
plot(t, vec(abs.(u[1, :]) .^ 2 + abs.(u[2, :]) .^ 2), "g-", label = L"A(t) = |u_{1}|^{2} + |u_{2}|^{2}")
xlabel("t")
legend()
title(L"rk4, \Delta t=0.01, \omega=\pi")
```

Thus, by choosing a stepsize which was only one order of magnitude smaller, our uncertainty in $A(t)$ is now something on the order of ~$10^{-6}$, as opposed to ~$10^{-2}$ like we had before. Although, what is rather peculiar is how the error in $A(t)$ looks almost linear for $h=0.1$, while it appears chaotic for $h=0.01$. It probably has something to do with some kind of polynomial error growth in the RK4 scheme if $h$ is too large, but once $h$ is chosen to be small enough, this no longer occurs, and we are left with much deeper, more inherent sources of error at each timestep.

```julia
using SparseArrays
using PyPlot
using LinearAlgebra

function Trapezoidal(N, T, u0, M)
    h = T / N
    u = copy(u0)
    t = h * collect(0 : N) 
    Atop = sparse(I + 0.5 * h * M)
    Abottom = sparse(I - 0.5 * h * M)
    for i = 1 : N
        u = Abottom \ (Atop * u)
    end
    return u
end

T  = 1.0
h = 0.1
N  = T / h

n = 100
dx = 2 * pi / n
r = 1 / dx ^ 2
M = r * spdiagm(-1 => ones(n - 1), 0 => -2 * ones(n), 1 => ones(n - 1))
#boundary conditions enforced here:
M[1, n] = r
M[n, 1] = r
xq = dx * collect(0 : n - 1)
u0 = exp.(-5 * (xq .- pi) .^ 2)

uT = Trapezoidal(N, T, u0, M)

figure(1, figsize = (5, 5))
plot(xq, u0, "r--", label = L"u(x, T = 0.0)")
plot(xq, uT, "m-", label = L"u(x, T = 1.0)")
legend()
title("Solution to the 1D heat equation using trapezoidal rule, h=0.1", pad = 10)
```

```julia
using SparseArrays
using PyPlot
using LinearAlgebra

function TrapezoidAB2(T, h0, u0, tol, M)
    t = 0.0
    h = h0
    u = copy(u0)
    upred = copy(u0)
    d = length(u0)
    uvec = zeros(d, 2)
    f = zeros(d, 2)
    uvec[:, 1] = copy(u0)
    f[:, 1] = M * u
    tvec = [t]
    
    #Initial step from trapezoidal rule
    Atop = sparse(I + 0.5 * h * M)
    Abottom = sparse(I - 0.5 * h * M)
    u = Abottom \ (Atop * u)
    t = t + h
    append!(tvec, t)
    uvec[:, 2] = copy(u)
    f[:, 2] = M * u
    didwewanttocontinue = false
    
    #Start while loop
    while (t < T)
        #Halt loop if we are already at t = T
        if (t + h >= T)
            append!(tvec, T)
            u = Abottom \ (Atop * u)
            break
        end
        #Check errors using Milne device, as detailed in lecture notes 3
        u = Abottom \ (Atop * u)
        upred = uvec[:, 2] + h * (1.5 * f[:, 2] - 0.5 * f[:, 1])
        error = norm(u - upred)
        println(error)
        if (error <= (1 / 6) * h * tol && didwewanttocontinue == false)
            #If our error is small enough, we may continue, doubling the step size
            t = t + h
            append!(tvec, t)
            uvec[:, 2] = copy(u)
            f[:, 2] = M * u
            h = 2 * h
            didwewanttocontinue = true
        elseif(error > 6 * h * tol)
            #If our error is too large, we will not continue, and halve the stepsize instead. 
            #In other words, try again! Do not change the stepsize history until we are satisfied with the stepsize.
            u = copy(uvec[:, 2])
            h = 0.5 * h
            uvec[:, 1] = Abottom \ (Atop * uvec[:, 1])
            f[:, 1] = M * uvec[:, 1]
        else
            #our error is good, so we continue as usual
            t = t + h
            append!(tvec, t)
            uvec[:, 1] = copy(uvec[:, 2])
            f[:, 1] = copy(f[:, 2])
            uvec[:, 2] = copy(u)
            f[:, 2] = M * u    
            didwewanttocontinue = false
        end
        if( h < 1e-10 )
            break
        end
    end
    return u, tvec
end

T  = 1.0
h0 = 0.001
tol = 1e-3

n = 100
dx = 2 * pi / n
r = 1 / dx ^ 2
M = r * spdiagm(-1 => ones(n - 1), 0 => -2 * ones(n), 1 => ones(n - 1))
#boundary conditions enforced here:
M[1, n] = r
M[n, 1] = r
xq = dx * collect(0 : n - 1)
u0 = exp.(-5 * (xq .- pi) .^ 2)

uT, tvec = TrapezoidAB2(T, h0, u0, tol, M)

figure(1, figsize = (5, 5))
plot(xq, u0, "r--", label = L"u(x, T = 0.0)")
plot(xq, uT, "m-", label = L"u(x, T = 1.0)")
legend()
title("Solution to the 1D heat equation using trapezoidal rule (monitored by AB2)", pad = 10)

figure(2, figsize = (5, 5))
plot(tvec[1 : end - 1], diff(tvec), "b-o")
xlabel("t")
ylabel("h")
```
