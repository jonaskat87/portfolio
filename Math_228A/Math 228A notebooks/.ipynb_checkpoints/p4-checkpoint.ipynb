## Problem 1

For a separable Hamiltonian system,

$$H(p,q) = T(p)+V(q),\quad p,q\in \mathbb{R}^d$$

with corresponding Hamiltonian equation

$$p'(t)=-\nabla_q V(q),\quad q'(t)=\nabla_p T(p),$$


prove that the symplectic Euler method 

$$
\begin{split}
p_{n+1} &= p_n - \Delta t \nabla_q V(q_n),\\
q_{n+1} &= q_n + \Delta t \nabla_p T(p_{n+1}),
\end{split}
$$

is 

a) an order 1 method.

b) a symplectic method.


**Proof (a):** We start off by taking simultaneous Taylor expansions of $p(t_{n+1})$ and $q(t_{n+1})$ about $p(t_{n})$ and $q(t_{n})$, respectively, and use the definitions of $p^{'}(t)$ and $q^{'}(t)$ in Hamilton's equations to further simplify our expressions.

$$
p\left(t_{n+1}\right)=p\left(t_{n}\right)+\left(t_{n+1}-t_{n}\right)p^{'}\left(t_{n}\right)+\left(t_{n+1}-t_{n}\right)^{2}\frac{p^{''}\left(t_{n}\right)}{2!}+O\left(\left(t_{n+1}-t_{n}\right)^{3}\right) \\ q\left(t_{n+1}\right)=q\left(t_{n}\right)+\left(t_{n+1}-t_{n}\right)q^{'}\left(t_{n}\right)+\left(t_{n+1}-t_{n}\right)^{2}\frac{q^{''}\left(t_{n}\right)}{2!}+O\left(\left(t_{n+1}-t_{n}\right)^{3}\right) \hspace{0.5cm} (1)
$$

Then, assuming a uniform stepsize, we define $\Delta t:=t_{n+1}-t_{n}$, such that $(1)$ becomes

$$
p\left(t_{n+1}\right)=p\left(t_{n}\right)+p^{'}\left(t_{n}\right)\Delta t+\frac{p^{''}\left(t_{n}\right)}{2}\left(\Delta t\right)^{2}+O(\left(\Delta t\right)^{3}) \\ q\left(t_{n+1}\right)=q\left(t_{n}\right)+q^{'}\left(t_{n}\right)\Delta t+\frac{q^{''}\left(t_{n}\right)}{2}\left(\Delta t\right)^{2}+O(\left(\Delta t\right)^{3}) \hspace{0.5cm} (2)
$$

Finally, from Hamilton's equations, we have that $p^{'}\left(t_{n}\right)\equiv-\nabla_{q} V\left(q\left(t_{n}\right)\right)$ and $q^{'}\left(t_{n}\right)\equiv\nabla_{p} T\left(p\left(t_{n}\right)\right)$, which means that $(2)$ can be further rewritten as

$$
p\left(t_{n+1}\right)=p\left(t_{n}\right)-\Delta t\nabla_{q} V\left(q\left(t_{n}\right)\right)+O(\left(\Delta t\right)^{2}) \\ q\left(t_{n+1}\right)=q\left(t_{n}\right)+\Delta t\nabla_{p} T\left(p\left(t_{n}\right)\right)+O(\left(\Delta t\right)^{2}) \hspace{0.5cm} (3)
$$

From here, we want to find the LTE (local truncation error) for the symplectic Euler method at each timestep. Suppose that, for some $n$, $p\left(t_{n}\right)=p_{n}$ and $q\left(t_{n}\right)=q_{n}$. Define the LTE's for $p_{n+1}$ and $q_{n+1}$ as $e_{p_{n+1}}:=\left|p\left(t_{n+1}\right)-p_{n+1}\right|_{2}$ and $e_{q_{n+1}}:=\left|q\left(t_{n+1}\right)-q_{n+1}\right|_{2}$, respectively, where $|\cdot|_{2}$ is the $2$-norm in $\mathbb{R}^d$. Then, from symplectic Euler method applied on the $n$th step,

$$
e_{p_{n+1}}=\left|p\left(t_{n}\right)-\Delta t\nabla_{q} V\left(q\left(t_{n}\right)\right)+O(\left(\Delta t\right)^{2})-\left(p_{n}-\Delta t \nabla_q V(q_{n})\right)\right|_{2}=\left|p_{n}-\Delta t\nabla_{q} V\left(q_{n}\right)+O(\left(\Delta t\right)^{2})-p_{n}+\Delta t \nabla_q V(q_{n})\right|_{2} \\ \Rightarrow \boxed{\therefore e_{p_{n+1}}=O(\left(\Delta t\right)^{2})} \hspace{0.5cm} (4) \\
e_{q_{n+1}}=\left|q\left(t_{n}\right)+\Delta t\nabla_{p} T\left(p\left(t_{n}\right)\right)+O(\left(\Delta t\right)^{2})-\left(q_{n}+\Delta t \nabla_p T(p_{n+1})\right)\right|_{2}=\left|q_{n}+\Delta t\nabla_{p} T\left(p_{n}\right)+O(\left(\Delta t\right)^{2})-q_{n}-\Delta t \nabla_p T(p_{n+1})\right|_{2} \\ \Rightarrow e_{q_{n+1}}=\left|\Delta t\left(\nabla_{p} T\left(p_{n}\right)-\nabla_p T(p_{n+1})\right)+O(\left(\Delta t\right)^{2})\right|_{2} \hspace{0.5cm} (5)
$$

We are done with showing the order of the method on our "momentum" variable $p$, but for our "position" variable $q$, we need to further simplify what we currently have. Taking the time derivative of both sides of the second equation for $(3)$, we get that

$$
q^{'}\left(t_{n+1}\right)=q^{'}\left(t_{n}\right)+\Delta t\frac{d}{dt}\left[\nabla_{p} T\left(p\left(t_{n}\right)\right)\right]+O(\left(\Delta t\right)^{2}) \Rightarrow \nabla_p T\left(p\left(t_{n+1}\right)\right)=\nabla_p T\left(p_{n}\right)+\Delta t\frac{d}{dt}\left[\nabla_{p} T\left(p\left(t_{n}\right)\right)\right]+O(\left(\Delta t\right)^{2}) \\ \Rightarrow \nabla_p T\left(p_{n}\right)=\nabla_p T\left(p\left(t_{n+1}\right)\right)-\Delta t\frac{d}{dt}\left[\nabla_{p} T\left(p\left(t_{n}\right)\right)\right]+O(\left(\Delta t\right)^{2}) \hspace{0.5cm} (6)
$$ 

Plugging in $(6)$ into $(5)$, we now have

$$
e_{q_{n+1}}=\left|\Delta t\left(\nabla_p T\left(p\left(t_{n+1}\right)\right)-\Delta t\frac{d}{dt}\left[\nabla_{p} T\left(p\left(t_{n}\right)\right)\right]-\nabla_p T(p_{n+1})\right)+O(\left(\Delta t\right)^{2})\right|_{2} \\ \Rightarrow e_{q_{n+1}}=\left|\Delta t\left(\nabla_p T\left(p\left(t_{n+1}\right)\right)-\nabla_p T(p_{n+1})\right)-\left(\Delta t\right)^{2}\frac{d}{dt}\left[\nabla_{p} T\left(p\left(t_{n}\right)\right)\right]+O(\left(\Delta t\right)^{2})\right|_{2}=\left|\Delta t\left(\nabla_p T\left(p\left(t_{n+1}\right)\right)-\nabla_p T(p_{n+1})\right)+O(\left(\Delta t\right)^{2})\right|_{2} \hspace{0.5cm} (7)
$$

Finally, if we take $\nabla_p T(p)$ to be Lipschitz continuous over a sufficiently large neighborhood in $\mathbb{R}^d$ enclosing $p_{n+1}$ and $p\left(t_{n+1}\right)$, then for some Lipschitz constant $L>0$, $\left|\nabla_p T\left(p\left(t_{n+1}\right)\right)-\nabla_p T(p_{n+1})\right|_{2}\leq L\left|p\left(t_{n+1}\right)-p_{n+1}\right|_{2}=Le_{p_{n+1}}=O(\left(\Delta t\right)^{2})$ (by $(4)$). (Or alternatively, one can take yet another Taylor expansion of $\nabla_p T(p)$ about $p_{n+1}$ and let $p=p\left(t_{n+1}\right)$.) Resultantly, applying the triangle inequality to $(7)$,

$$
e_{q_{n+1}}=\left|\Delta t\left(\nabla_p T\left(p\left(t_{n+1}\right)\right)-\nabla_p T(p_{n+1})\right)+O(\left(\Delta t\right)^{2})\right|_{2}\leq O(\left(\Delta t\right)^{2}) + \Delta t\left|\nabla_p T\left(p\left(t_{n+1}\right)\right)-\nabla_p T(p_{n+1})\right|_{2}\leq O(\left(\Delta t\right)^{2}) + O(\left(\Delta t\right)^{3}) \\ \Rightarrow \boxed{\therefore e_{q_{n+1}}=O(\left(\Delta t\right)^{2})} \hspace{0.5cm} (8)
$$

By $(4)$ and $(8)$, the LTE in both $p$ and $q$ is 2nd-order (i.e. $O(\left(\Delta t\right)^{2})$), so globally, the symplectic Euler method is 1st-order. Q.E.D.

**Proof (b):** First, we take the symplectic Euler method:

$$
p_{n+1}=p_{n}-\Delta t\nabla V\left(q_{n}\right) \\ q_{n+1}=q_{n}+\Delta t\nabla T\left(p_{n+1}\right) \hspace{0.5cm} (1)
$$

Now, we define the Jacobian of the dynamical system governed by $(1)$ as $\Phi_{n}:=\begin{pmatrix}\frac{\partial p_{n+1}}{\partial p_{n}} & \frac{\partial p_{n+1}}{\partial q_{n}} \\ \frac{\partial q_{n+1}}{\partial p_{n}} & \frac{\partial q_{n+1}}{\partial q_{n}}\end{pmatrix}$, where $I$ is defined to be the identity matrix in $d$-dimensions. Then, $(1)$ gives us its entries as the following, defining $\nabla^{2}$ to be the Hessian matrix of whatever it is applied to:

$$
\frac{\partial p_{n+1}}{\partial p_{n}}=\frac{\partial}{\partial p_{n}}\left[p_{n}-\Delta t\nabla V\left(q_{n}\right)\right]=I \\ \frac{\partial p_{n+1}}{\partial q_{n}}=\frac{\partial}{\partial q_{n}}\left[p_{n}-\Delta t\nabla V\left(q_{n}\right)\right]=-\Delta t\nabla^{2} V\left(q_{n}\right) \\ \frac{\partial q_{n+1}}{\partial p_{n}}=\frac{\partial}{\partial p_{n}}\left[q_{n}+\Delta t\nabla T\left(p_{n+1}\right)\right]=\frac{\partial p_{n+1}}{\partial p_{n}}\frac{\partial}{\partial p_{n+1}}\left[q_{n}+\Delta t\nabla T\left(p_{n+1}\right)\right]=\left(I\right)\left(\Delta t\nabla^{2} T\left(p_{n+1}\right)\right)=\Delta t\nabla^{2} T\left(p_{n+1}\right) \\ \frac{\partial q_{n+1}}{\partial q_{n}}=\frac{\partial}{\partial q_{n}}\left[q_{n}+\Delta t\nabla T\left(p_{n+1}\right)\right]=I+\Delta t\frac{\partial p_{n+1}}{\partial q_{n}}\frac{\partial}{\partial p_{n+1}}\left[\nabla T\left(p_{n+1}\right)\right]=I-\left(\Delta t\right)^{2}\nabla^{2} V\left(q_{n}\right)\nabla^{2} T\left(p_{n+1}\right) \\ \Rightarrow \Phi_{n}=\begin{pmatrix}I & -\Delta t\nabla^{2} V\left(q_{n}\right) \\ \Delta t\nabla^{2} T\left(p_{n+1}\right) & I-\left(\Delta t\right)^{2}\nabla^{2} V\left(q_{n}\right)\nabla^{2} T\left(p_{n+1}\right)\end{pmatrix} \hspace{0.5cm} (2)
$$

For visual ease, we define $X:=\Delta t\nabla^{2}T\left(p_{n+1}\right)$ and $Y:=-\Delta t\nabla^{2}V\left(q_{n}\right)$. Then, the Jacobian, $\Phi_{n}$, as written in $(2)$, can be further decomposed as follows:

$$
\Phi_{n}\equiv\begin{pmatrix}I & Y \\ X & I+XY\end{pmatrix}=\begin{pmatrix}I & 0 \\ X & I\end{pmatrix}\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix} \hspace{0.5cm} (3)
$$

Now, we want to show that the flow map described by $(1)$ is symplectic. In other words, we want to show that our Jacobian, $\Phi_{n}$, satisfies the equation $\Phi_{n}^{T}J\Phi_{n}=J \hspace{0.15cm} (\star)$, where $J$ is the skew-symmetric matrix $J:=\begin{pmatrix}0 & I \\ -I & 0\end{pmatrix}$. To do this, we prove a useful lemma, as suggested in class.

*Lemma*: For all matrices $A\in\mathbb{R}^{d\times d}$ such that A is symmetric (i.e. $A=A^{T}$), we have that

$$
\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}^{T}J\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}=J=\begin{pmatrix}I & A \\ 0 & I\end{pmatrix}^{T}J\begin{pmatrix}I & A \\ 0 & I\end{pmatrix} \hspace{0.5cm} (*)
$$

where $J$ is the same as above.

*Proof of lemma*: We proceed by directly computing the left- and right-hand sides of $(*)$.

$$
\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}^{T}J\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}=\begin{pmatrix}I^{T} & A^{T} \\ 0 & I^{T}\end{pmatrix}\begin{pmatrix}0 & I \\ -I & 0\end{pmatrix}\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}=\begin{pmatrix}I & A \\ 0 & I\end{pmatrix}\begin{pmatrix}0 & I \\ -I & 0\end{pmatrix}\begin{pmatrix}I & 0 \\ A & I\end{pmatrix} \\ \Rightarrow \begin{pmatrix}I & 0 \\ A & I\end{pmatrix}^{T}J\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}=\begin{pmatrix}-A & I \\ -I & 0\end{pmatrix}\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}=\begin{pmatrix}0 & I \\ -I & 0\end{pmatrix} \Rightarrow \boxed{\therefore \begin{pmatrix}I & 0 \\ A & I\end{pmatrix}^{T}J\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}=J} \\ \begin{pmatrix}I & A \\ 0 & I\end{pmatrix}^{T}J\begin{pmatrix}I & A \\ 0 & I\end{pmatrix}=\begin{pmatrix}I^{T} & 0 \\ A^{T} & I^{T}\end{pmatrix}\begin{pmatrix}0 & I \\ -I & 0\end{pmatrix}\begin{pmatrix}I & A \\ 0 & I\end{pmatrix}=\begin{pmatrix}I & 0 \\ A & I\end{pmatrix}\begin{pmatrix}0 & I \\ -I & 0\end{pmatrix}\begin{pmatrix}I & A \\ 0 & I\end{pmatrix} \\ \Rightarrow \begin{pmatrix}I & A \\ 0 & I\end{pmatrix}^{T}J\begin{pmatrix}I & A \\ 0 & I\end{pmatrix}=\begin{pmatrix}0 & I \\ -I & A\end{pmatrix}\begin{pmatrix}I & A \\ 0 & I\end{pmatrix}=\begin{pmatrix}0 & I \\ -I & 0\end{pmatrix} \Rightarrow \boxed{\therefore \begin{pmatrix}I & A \\ 0 & I\end{pmatrix}^{T}J\begin{pmatrix}I & A \\ 0 & I\end{pmatrix}=J}
$$

Therefore, both sides of $(*)$ have been shown to equal $J$. Q.E.D.

With the lemma proven, we return back to $(3)$ and show that $(\star)$ holds. Note also that, given that $V$ and $T$ are smooth enough, their respective Hessian matrices $\nabla^{2}V\left(q_{n}\right)$ and $\nabla^{2}T\left(p_{n+1}\right)$ should be symmetric. Thus, $X=X^{T}$ and $Y=Y^{T}$, which are facts that we will use below.

$$
\Phi_{n}^{T}J\Phi_{n}=\left[\begin{pmatrix}I & 0 \\ X & I\end{pmatrix}\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix}\right]^{T}J\begin{pmatrix}I & 0 \\ X & I\end{pmatrix}\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix} \\ \Phi_{n}^{T}J\Phi_{n}=\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix}^{T}\begin{pmatrix}I & 0 \\ X & I\end{pmatrix}^{T}J\begin{pmatrix}I & 0 \\ X & I\end{pmatrix}\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix}=\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix}^{T}\left[\begin{pmatrix}I & 0 \\ X & I\end{pmatrix}^{T}J\begin{pmatrix}I & 0 \\ X & I\end{pmatrix}\right]\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix} \hspace{0.5cm} (4)
$$

By the lemma, the term in the brackets is equal to $J$, so $(4)$ becomes the following, which we again simplify by our lemma:

$$
\Phi_{n}^{T}J\Phi_{n}=\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix}^{T}J\begin{pmatrix}I & Y \\ 0 & I\end{pmatrix} \Rightarrow \boxed{\therefore \Phi_{n}^{T}J\Phi_{n}=J}
$$

Therefore, since the Jacobian of the flow map is symplectic, it follows that the symplectic Euler method is also symplectic. Q.E.D.

## Problem 2

Consider the ODE 

$$u'(t) = \lambda (u − \cos t) − \sin t, \quad u(0) = u_0,$$

with the exact solution

$$u(t) = e^{\lambda t}(u_0 − 1) + \cos t.$$

Solve the equation and compare with the exact solution using 

a) The trapezoidal method

b) The backward Euler method

for $0\le t \le 3$. Use the parameter $\lambda=-10^6,u_0=1.5,\Delta t = 0.1$. Repeat with $\Delta t=0.001$. Explain the difference in terms of the behavior $R(z)$ at $z=\infty$.

**Part a**

(For both parts, I will derive the trapezoidal and backward Euler numerical methods for this ODE, then graph them as desired for the problem afterwards.)

For the trapezoidal method, we basically approximate the finite difference for $u$ on the interval $[t_{n}, t_{n+1}]$ as an average of the derivatives for $u$, $u^{'}(t):=f\left(u,t\right)$, at $t_{n}$ (i.e. at the beginning of the time interval) and at $t_{n+1}$ (i.e. at the end). Thus, defining a uniform step size $h:=t_{n+1}-t_{n} \forall n$, we have that

$$
\frac{u_{n+1}-u_{n}}{t_{n+1}-t_{n}}=\frac{u_{n+1}-u_{n}}{h}=\frac{f\left(u_{n+1},t_{n+1}\right)+f\left(u_{n},t_{n}\right)}{2} \Rightarrow u_{n+1}=u_{n}+\frac{h}{2}\left(f\left(u_{n+1},t_{n+1}\right)+f\left(u_{n},t_{n}\right)\right) \hspace{0.5cm} (1)
$$

In our case, $f\left(u, t\right)\equiv \lambda(u−\cos t)−\sin t$, so $(1)$ becomes

$$
u_{n+1}=u_{n}+\frac{h}{2}\left(\lambda \left(u_{n}-\cos t_{n}\right)-\sin t_{n}+\lambda \left(u_{n+1}-\cos t_{n+1}\right)-\sin t_{n+1}\right)=u_{n}+\frac{h}{2}\left(\lambda \left(u_{n}-\cos t_{n}\right)-\sin t_{n}+\lambda \left(u_{n+1}-\cos (t_{n}+h)\right)-\sin (t_{n}+h)\right) \\ \Rightarrow \left(1-\frac{h\lambda}{2}\right)u_{n+1}=\left(1+\frac{h\lambda}{2}\right)u_{n}+\frac{h}{2}\left(\lambda \left(u_{n}-\cos t_{n}\right)-\sin t_{n}-\lambda\cos (t_{n}+h)-\sin (t_{n}+h)\right) \\ \Rightarrow \boxed{\therefore u_{n+1}=\frac{2+h\lambda}{2-h\lambda}u_{n}-\frac{h}{2-h\lambda}\left(\lambda\left(\cos (t_{n}+h)+\cos (t_{n})\right)+\sin (t_{n}+h)+\sin(t_{n})\right)} 
$$

**Part b**

In the case of the backward Euler method, instead of taking an average of the derivative values at the endpoints of the interval $[t_{n}, t_{n+1}]$, we just use the derivative at $t_{n+1}$, or $u^{'}(t_{n+1}):=f\left(u_{n+1},t_{n+1}\right)$, to approximate the finite difference for $u$. Thus, defining the same $h:=t_{n+1}-t_{n} \forall n$ as before,

$$
\frac{u_{n+1}-u_{n}}{t_{n+1}-t_{n}}=\frac{u_{n+1}-u_{n}}{h}=f\left(u_{n+1},t_{n+1}\right) \Rightarrow u_{n+1}=u_{n}+hf\left(u_{n+1},t_{n}+h\right)=u_{n}+h\left(\lambda\left(u_{n+1}−\cos (t_{n}+h)\right)−\sin (t_{n}+h)\right) \\ \Rightarrow \left(1-h\lambda\right)u_{n+1}=u_{n}-h\left(\lambda\cos (t_{n}+h)+\sin (t_{n}+h)\right) \Rightarrow \boxed{\therefore u_{n+1}=\frac{u_{n}-h\left(\lambda\cos (t_{n}+h)+\sin (t_{n}+h)\right)}{1-h\lambda}}
$$

```julia
function BackEuler(N, T, u0, lambda)
    h = T / N
    u = zeros(N + 1)
    t = h * collect(0 : N)
    u[1] = u0
    R = 1 / (1 - h * lambda)
    for i = 1 : N
        u[i + 1] = R * (u[i] - h * (lambda * cos(t[i + 1]) + sin(t[i + 1])))
    end
    return u
end

function Trape(N, T, u0, lambda)
    h = T / N
    u = zeros(N + 1)
    t = h * collect(0 : N)
    u[1] = u0
    R1 = (2 + h * lambda) / (2 - h * lambda)
    R2 = h / (2 - h * lambda)
    for i = 1 : N
        u[i + 1] = R1 * u[i] - R2 * (lambda * (cos(t[i + 1]) + cos(t[i])) + sin(t[i + 1]) + sin(t[i]))
    end
    return u
end


u0 = 1.5
lambda = -1.0e+6
T = 3.0
h = 0.1
N = round(Int64, T / h)
t = collect(0 : N) * h
uexact = exp.(lambda * t) * (u0 - 1) + cos.(t)
uBackEuler = BackEuler(N, T, u0, lambda)
uTrape     = Trape(N, T, u0, lambda)

using PyPlot
figure(1, figsize=(5,5))
plot(t, uTrape, "r-d", label = "Trape")
plot(t, uexact, "k-<", label = "exact")
plot(t, uBackEuler, "b-o", label = "BackEuler")
legend()

h = 0.001
N = round(Int64, T / h)
t = collect(0 : N) * h
uexact = exp.(lambda * t) * (u0 - 1) + cos.(t)
uBackEuler = BackEuler(N, T, u0, lambda)
uTrape     = Trape(N, T, u0, lambda)
figure(2, figsize = (5, 5))
plot(t, uTrape, "r-", label = "Trape")
plot(t, uexact, "k-", label = "exact")
plot(t, uBackEuler, "b-", label = "BackEuler")
legend()
```

Note how, in either case, the backward Euler method clearly outperformed trapezoidal rule in terms of numerical stability, despite what error analysis might suggest. With a stepsize of $\Delta t=0.1$, the trapezoidal rule was highly unstable and oscillated pretty radically around the actual solution, while comparatively, backward Euler converged almost exactly to the analytical solution. Even with a stepsize of $\Delta t=0.001$, trapezoidal rule was still very unstable for the first thousand or so timesteps (i.e. up to around $t=1.0$), exhibiting the same kind of unstable oscillation, before finally converging to the analytical solution. Still, of course, backward Euler converged pretty instantly.

The ODE we are solving here is an example of what one might call a stiff equation. Stiff equations require special attention to solve numerically, as many of the usual numerical methods we apply on them do not behave well. This is especially the case when we have a system that grows or decreases exponentially at a large rate, such as this one, where $\lambda$ is very large in magnitude and very negative. 

In many cases, when dealing with stiff equations, one might only need to consider A-stability. Note that backward Euler and trapezoidal rule are both A-stable for the value of $z:=h\lambda$ used in this problem, since $z=-10^{5}$ in the first case and $z=-10^{3}$ in the second (i.e., $z<<0$.). Trapezoidal rule is A-stable whenever $Re(z)<0$, and backward Euler is A-stable whenever $z$ lies outside of the unit circle centered at $z=1$ in the complex plane. However, while backward Euler method worked very nicely, trapezoidal rule, despite being A-stable, performed very poorly. 

When working with stiff equations, it is also important to consider other, often stronger forms of stability. This is because A-stability only guarantees that our numerical solution will converge to the actual solution as $h\rightarrow 0$. It still does not tell us *how* small we will need to make $h$ for this to happen, especially when, as in this problem, the Jacobian of our system has eigenvalues which are very large in magnitude. We may still get cases where $h$ would need to be exceedingly small for our numerical scheme to stabilize quick enough, especially for a "very stiff* problem.

This is where **L-stability** is crucial to consider. As proved in lecture, L-stability actually implies A-stability, and is therefore a stronger condition. I claim that backward Euler method is indeed L-stable, while trapezoidal rule is *not* L-stable, which explains the behavior of both numerical methods in this problem. I will show this here.

**Backward Euler method:**

Consider the test problem $u^{'}=\lambda u$. Applying backward Euler to it with stepsize $h:=t_{n+1}-t_{n}$, we get that $u_{n+1}=u_{n}+h\lambda u_{n+1} \Rightarrow\left(1-h\lambda\right)u_{n+1}=u_{n}\Rightarrow u_{n+1}=\frac{1}{1-h\lambda}u_{n}$. Now, defining $z:=h\lambda$, we get that $u_{n+1}=\frac{1}{1-z}u_{n}:=R(z)u_{n}$. $R(z)$ is what is often called the "stability function." If a method is L-stable, then $\lim _{z\rightarrow\infty}{\left|R(z)\right|}=0$. Checking $\lim _{z\rightarrow\infty}{\left|R(z)\right|}$ for backward Euler method, we have that, 

$$
\forall \varepsilon>0, \hspace{0.15cm} z>\frac{1}{\varepsilon}\Rightarrow 0\leq \left|R(z)\right|=\frac{1}{\left|1-z\right|}<\frac{1}{\left|-z\right|}=\frac{1}{|z|}<\frac{1}{\frac{1}{\epsilon}}=\varepsilon
$$

Thus, by definition of limit, $\Rightarrow\boxed{\therefore\lim _{z\rightarrow\infty}{\left|R(z)\right|}=\lim _{z\rightarrow\infty}{\frac{1}{\left|1-z\right|}}=0}$. Therefore, backward Euler is L-stable wherever it is A-stable (stated above). Q.E.D.

**Trapezoidal rule:**

Now, consider the same test problem as above. This time, applying trapezoidal rule, we get that $u_{n+1}=u_{n}+\frac{h}{2}\left(\lambda u_{n}+\lambda u_{n+1}\right)\Rightarrow\left(1-\frac{h\lambda}{2}\right)u_{n+1}=\left(1+\frac{h\lambda}{2}\right)u_{n}\Rightarrow u_{n+1}=\frac{2+h\lambda}{2-h\lambda}u_{n}$. Again, defining $z:=h\lambda$, we get that $u_{n+1}=\frac{2+z}{2-z}u_{n}:=R(z)u_{n}$, so our "stability function" for trapezoidal rule is now $R(z)=\frac{2+z}{2-z}$. Checking $\lim _{z\rightarrow\infty}{\left|R(z)\right|}$ in this case, we have that,

$$
\forall \varepsilon>0, \hspace{0.15cm} z>2+\frac{4}{\varepsilon} \Rightarrow\left|\left|R(z)\right|-1\right|=\left|\left|\frac{2+z}{2-z}\right|-1\right|=\left|\frac{|2+z|-|2-z|}{|2-z|}\right| \hspace{0.5cm} (1)
$$

(Now, since $z>0$ eventually for this limit, $(1)$ becomes the following.)

$$
\left|\left|R(z)\right|-1\right|=\left|\frac{z+2-(z-2)}{z-2}\right|=\left|\frac{4}{z-2}\right|<\frac{4}{2+\frac{4}{\varepsilon}-2}=\frac{4}{\frac{4}{\varepsilon}}=\varepsilon
$$

Again, by definition of limit, $\Rightarrow\boxed{\therefore\lim _{z\rightarrow\infty}{\left|R(z)\right|}=\lim _{z\rightarrow\infty}{\left|\frac{2+z}{2-z}\right|}=1 \neq 0}$. Trapezoidal rule is **not** L-stable, as we claimed earlier. Q.E.D.

$\therefore$ The L-stability of both of these methods explains the qualitative behavior of our numerical solutions, as shown in the graph above; backward Euler method is L-stable, while trapezoidal rule is not.

## Problem 3

Use the following counter example to show that the trapezoidal rule is NOT B-stable.

Consider $\dot{u}=f(u)$ and 

$$f(u) = \begin{cases}
-u, & u \ge 0,\\
-u/\varepsilon, & u < 0.
\end{cases}$$

Consider the initial data to be $u_0 = 0, v_0 = -\varepsilon$.   Choose some proper $\varepsilon$ and step size $h$ to show that it is possible to have

$$|u_1-v_1| > |u_0-v_0|.$$

(Note: B-stability is applicable to any step size $h$).


**Proof:**

We start off by noting that $\left|u_{0}-v_{0}\right|=\left|0-\left(-\varepsilon\right)\right|=\varepsilon$. Also, if we want to test B-stability, $\varepsilon>0$, because B-stability is defined for contractive ODEs. Therefore, our goal is to find a stepsize $h>0$ such that, $\forall\varepsilon>0$, $\left|u_{1}-v_{1}\right|>\varepsilon\hspace{0.15cm}(1)$. I claim that, as long as $h>\frac{4\varepsilon}{1-\varepsilon}$, for $\varepsilon>0, \varepsilon\neq 1$, $(1)$ will hold, and therefore, we will find a number of counterexamples showing that the trapezoidal rule is NOT B-stable.

First off, note that, if $u_{0}=0$, then $f\left(u_{0}\right)=0$, which implies that $u=0$ is a fixed point for this ODE. Therefore, $u_{k}=\ldots=u_{0}=0\hspace{0.15cm}\forall k\geq0$. On the other hand, trapezoidal rule gives that 

$$
v_{n+1}=v_{n}+\frac{h}{2}\left(f\left(v_{n}\right)+f\left(v_{n+1}\right)\right)\Rightarrow v_{1}=v_{0}+\frac{h}{2}\left(f\left(v_{0}\right)+f\left(v_{1}\right)\right)=-\varepsilon+\frac{h}{2}\left(f\left(-\varepsilon\right)+f\left(v_{1}\right)\right) \hspace{0.5cm} (2)
$$

Given that $\varepsilon>0$, $-\varepsilon<0$, which implies that $f\left(-\varepsilon\right)=-\frac{-\varepsilon}{\varepsilon}=1$. Thus, $(2)$ becomes $v_{1}=-\varepsilon+\frac{h}{2}\left(1+f\left(v_{1}\right)\right)\hspace{0.15cm}(3)$. From here, note that the value of $f\left(v_{1}\right)$ depends on the sign of $v_{1}$. Thus, we divide our simplification of $(3)$ into two cases:

*Case 1:* Suppose that $v_{1}<0$. Then,
$$
v_{1}=\frac{h}{2}-\varepsilon+\frac{h}{2}\left(-\frac{v_{1}}{\varepsilon}\right)\Rightarrow\left(1+\frac{h}{2\varepsilon}\right)v_{1}=\frac{h}{2}-\varepsilon\Rightarrow v_{1}=\frac{h-2\varepsilon}{\frac{h}{\varepsilon}+2}
$$
Now, as a reminder, we want to find an $h$, such that $\left|v_{1}\right|>\varepsilon>0$. Thus, either $\frac{h-2\varepsilon}{\frac{h}{\varepsilon}+2}>\varepsilon$ or $\frac{h-2\varepsilon}{\frac{h}{\varepsilon}+2}<-\varepsilon$. 

In the first case, we get that $h-2\varepsilon>2\varepsilon+h\Rightarrow-2\varepsilon>2\varepsilon\Rightarrow\varepsilon<0$, which is a contradiction, because we initially assumed that $\varepsilon>0$, so this cannot be true. For the second, we have that $2\varepsilon-h>2\varepsilon+h\Rightarrow-h>h\Rightarrow h<0$, which is also a contradiction, since we are taking $h>0$. Thus, $\forall\varepsilon>0$, $\nexists h>0$ such that $\left|u_{1}-v_{1}\right|>\varepsilon$. We now find a counterexample that falls under case 2.

*Case 2:* Suppose that $v_{1}\geq 0$. Then,
$$
v_{1}=\frac{h}{2}-\varepsilon+\frac{h}{2}\left(-v_{1}\right)\Rightarrow\left(1+\frac{h}{2}\right)v_{1}=\frac{h}{2}-\varepsilon\Rightarrow v_{1}=\frac{h-2\varepsilon}{h+2}
$$
As in case 1, we again want to find an $h$, such that $\left|v_{1}\right|>\varepsilon>0$, so if this is the case, then either $\frac{h-2\varepsilon}{h+2}>\varepsilon$ or $\frac{h-2\varepsilon}{h+2}<-\varepsilon$. For the first, we have that $h-2\varepsilon>h\varepsilon+2\varepsilon\Rightarrow h>\frac{4\varepsilon}{1-\varepsilon}$, and in the second, we get that $2\varepsilon-h>h\varepsilon+2\varepsilon\Rightarrow \varepsilon<-1$. The second is a contradiction, since we have taken $\varepsilon>0$, so we are left with the first condition, as claimed earlier.

To check that this condition is sufficient, assuming the inequality on $h$ (which only holds if $v_{1}\geq0$), we note the following chain of inequalities:

$$
\left|u_{1}-v_{1}\right|=\left|0-v_{1}\right|=v_{1}=\frac{h-2\varepsilon}{h+2}>\frac{\frac{4\varepsilon}{1-\varepsilon}-2\varepsilon}{h+2}=2\varepsilon\left(\frac{1+\varepsilon}{1-\varepsilon}\right)\left(h+2\right)>2\varepsilon\left(\frac{1+\varepsilon}{1-\varepsilon}\right)\left(\frac{4\varepsilon}{1-\varepsilon}+2\right)=4\varepsilon\left(\frac{1+\varepsilon}{1-\varepsilon}\right)^{2}>4\varepsilon>\varepsilon=\left|u_{0}-v_{0}\right|
$$

Thus, $\forall h>\frac{4\varepsilon}{1-\varepsilon}$, taking $h>0$ and $\varepsilon>0$ (so, if $\frac{4\varepsilon}{1-\varepsilon}<0$, we just take some $h>0$), we have that $\left|u_{1}-v_{1}\right|>\left|u_{0}-v_{0}\right|$, as desired. Therefore, this counterexample demonstrates that trapezoidal rule is *not* B-stable, since when applied to this contractive ODE (i.e. when $\varepsilon>0$), there exists some values of $h$ such that $\left|u_{n+1}-v_{n+1}\right|>\left|u_{n}-v_{n}\right|$ for some $n$ (in this case, for $n=0$). Q.E.D.

## Problem 4

Consider two atoms with Lennard-Jones interaction

$$V(x_1,x_2)=w(|x_1-x_2|),\quad w(r)=0.01\left(\frac{1}{r^{12}}-\frac{1}{r^6}\right).$$

a) Use the 1-step (2nd order) Gauss-Legendre method to solve the Hamiltonian system

$$x_1''(t) = -\nabla_{x_1} V(x_1,x_2), \quad x_2''(t) = -\nabla_{x_2} V(x_1,x_2),$$

with initial condition

$$x_1(0) = 0.7,\quad x_2(0)=-0.7, \quad v_1(0)=v_2(0)=0.0.$$

Use time step $h=0.2$ on the time interval $[0,50.0]$. Plot the distance $x_1(t)-x_2(t)$ along the trajectory, and plot the energy along the trajectory
$$E(t)=\frac12 (x_1'(t))^2 + \frac12 (x_2'(t))^ 2 + V(x_1(t),x_2(t)).$$

b) Do the same calculation using the Velocity-Verlet method, (and observe the ease of implementation compared to implicit methods).



**Part a**

We start off by turning this system with two degrees of freedom into a four-dimensional system by defining "velocity" variables $v_{1}:=x^{'}_{1}$ and $v_{2}:=x^{'}_{2}$. Then, noting that $v^{'}_{1}=x^{''}_{1}=-\frac{\partial}{\partial x_{1}}\left[V(x_1,x_2)\right]$ and $v^{'}_{2}=x^{''}_{2}=-\frac{\partial}{\partial x_{2}}\left[V(x_1,x_2)\right]$, our system can be written in the following, equivalent form:

$$
\overrightarrow{u}^{'}:=\begin{pmatrix}x^{'}_{1} \\ x^{'}_{2} \\ v^{'}_{1} \\ v^{'}_{2} \end{pmatrix}=\begin{pmatrix}v_{1} \\ v_{2} \\ -\frac{\partial V}{\partial x_{1}} \\ -\frac{\partial V}{\partial x_{2}} \end{pmatrix}\equiv\begin{pmatrix}v_{1} \\ v_{2} \\ 0.06\left(\frac{2}{\left|x_{1}-x_{2}\right|^{13}}-\frac{1}{\left|x_{1}-x_{2}\right|^{7}}\right) \\ -0.06\left(\frac{2}{\left|x_{1}-x_{2}\right|^{13}}-\frac{1}{\left|x_{1}-x_{2}\right|^{7}}\right) \end{pmatrix}:=f\left(\overrightarrow{u}\right) \hspace{0.5cm} (1)
$$

Now, we want to apply the 2nd-order Gauss-Legendre method (GL2, or equivalently, the implicit midpoint method) for this system of first-order ODEs. What is nice about this method is how it is a symplectic integrator, so it should roughly conserve the energy (i.e. the Hamiltonian) along the trajectory. In an autonomous system such as this one, GL2 is defined as $\overrightarrow{u_{n+1}}=\overrightarrow{u_{n}}+hf\left(\frac{1}{2}\left(\overrightarrow{u_{n}}+\overrightarrow{u_{n+1}}\right)\right)$. However, since $f$ is nonlinear and not explicitly solvable, we implement a fixed-point iterative solver to approximate $\overrightarrow{u_{n+1}}$ for each timestep, halting the iterations in $i$ once the absolute error between successive iterations falls below a preset tolerance, which we set to be $10^{-12}$ in the code below. We expect that, with each increased iteration, the magnitude of this error should decrease. The fixed-point iteration looks like this:

$$
\overrightarrow{u^{i+1}_{n+1}}=\overrightarrow{u_{n}}+hf\left(\frac{1}{2}\left(\overrightarrow{u_{n}}+\overrightarrow{u^{i}_{n+1}}\right)\right) \hspace{0.5cm} (2)
$$

$f$ here is the same as defined in $(1)$. Once $\left|\overrightarrow{u^{N}_{n+1}}-\overrightarrow{u^{N-1}_{n+1}}\right|<10^{-12}$ for some value of $N$, $(2)$ implies that we should use this approximation: $\overrightarrow{u_{n+1}}\approx\overrightarrow{u_{n}}+hf\left(\frac{1}{2}\left(\overrightarrow{u_{n}}+\overrightarrow{u^{N}_{n+1}}\right)\right)$. To make an initial guess for $\overrightarrow{u^{i=0}_{n+1}}$ at each $n$th timestep, we should let $\overrightarrow{u^{0}_{n+1}}=\overrightarrow{u_{n}}$, because this is guaranteed to be at least somewhat close to $\overrightarrow{u_{n+1}}$, which means that it is most likely to allow converge when iterated. (A matter of fact, $\overrightarrow{u^{1}_{n+1}}=\overrightarrow{u_{n+1}}$ if we were to use forward Euler method, so the error in our initial guess is actually bounded.) 

Anyways, we apply this numerical scheme to the Lennard-Jones interaction, as defined above, for $h=0.2$.

```julia
using LinearAlgebra

function force(x)
    r = abs(x[1] - x[2])
    f1 = 0.01 * 6 * (2 * r ^ (-13) - r ^ (-7))
    f2 = -f1
    return [f1, f2]
end

function potential(x)
    r = abs(x[1] - x[2])
    return 0.01 * (r ^ (-12) - r ^ (-6))
end
    
function gl1(N, T, x0, v0, Niter)
    h = T / N
    u = zeros(4, N + 1)
    fu = zeros(4, N + 1)
    V = zeros(N + 1)
    u[:, 1] = [x0; v0]
    fu[:, 1] = [v0; force(x0)]
    V[1] = potential(u[:, 1])
    tol = 1e-12
    for i = 1 : N
        ui = u[:, i] #initial value at each timestep (i.e. u_{n})
        um = ui #middle value at each timestep (what is being iterated, hopefully approaching u_{n+1})
        uf = ui #final value at each timestep (our best guess at u_{n+1})
        iter = 1
        #Fixed point iteration to solve the nonlinear equation at each timestep
        for j = 1 : Niter
            uf = ui + h * [0.5 * (ui[3] + um[3]); 0.5 * (ui[4] + um[4]); force([0.5 * (ui[1] + um[1]); 0.5 * (ui[2] + um[2])])]
            if(norm(uf - um) < tol)
                break
            end
            um = uf
            iter += 1
        end
        u[:, i + 1] = uf
        V[i + 1] = potential(u[:, i + 1])
    end
    return u[1 : 2, :], u[3 : 4, :], V
end

T = 50.0
dt = 0.2
N = round(Int64, T / dt)
t = collect(0 : N) * dt
x0 = [0.7, -0.7]
v0 = [0.0, 0.0]

x, v, Vpot = gl1(N, T, x0, v0, 100)

using PyPlot
figure(1, figsize = (5, 5))
st = 1
plot(t, vec(x[1, :] - x[2, :]), "b-o")
xlabel(L"t")
ylabel(L"x_1-x_2")
title("GL1")

Et = zeros(N + 1)
Et = 0.5 * vec(v[1, :] .^ 2 + v[2, :] .^ 2) + Vpot

figure(2, figsize = (5, 5))
st = 1
plot(t, Et, "g-")
xlabel(L"t")
ylabel(L"E(t)")
title("GL1")
```

We see that the energy is roughly conserved along the trajectory, but not exactly. Actually, it appears that the deviation from energy conservation roughly seems to spike whenever the change in the displacement $x_{1}-x_{2}$ is the greatest, demonstrating how this error might depend off of some higher-order derivatives in $x_{1}(t)$ and $x_{2}(t)$.

**Part b**

An especially convenient things about the Velocity-Verlet method is that the truncation error in position is $O\left(h^{4}\right)$ (and in velocity, $O\left(h^{2}\right)$), yet the method does not require a nonlinear solver at each timestep. Although, it is important to note that the global error is still $O\left(h^{2}\right)$ in both position and velocity. Despite this, the implementation is much simpler than the above Gauss-Legendre method, which is implicit and therefore requires a nonlinear solver at each timestep for this problem, yet still having the same global error of $O\left(h^{2}\right)$.

In general, the Velocity-Verlet algorithm is the following, defining $t_{n+1}:=t_{n}+h$:

$$
\overrightarrow{x_{n+1}}=\overrightarrow{x_{n}}+h\overrightarrow{v_{n}}+\frac{h^{2}}{2}\overrightarrow{a_{n}} \\
\overrightarrow{v_{n+1}}=\overrightarrow{v_{n}}+h\left(\frac{\overrightarrow{a_{n+1}}+\overrightarrow{a_{n}}}{2}\right) \hspace{0.5cm} (1)
$$

Note we assume that $\overrightarrow{a}=\overrightarrow{a}(t,\overrightarrow{x})$, and therefore we can calculate $\overrightarrow{a_{n+1}}\equiv\overrightarrow{a}(t_{n+1},\overrightarrow{x_{n+1}})$. For the Lennard-Jones interaction, we should let $\overrightarrow{x}:=\begin{pmatrix}x_{1} \\ x_{2} \end{pmatrix}$, $\overrightarrow{v}:=\begin{pmatrix}x^{'}_{1} \\ x^{'}_{2} \end{pmatrix}:=\begin{pmatrix}v_{1} \\ v_{2} \end{pmatrix}$, and $\overrightarrow{a}:=\begin{pmatrix}x^{''}_{1} \\ x^{''}_{2} \end{pmatrix}=\begin{pmatrix}0.06\left(\frac{2}{\left|x_{1}-x_{2}\right|^{13}}-\frac{1}{\left|x_{1}-x_{2}\right|^{7}}\right) \\ -0.06\left(\frac{2}{\left|x_{1}-x_{2}\right|^{13}}-\frac{1}{\left|x_{1}-x_{2}\right|^{7}}\right) \end{pmatrix}$. Then, $(1)$ becomes:

$$
\begin{pmatrix}x^{n+1}_{1} \\ x^{n+1}_{2} \end{pmatrix}=\begin{pmatrix}x^{n}_{1} \\ x^{n}_{2} \end{pmatrix}+h\begin{pmatrix}v^{n}_{1} \\ v^{n}_{2} \end{pmatrix}+\frac{h^{2}}{2}\begin{pmatrix}0.06\left(\frac{2}{\left|x^{n}_{1}-x^{n}_{2}\right|^{13}}-\frac{1}{\left|x^{n}_{1}-x^{n}_{2}\right|^{7}}\right) \\ -0.06\left(\frac{2}{\left|x^{n}_{1}-x^{n}_{2}\right|^{13}}-\frac{1}{\left|x^{n}_{1}-x^{n}_{2}\right|^{7}}\right) \end{pmatrix} \\
\begin{pmatrix}v^{n+1}_{1} \\ v^{n+1}_{2} \end{pmatrix}=\begin{pmatrix}v^{n}_{1} \\ v^{n}_{2} \end{pmatrix}+\frac{h}{2}\begin{pmatrix}0.06\left(\frac{2}{\left|x^{n+1}_{1}-x^{n+1}_{2}\right|^{13}}-\frac{1}{\left|x^{n+1}_{1}-x^{n+1}_{2}\right|^{7}}+\frac{2}{\left|x^{n}_{1}-x^{n}_{2}\right|^{13}}-\frac{1}{\left|x^{n}_{1}-x^{n}_{2}\right|^{7}}\right) \\ -0.06\left(\frac{2}{\left|x^{n+1}_{1}-x^{n+1}_{2}\right|^{13}}-\frac{1}{\left|x^{n+1}_{1}-x^{n+1}_{2}\right|^{7}}+\frac{2}{\left|x^{n}_{1}-x^{n}_{2}\right|^{13}}-\frac{1}{\left|x^{n}_{1}-x^{n}_{2}\right|^{7}}\right) \end{pmatrix} \hspace{0.5cm} (2)
$$

$(2)$ is the scheme that we implement below, using the same parameters as for part a.

```julia
function verlet(N, T, x0, v0)
    h = T / N
    hsq = h ^ 2
    x = zeros(2, N + 1)
    v = zeros(2, N + 1)
    a = zeros(2, N + 1)
    Vpot = zeros(N + 1)
    x[:, 1] = x0 #position
    v[:, 1] = v0 #velocity
    a[:, 1] = force(x0) #acceleration, because mass has been normalized, so force=acceleration
    Vpot[1] = potential(x0)
    for i = 1 : N
        x[:, i + 1] = x[:, i] + h * v[:, i] + 0.5 * hsq * a[:, i]
        a[:, i + 1] = force(x[:, i + 1])
        v[:, i + 1] = v[:, i] + 0.5 * h * (a[:, i] + a[:, i + 1])
        Vpot[i + 1] = potential(x[:, i + 1])
    end
    return x, v, Vpot
end

T = 50.0
dt = 0.2
N = round(Int64, T / dt)
t = collect(0 : N) * dt
x0 = [0.7, -0.7]
v0 = [0.0, 0.0]

x, v, Vpot = verlet(N, T, x0, v0)

Et = zeros(N + 1)
Et = 0.5 * vec(v[1, :] .^ 2 + v[2, :] .^ 2) + Vpot

using PyPlot
figure(1, figsize = (5, 5))
st = 1
plot(t, vec(x[1, :] - x[2, :]), "b-o")
xlabel(L"t")
ylabel(L"x_1-x_2")
title("Verlet")

figure(2, figsize = (5, 5))
st = 1
plot(t, Et, "g-")
xlabel(L"t")
ylabel(L"E(t)")
title("Verlet")
```

Both of the plots of the displacement vs. time (i.e. $x_{1}-x_{2}$ vs. $t$), look nearly identical, demonstrating how our schemes show roughly the same degree of accuracy (which is probably pretty precise as well). Furthermore, the energy again looks roughly conserved, and as before, we see that the energy appears to deviate the most from something constant whenever the change in the displacement is the greatest, indicating perhaps some kind of error in the energy that is dependent off of higher-order derivatives in $\overrightarrow{x}(t)$.

Despite favorable performance of both schemes, the Velocity-Verlet algorithm was far easier to implement than the 2nd-order Gauss-Legendre method, showing one of its advantages.

## Problem 5

Solve the van der Pol equation on the interval $[0,T]$

$$
u''=\frac{1}{\varepsilon}(1-u^2)u'-u.
$$
where $\varepsilon=0.1,u(0)=2,u'(0)=0$,$T=60.0$.

Implement a 2-step Gauss-Legendre method named `gl2Newton`. Solve the nonlinear equation with Newton's method. The stopping criterion for Newton's method is the same as that for the fixed point iteration. The function  should return the solution, as well as the number of iterations for the fixed point iteration for each time step. Plot a figure with x axis being the discrete time step, and y axis being the number of iterations for each time step.

Plot the number of iterations for each time step for $h=0.01$ and $h=0.1$. How does this compare with the case when the fixed point iteration is used to solve the equation?


First off, since this is a second-order nonlinear ordinary differential equation, we transform the system into an equivalent two-dimensional system by introducing a new variable (think of this as a velocity), $v(t):=\dot{u}(t)$. Then, defining $\overrightarrow{u}:=\begin{pmatrix} u \\ v \end{pmatrix}$ and $\frac{\partial \overrightarrow{u}}{\partial t}:=\dot{\overrightarrow{u}}=\begin{pmatrix} \dot{u} \\ \dot{v} \end{pmatrix}:=f\left(t, \overrightarrow{u}\right)$, our dynamical system can be written as the following equivalent two-dimensional system:

$$
\dot{\overrightarrow{u}}=\begin{pmatrix} v \\ \frac{1}{\varepsilon}\left(1-u^{2}\right)v-u \end{pmatrix}=f\left(t, \overrightarrow{u}\right) \hspace{0.5cm} (1)
$$

Before applying GL2 (i.e. the 2-step Gauss-Legendre method), let us define a function (in Julia) which evaluates $f\left(t, \overrightarrow{u}\right)$ for given values of $\overrightarrow{u}$ and $\varepsilon$.

```julia
function vdp(u, eps)
    return [u[2], (1 / eps) * (u[2] * (1 - u[1]^2)) - u[1]]
end
```

The Julia function we defined here will be very useful later. Next, given an IVP of the form $\dot{\overrightarrow{u}}=\overrightarrow{f}\left(t, \overrightarrow{u}\right)$ with initial condition $\overrightarrow{u(0)}:=\overrightarrow{u_{0}}$, we define our GL2 scheme as the following, where $f$ is as defined above in $(1)$:

$$
\overrightarrow{k_{1}}=f\left(\overrightarrow{u_{n}}+h\left(a_{11}\overrightarrow{k_{1}}+a_{12}\overrightarrow{k_{2}}\right)\right)=f\left(\overrightarrow{u_{n}}+h\left(\frac{1}{4}\overrightarrow{k_{1}}+\left(\frac{1}{4}-\frac{\sqrt{3}}{6}\right)\overrightarrow{k_{2}}\right)\right) \\ \overrightarrow{k_{2}}=f\left(\overrightarrow{u_{n}}+h\left(a_{21}\overrightarrow{k_{1}}+a_{22}\overrightarrow{k_{2}}\right)\right)=f\left(\overrightarrow{u_{n}}+h\left(\left(\frac{1}{4}+\frac{\sqrt{3}}{6}\right)\overrightarrow{k_{1}}+\frac{1}{4}\overrightarrow{k_{2}}\right)\right) \\ \overrightarrow{u_{n+1}}=\overrightarrow{u_{n}}+h\left(b_{1}\overrightarrow{k_{1}}+b_{2}\overrightarrow{k_{2}}\right)=\overrightarrow{u_{n}}+\frac{h}{2}\left(\overrightarrow{k_{1}}+\overrightarrow{k_{2}}\right) \hspace{0.5cm} (2)
$$

Solving $(2)$ at each timestep requires solving a nonlinear system of equations for $\overrightarrow{k_{1}}$ and $\overrightarrow{k_{2}}$, which is not explicitly solvable. To remedy this, we at first define the multivariable function $F:\mathbb{R}^4\rightarrow\mathbb{R}^4$:

$$
F=F\left(\overrightarrow{y_{1}},\overrightarrow{y_{2}},f\left(\overrightarrow{u_{n}}\right)\right):=\begin{pmatrix} \overrightarrow{y_{1}}-f\left(\overrightarrow{u_{n}}+h\left(a_{11}\overrightarrow{y_{1}}+a_{12}\overrightarrow{y_{2}}\right)\right) \\ \overrightarrow{y_{2}}-f\left(\overrightarrow{u_{n}}+h\left(a_{21}\overrightarrow{y_{1}}+a_{22}\overrightarrow{y_{2}}\right)\right) \end{pmatrix}:=\begin{pmatrix} F_{1}\left(\overrightarrow{y_{1}},\overrightarrow{y_{2}},f\left(\overrightarrow{u_{n}}\right)\right) \\ F_{2}\left(\overrightarrow{y_{1}},\overrightarrow{y_{2}},f\left(\overrightarrow{u_{n}}\right)\right)\end{pmatrix} \hspace{0.5cm} (3)
$$

We have defined $F$ such that, if $\overrightarrow{y_{1}}=\overrightarrow{k_{1}}$ and $\overrightarrow{y_{2}}=\overrightarrow{k_{2}}$, then $F=F\left(\overrightarrow{k_{1}},\overrightarrow{k_{2}},f\left(\overrightarrow{u_{n}}\right)\right)=0$. Therefore, to find the roots of $F$, we can use multivariable Newton's method in $\mathbb{R}^4$, which is defined iteratively as $\overrightarrow{y}^{i+1}=\overrightarrow{y}^{i}-\left[J_{F}\left(\overrightarrow{y}^{i}\right)\right]^{-1}F\left(\overrightarrow{y}^{i}\right) \hspace{0.15cm} (4)$, where $\overrightarrow{y}=\begin{pmatrix}\overrightarrow{y_{1}} \\ \overrightarrow{y_{2}}\end{pmatrix}$ and $J_{F}\left(\overrightarrow{y}^{i}\right)$ is the Jacobian of F, or $J_{F}\left(\overrightarrow{y}^{i}\right):=\begin{pmatrix}\frac{\partial F_{1}}{\partial \overrightarrow{y_{1}}} \frac{\partial F_{1}}{\partial \overrightarrow{y_{2}}} \\ \frac{\partial F_{2}}{\partial \overrightarrow{y_{1}}} \frac{\partial F_{2}}{\partial \overrightarrow{y_{2}}}\end{pmatrix}$. (Note that, for simplicity's sake, our Jacobian is written as a $4\times4$ block matrix, where each block is a $2\times2$ matrix. Thus, each block is effectively also a Jacobian, but $2\times2$.) Given how $F_{1}$ and $F_{2}$ are defined in $(3)$, we can apply the chain rule on $f$ within $F$ for each of the partial derivatives to obtain:

$$
J_{F}\left(\overrightarrow{y}^{i}\right)=\begin{pmatrix}I-ha_{11}J_{f}\left(\overrightarrow{u_{n}}+h\left(a_{11}\overrightarrow{y_{1}}^{i}+a_{12}\overrightarrow{y_{2}}^{i}\right)\right) \hspace{0.6cm} -ha_{12}J_{f}\left(\overrightarrow{u_{n}}+h\left(a_{11}\overrightarrow{y_{1}}^{i}+a_{12}\overrightarrow{y_{2}}^{i}\right)\right) \\ -ha_{21}J_{f}\left(\overrightarrow{u_{n}}+h\left(a_{21}\overrightarrow{y_{1}}^{i}+a_{22}\overrightarrow{y_{2}}^{i}\right)\right) \hspace{0.6cm} I-ha_{22}J_{f}\left(\overrightarrow{u_{n}}+h\left(a_{21}\overrightarrow{y_{1}}^{i}+a_{22}\overrightarrow{y_{2}}^{i}\right)\right)\end{pmatrix} \hspace{0.5cm} (5)
$$

Note that $J_{f}(\cdot)$ is the Jacobian of $f$, evaluated at $\cdot$ and $I$ is the $2\times2$ identity matrix. Before we move on, we should define a function (in Julia) that will evaluate $J_{f}(\cdot)$ at $\cdot$. 

```julia
function Jacobian(u, eps)
    return [0 1 ; (- 1 / eps) * (2 * u[1] * u[2] + 1) (1 / eps) * (1 - u[1] ^ 2)]
end
```

Now, given Newton's method as outlined in $(4)$, we can solve the linear system $J_{F}\left(\overrightarrow{y}^{i}\right)\overrightarrow{\delta}^{i}=-F\left(\overrightarrow{y}^{i}\right) \hspace{0.15cm}$ at each $i$th step of our iteration for $\overrightarrow{\delta}^{i}\in\mathbb{R}^4$. If this system of equations admits a solution, then $(4)$ just becomes $\overrightarrow{y}^{i+1}=\overrightarrow{y}^{i}+\overrightarrow{\delta}^{i}$.

From here, as with fixed-point iteration, we continue iterating Newton's method until $\left|\overrightarrow{y^{N}}-\overrightarrow{y^{N-1}}\right|<10^{-12}$ for some value of $N$. Then, $(2)$ implies that we should use this approximation: $\overrightarrow{u_{n+1}}\approx\overrightarrow{u_{n}}+h\left(b_{1}\overrightarrow{y_{1}^{N}}+b_{2}\overrightarrow{y_{2}^{N}}\right)=\overrightarrow{u_{n}}+\frac{h}{2}\left(\overrightarrow{y_{1}^{N}}+\overrightarrow{y_{2}^{N}}\right)$. To make an initial guess for Newton's method, we will let $\overrightarrow{y_{1}^{i=0}}=\overrightarrow{y_{2}^{i=0}}=f\left(\overrightarrow{u_{n}},t_{n}\right)$ for each $n$th timestep. 

Note that, since we have to run this procedure at each timestep, which involves inverting a $4\times4$ matrix upon each iteration, one might initially think that fixed-point iteration is much less computationally intensive. However, we will soon discover that Newton's method converges considerably faster, especially for a larger timestep, $h$. As with problem 4 of problem set 3, we test this numerical method out for $h=0.01$.

```julia
using LinearAlgebra

function gl2Newton(N, T, u0, eps, maxIter)
    #setup and coefficients from the Butcher tableau for 2-step Gauss-Legendre method
    h = T / N
    A = [0.25 0.25 - (sqrt(3) / 6) ; 0.25 + (sqrt(3) / 6) 0.25]
    b = [0.5 0.5]
    u = zeros(2, N + 1)
    u[:, 1] = u0
    fu = zeros(2, N + 1)
    iterations = zeros(N)
    tol = 1e-12
    for i = 1 : N
        #Multivariable Newton's method to solve for k1 and k2
        iter = 1
        k = [vdp(u[:, i], eps); vdp(u[:, i], eps)]
        while iter < maxIter
            u1m = u[:, i] + h * (A[1, 1] * k[1 : 2] + A[1, 2] * k[3 : 4])
            u2m = u[:, i] + h * (A[2, 1] * k[1 : 2] + A[2, 2] * k[3 : 4])
            #We are trying to find roots for F(k), so we want it as close to 0 as possible
            Fk = [k[1 : 2] - vdp(u1m, eps); k[3 : 4] - vdp(u2m, eps)]
            if(norm(Fk) < tol)
                break
            end
            JF = [I - h * Jacobian(u1m, eps) * A[1, 1] -h * Jacobian(u1m, eps) * A[1, 2] ; 
                -h * Jacobian(u2m, eps) * A[2, 1] I - h * Jacobian(u2m, eps) * A[2, 2]]
            y = -JF \ Fk
            k = k + y
            iter += 1 
        end
        u[:, i + 1] = u[:, i] + h * (b[1] * k[1 : 2] + b[2] * k[3 : 4])
        iterations[i] = iter
    end
    return u, iterations
end 


##############################
T = 60.0
dt = 0.01
N = round(Int64, T / dt)
t = collect(0 : N) * dt
eps = 0.1
w = pi
u0 = [2.0, 0.0]

# GL 2-step Newton
ugl2Newton,Niter = gl2Newton(N, T, u0, eps, 100)

figure(1, figsize = (5, 5))
plot(t, vec((ugl2Newton[1, :])), "r-")
xlabel(L"t")
ylabel(L"u(t)")

figure(2, figsize = (5, 5))
plot(t[2 : N + 1], Niter, "b-")
xlabel(L"t")
ylabel(L"N_{iter}")
title("h=0.01", pad = 15)
```

Here, for Newton's method, the number of iterations required at each timestep are considerably more consistent than with fixed-point iteration, where they fluctuated between $7$ and $13$ steps. Note that the maximum number of timesteps occurred in either case wherever $u$ was changing the most, which is expected, as the difference between successive timesteps will be the largest in those regions. In this case, given that the number of iterations required was only slightly less than for fixed-point iteration, it might be computationally less intensive to use fixed-point iteration, because we do not have to invert a matrix at each iteration, for each timestep. Still, the methods are roughly comparable, because the matrices are pretty small.

However, when we test our numerical method for $h=0.1$ in the next case, we will see a considerable improvement.

```julia
T = 60.0
dt = 0.1
N = round(Int64, T / dt)
t = collect(0 : N) * dt
eps = 0.1
w = pi
u0 = [2.0, 0.0]

# GL 2-step Newton
ugl2Newton,Niter = gl2Newton(N, T, u0, eps, 100)

figure(1, figsize = (5, 5))
plot(t, vec((ugl2Newton[1, :])), "r-")
xlabel(L"t")
ylabel(L"u(t)")

figure(2, figsize = (5, 5))
plot(t[2 : N + 1], Niter, "b-")
xlabel(L"t")
ylabel(L"N_{iter}")
title("h=0.1", pad = 15)
```

For Newton's method, the number of iterations required at each timestep fluctuated between $6$ and $9$, and again, as expected, they peaked where the magnitude of $u$ was changing the most. However, when using fixed-point iteration on problem set 3, the number of iterations sometimes went up to over $250$ at each timestep. Thus, one can clearly see how, in cases where decreasing the stepsize $h$ is computationally expensive, it may be much more beneficial to use Newton's method over fixed-point iteration, even if the implementation was more difficult.
